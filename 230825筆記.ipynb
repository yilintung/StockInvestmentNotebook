### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】夜盤大跌，留意賣壓  
昨天盤前提到「挑戰反壓，留意要件」，加權指數要挑戰短線轉強，最基本的3要求分別為：出現4個以上的主流類股、短壓16650能否站回、成交量必須要放大，才會真的回到多方結構。而收盤前提到，成交量高達4511億屬於爆量，既然爆量就不能留長上影線，否則未來這個位置就會變成短線的套牢量。  
收盤前也曾提到，若加權指數低於16708，就算長上影線，白天雖然漲幅有193點，收在16770的位置，但夜盤期指隨即轉弱大跌213點，如果將白天和夜盤合併觀察，算是有出現長上影線。  
本周前4個交易日，雖然加權指數都稍微偏多，但櫃買指數卻是開高走低的走勢，明顯有拉大出小的現象。拉大出小是指指力作手會趁市場氛圍樂觀，加權指數當天表現強勢，藉此掩護個股出貨現象。因此如果持有的是中小型股居多，那應該同步關注櫃買指數，避免加權指數帶來的誤判。  
因此雖然表面指數強漲，但不管從國際股市、類股結構、籌碼面…，都還沒有完全符合轉強翻多的條件，這一兩天盤是真的是峰迴路轉，感覺弱勢盤整，卻突然拉一根長紅，但夜盤期指卻又大幅修正，沒意外的話昨天的大量，反而會成為短線壓力，如果判斷力不足，昨天就貿然積極進場，容易有短套的風險。  
所以又回這個關鍵的16650位置，仍是關鍵的壓力區，今天基本上沒意外的話，會吃掉昨天的漲幅，而是否宣告昨天是假突破，就要看今天或後幾日，有沒有跌破16650。基本上今天就觀察上面的重點即可，其他內容以其後市的看法，等待周末影音解盤，再作完整補充。 (出處：K線捕手楊忠憲粉絲團)
