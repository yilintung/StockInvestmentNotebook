用停損做模擬，當使用10萬本金進行交易時，10%的虧損只有1萬，但是當本金放大到100萬時，雖然停損一樣是10%，但是絕對金額已經來到10萬。停損1萬元可能不痛不癢，但是停損10萬，就需要一點訓練，更遑論當資金部位放大到1千萬時，10%停損會高達100萬，是很多人1-2年的薪水，要能紀律停損，必須有相當的智慧與勇氣。  
要如何停損1萬到10萬，甚至到達百萬，都能維持紀律，要把握4個重點：交易次數累積、停損技巧鍛鍊、提高選股勝率、漸進式放大。首先，我認為至少要有100檔以上的交易經驗，才能有紮實的經驗值，所謂一檔，是指試單、加碼、買足，甚至中間經過調節買賣，最後全部出清，這樣的過程只能算是一檔，當然能夠超過100檔更好，次數多多多益善。第二是停損的技巧，上過課同學就知道有8種，而8種之中能用於實戰的只有5種，必須要完全清楚使用的時機和要點。  
第三是選股勝率，在國際股市多頭和台股指數多頭時，選股勝率要在7成以上，如果未達標準，選股技巧可能要再精進。最後則是漸進式的擴大資金，假如一開始用10萬元本金交易，操作一段時間後相當熟練，可以放大資金到30萬，一段時間之後再放大到50萬，再來是80萬，最後才是100萬，不要一次性的從10萬放大到100萬。10萬本金的心理素質，不見得能應付100萬資金的變化。
