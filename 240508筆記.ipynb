### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線震盪，留意壓力  
加權指數昨天雖然仍然上漲了130點，但由於距離前高20883不到300點，盤中開始出現賣壓，震盪相當劇烈，與周末影音解盤提醒的一樣，越接近前高要越保守。指數反彈超過10個交易日，進入中期多空關鍵，大家更關心盤勢究竟有機會突破前高，中期趨勢由相形轉為多方？或是遇壓再度回測支撐？  
繼前兩周就已經翻空的重電和電纜族群，昨天營建和電動車族群也開始轉弱現象。多方要續漲，波段最強的類股轉弱後，必須出現新主流遞補，以目前類股結構來說，34大類股中，比例最多的與指數一樣，有22類股屬於短線反彈中期箱型，多方類股大約只有6類，真正要翻多至少要有10類轉強，所以從類股結構還看不出續攻訊號。  
其他挑戰20883前高的要件包含：價量關係要在10日均量之上，增加到4500億是最低標準；期貨空單口數降低到1萬口以下；現貨籌碼，法人買超增幅持續擴大；台幣升值，20日線下彎。和昨天一樣，上述條件只有法人籌碼買盤增加，這個籌碼條件改善，其餘多方條件都還沒有出現。  
既然已經進入壓力區，密切觀察不必過度躁進，靜待趨勢明朗即可，持股比重高者可以部份先獲利了結，做預防性減碼。另外要多加入一個觀察，在多頭時永遠是利多測頭部，正常情況利多上漲、利多上漲、利多上漲…，但終有一天會出現利多不漲，這個時候更要留意止漲訊號，慎防接著出現空方轉折訊號。 (出處：K線捕手楊忠憲粉絲團)
