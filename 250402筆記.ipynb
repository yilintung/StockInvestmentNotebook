📊 在股票交易的世界裡，許多人常常會問：「老師，你是不是都做短線交易？」 其實，這並非個人的選擇，而是市場決定的！我的交易模式以技術分析為基礎，當市場給出買賣訊號時，必須果斷執行，因此短線操作成為必然。然而，這並不代表這種模式只適用於我，而是所有想要透過波段或短線賺取價差的投資人，都必須遵循的交易原則。換句話說，股票的走勢決定了交易的長短，而不是投資人單方面的意願。  
【許多人的「長線投資」其實是被迫的】  
🔍 大多數散戶的投資模式往往是這樣的：  
📢 尋找「明牌」：透過網路、電視、雜誌，甚至是親友推薦來決定買哪支股票。  
🤔 憑感覺進場：沒有經過精確統計與驗證，只是覺得這支股票「好像不錯」，就貿然買入。  
💤 買進後「養著」：不關心技術指標、型態、籌碼或基本面，只是每天看著價格上下波動。  
📺 聽消息操作：依賴新聞報導或所謂「有內線消息」的人來決定下一步該怎麼做。  
📉 如果股票漲了，這類投資人可能會因為賺錢而賣出，但如果跌了呢？大多數人選擇「養著」，而且還會安慰自己：「沒關係，長期持有就好了。」更嚴重的是，當股價持續下跌，部分人會選擇「向下攤平」，從小套房變成大套房，甚至演變成總統套房！

💡 很多人原本的計畫是短線操作，想要投機賺取價差，結果卻因為不懂得何時該買、何時該賣，最終被迫變成長期持有。他們告訴自己：「沒關係，我做的是長線投資。」但事實上，這樣的長期持有，並非基於合理的投資策略，而是無奈之舉。  
📌 長期下來，這種不良習慣讓投資組合內的股票越買越多，最終變成一籃子股票，而投資人自己卻不知道該如何應對，索性選擇忽視，甚至關掉電腦，不願面對虧損。然而，這樣的方式真的能讓你獲利嗎？  
【真正的長線投資，是有策略的】  
📈 真正的長線投資，應該基於企業的基本面與未來成長潛力，而不是因為「不想賣」或「不知如何賣」才被迫持有。因此，無論你是短線交易者還是長線投資者，都應該建立明確的交易策略，遵守買賣規則，而不是讓股票的價格變動來主導你的決策。

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import mplfinance as mpf
import numpy as np
import myutils
```

```python
def date_to_index(df_in,date_in) :
    number_array = df_in.index == date_in
    idx = 0
    for number_index in number_array :
        if number_index == True :
            break
        idx = idx + 1
    return idx
```

```python
##### 使用2024年7月22日到今日(2025年4月2日)收盤之日線圖資料 #####

# 載入從「好神通PLUS」輸出的Excel檔
df = pd.read_excel('./data/SINO_TSE_250402.xlsx')

# 保存K棒的基本資訊(開、高、低、收、量)
df_k_line = df.drop(columns=['SMA5','SMA10','SMA20','SMA60','SMA120','SMA240','MA5','MA10'])

# 保存價的均線
df_sma = df.drop(columns=['開盤價','最高價','最低價','收盤價','成交量','MA5','MA10']) 

# 將K棒的Columns的名稱由中文改為英文
df_k_line = df_k_line.rename(columns={'時間':'Date','開盤價':'Open','最高價':'High','最低價':'Low','收盤價':'Close','成交量':'Volume'})

# 將價的均線的Columns的名稱由中文改為英文
df_sma = df_sma.rename(columns={'時間':'Date'})

# 將K棒的Date設為Index
df_k_line.set_index(df_k_line['Date'],inplace=True)
df_k_line = df_k_line.drop(columns=['Date'])

# 將價的均線的Date設為Index
df_sma.set_index(df_sma['Date'],inplace=True)
df_sma = df_sma.drop(columns=['Date'])
```

```python
### 基準：股票老師所繪製的線條 ###

# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定支撐線
support_line_start_date  = '2024-08-02'
support_line_end_date    = df_k_line.iloc[-1].name.strftime("%Y-%m-%d")
support_line_x           = [date_to_index(df_k_line,'2024-08-08'),date_to_index(df_k_line,'2024-09-10'),date_to_index(df_k_line,'2024-09-11')]
support_line_y           = [df_k_line.loc['2024-08-08']['Open'],df_k_line.loc['2024-09-10']['Close'],df_k_line.loc['2024-09-11']['Close']]
slope,intercept          = np.polyfit(support_line_x,support_line_y,1)
support_line_start_price = slope * date_to_index(df_k_line,support_line_start_date) + intercept
support_line_end_price   = slope * date_to_index(df_k_line,support_line_end_date) + intercept

seq_of_seq_of_points=[
    [(support_line_start_date,support_line_start_price),(support_line_end_date,support_line_end_price)]
]

# 設定移動平均線
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.6,color='xkcd:maroon'),
    "SMA10": mpf.make_addplot(df_sma['SMA10'],width=0.6,color='xkcd:cyan'),
    "SMA20": mpf.make_addplot(df_sma['SMA20'],width=0.6,color='xkcd:violet'),
    "SMA60": mpf.make_addplot(df_sma['SMA60'],width=0.6,color='xkcd:salmon'),
    "SMA120": mpf.make_addplot(df_sma['SMA120'],width=0.6,color='xkcd:blue grey'),
    "SMA240": mpf.make_addplot(df_sma['SMA240'],width=0.6,color='xkcd:sky blue')
            }

# 繪製K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=2.0, colors='xkcd:bright blue', alpha=0.6), volume=True, datetime_format='%Y-%m-%d')
mpf.plot(df_k_line,**kwargs)
```

```python
from PIL import Image

def get_first_last(mask, axis: int):
    """ Find the first and last index of non-zero values along an axis in `mask` """
    mask_axis = np.argmax(mask, axis=axis) > 0
    a = np.argmax(mask_axis)
    b = len(mask_axis) - np.argmax(mask_axis[::-1])
    return int(a), int(b)

def crop_borders(img, crop_color):
    np_img = np.array(img)
    mask = (np_img != crop_color)[..., 0]  # compute a mask
    x0, x1 = get_first_last(mask, 0)       # find boundaries along x axis
    y0, y1 = get_first_last(mask, 1)       # find boundaries along y axis
    return img.crop((x0, y0, x1, y1))
```
