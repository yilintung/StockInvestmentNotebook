### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】持續偏空，做多不易  
雖然前天上漲107點，但有提醒投資人「單日偏多，要件不齊」，因此整體來說，只能算是多方抵抗，反攻要件不具備，還稱不上轉強或翻多訊號，上漲只是短期現象，「隨時可能回吐漲幅」。  
加權指數昨天隨即開低走低，終場下跌176點，成交量2397億，收在16276點，紮實的吃掉前一日漲幅，加上台幣續貶，再度壓回16400之下。短中期不論籌碼面或技術面持續偏空，櫃買指數也相仿的現象，因此不論各大類股或個股，普遍都表現弱勢。  
很多新手都會有一種迷思，不用管國際股市好壞，加權指數以及上櫃指數指數的多空強弱和位階，只要挑對好的股票就可以做多，交易時不用管指數多空強弱和位階，只要挑對好的股票就可以做多，稱之為「選股不選市」。
認為只要挑到強勢股，就是天天all in，不要管甚麼國際股是跟大盤，但是觀察近期的盤勢，應該會很有體會，選股一定也要選市，因為當趨勢翻空，覆巢之下無完卵。這也是為什麼每天投資不是直接看個股，一定是先看國際股市以及大盤、類股。  
因為個股再強，強不過類股；類股再強，強不過大盤；大盤再強，強不過國際股市。選股一定要先選市(勢)，不然市(勢)不對，還硬要逆勢操做，那就會很辛苦了。大盤的趨勢多空強弱，會影響究竟要做多、做空，還有資金比重的調節。  
經過昨天續弱走勢，歐美股市大多已經做頭，而亞股中除日本之外，也都已經做頭，中期空方趨勢越來越強，因此未來續跌的機率相當高，加上昨天夜盤期指續弱，今天就會來測試上周五前低支撐，加上再過2個交易日就放假，多看少做不要過度交易。 (歡迎分享，請載明出處K線捕手楊忠憲粉絲團)

許多分析家和經紀人在評價有價證券時，將之區分成投機型和保守型，我認為這種區分過於膚淺，甚至是錯誤的。重點不在大家投資股票的品質，而是投資的數量，如果一位大資本家用少量的錢，購買一種相對沒有把握的科技股，那不算是投機，而是計算過風險的保守投資。但如果是一個小人物用超過自己資產的金額，也就是透過融資購買「最安全的」標準股，才是非常危險的投機。大家必須時刻牢記，投機型和保守型的區別，只是比例上的問題。

同樣，我認為分析家經常說的「繼續持有」建議，完全是胡鬧。如果證券帳戶，放著我已不會再買進的股票，那我便必須賣掉，有什麼理由要保留顯然已不會再上漲的股票？經紀人犯了一個幾乎每個證券交易者都會犯的錯誤，即以自己買股票的角度來評價股票。如果他以一百元的價格買了一種股票，等漲到二百元時，他就認為股票價格貴了；如果跌到五十元，他又會覺得股票價格便宜了。股票價格是貴還是便宜，取決於基本數據和企業的發展前景。按這種標準，投機人士必須盡可能客觀評價一種股票，至於是什麼時候進場的，並沒有太大作用，證券市場不會顧及這一點。即使已經上漲1000%的股票仍可能值得購買，而已跌了80%的股票可能還是不值得買進。
