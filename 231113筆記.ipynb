【不要期待指數上漲會拯救個股】  
研究國際指數跟大盤，目的是決定要做多還是做空、判斷風險高低、決定投入的資金比重，但有一些投資人是這樣想的，觀察指數只是希望自己手中套牢和虧損個股，會不會跟著上漲？也就是期待指數上漲能拯救套牢和虧損的個股，但這樣是很危險的思考路徑。  
加權指數或是櫃買指數本質上，是所有上市櫃股票漲跌表現的集合體，一般情況，當指數表現好的時候，大多數個股表現也不會太差，但有些時候指數結構出現問題的時候，可能就不見得如此。比方現在32大類股中，維持盤整的共計9類。持續弱勢偏空的有4類。中期偏空短線反彈的有14類。偏多方的僅有紡織纖維、營建、網通、電子通路、居家生活，共計4類。  
指數是呈現所有個股的總體表現，但是個別股仍然會有自己的技術線型跟籌碼變化，因此上市櫃近2千檔股票，會有多、有空、有強、有弱，不能一概而論。所以即使指數止穩，甚至轉強了，但持股倘若中期翻空甚至做頭，而且籌碼弱勢，完全看不到支撐，即使指數反彈、轉強，也救不了你的個股，所以還是要理性客觀看待手中期股，並且擬訂明確交易策略，不要過度期待指數上漲會拯救你的空頭弱勢股。
