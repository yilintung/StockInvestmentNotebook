【[究竟盤中看幾次，比較剛好呢？](https://www.facebook.com/kevinyang20180801/posts/pfbid0k4QgqRMfNj1n29kQMMcMoGvmc8jdZN2Hjem1DUE6SSTvi4x4ZzQq14cMNLfRysoKl)】  
有很多同學問過我，我會一直盯盤嗎？或是應該看幾次盤比較剛好？  
我發現很多同學，看盤非常頻繁，其實凡事過猶不及，完全都不關心，當然是不好，可是關心過度，也會變成適合其反，很多人明明自認為是做長線的，卻盯得好像做當沖一樣，又或者明明做短線交易，卻沒有看盤習慣，這都是很危險的。  
但究竟看幾次盤才比較剛好呢？說真的，肯定是沒有標準答案的，因為會牽涉到交易周期長短，做得越短必然盯的越兇，所以以下僅就我個人操作習慣，作為模式的參考。  
我個人以做現股為主，沒有衍生性商品或是做空，當持股比重高時，一天最多看個5次，當持股比重低時，一天看2-3次。先以持股比重低做模擬，首先是第一次是盤前到開盤後10分鐘，其實盤前大約期貨開盤時到開盤後10分鐘關注一下，當天走勢心裡就有底，這段時間也是最重要的，其次當然是收盤，大概收盤前10分鐘關注一下，驗證今天看法。  
而且看盤也是幾分鐘，把重點瀏覽一下，就結束了。如果你持股很低，基本上就是上述的這兩次，其實就夠了，最多再加一次中午11-12點。但如果持股比重很高，盤勢波動又很大，那每間隔1小時看一次，其實也就足夠，至於盤中看甚麼以及細節，課程和我的專書都有寫，就不贅述。
