### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】中期空方，測試支撐  
昨天盤前曾提醒，「加權指數中期趨勢，關鍵支撐頸線瀕臨跌破，一旦有效跌破，中期大箱型區間是否跌破，一旦支撐失守，會形成新的中期頭部，出現新一波續跌，但如果依照國際股市跟櫃買指數表現，跌破機率相對高。」  
加權指數昨天盤中跌點逼近300點，終場收斂跌勢，但仍下跌207點，成交量3295億，技術面出現中期頭部現象，外資、融資、主力指標均站在賣方，比較值得注意的是投信連續兩周，持續逢低承接。  
昨除了跌勢較大之外，原先提到的中長期空方續跌的輔助條件，觀察台幣32.18關鍵支撐是否貶破、石油是否持續上漲、外資未平倉空單是否超過一萬口，昨天外資未平倉空單口數首度攀升到11208口。  
本波段領跌的櫃買指數，昨天也出現2.29%的跌幅，相較於加權指數跌幅更大，不過由於櫃買指數20日夫乖離已經來到6.15%，而且已經接近前波支撐和黃金分割率的0.382，短線會比較容易有支撐，甚至出現反彈。  
短中期空方基本上已經確認，過程中或有機會小漲或反彈，不過不影響空方趨勢，但持股比重高的投資人，還是會想問是否有由空翻多的機會？基本上最少要有兩個條件，類股結構是否能出現4個以上的主流類股，以及16650能否站回，可以自行做觀察。 (出處：K線捕手楊忠憲粉絲團)
