```python
import numpy as np
import mysys
import markdown
import base64
from IPython.core.display import HTML
```

```python
def stock_analysis( analysis, stock_id, buttom_pattern = False) :
    results,images = analysis.analysis(stock_id)
    if buttom_pattern is True and '底部型態' in images :
        print('底型反轉交易策略：')
        display(images['底部型態'])
    print('解盤內容：')
    if '整體評價' in images :
        display(images['整體評價'])
    result_md   = results.to_markdown(tablefmt="grid")
    result_html = markdown.markdown(result_md, extensions=['markdown_grid_tables:GridTableExtension'])
    display(HTML(result_html))
```

```python
# 支撐線
def line_drawing_callback( range_prices) :

    seq_of_seq_of_points = []
    linewidths           = []
    colors               = []
    
    # 設定上升趨勢線
    support_line_start_date  = '2025-04-01'
    support_line_end_date    = range_prices.iloc[-1].name.strftime("%Y-%m-%d")
    support_line_x           = [mysys.DateToIndex(range_prices,'2025-04-22'),mysys.DateToIndex(range_prices,'2025-08-21')]
    support_line_y           = [range_prices.loc['2025-04-22']['Close'],range_prices.loc['2025-08-21']['Open'] + ((range_prices.loc['2025-08-21']['Close'] - range_prices.loc['2025-08-21']['Open'])/2)]
    slope,intercept          = np.polyfit(support_line_x,support_line_y,1)
    support_line_start_price = slope * mysys.DateToIndex(range_prices,support_line_start_date) + intercept
    support_line_end_price   = slope * mysys.DateToIndex(range_prices,support_line_end_date) + intercept
    seq_of_seq_of_points.append([(support_line_start_date,support_line_start_price),(support_line_end_date,support_line_end_price)])
    linewidths.append(1.5)
    colors.append('xkcd:electric blue')

    # 設定上升趨勢線
    support_line_start_date  = '2025-04-21'
    support_line_end_date    = range_prices.iloc[-1].name.strftime("%Y-%m-%d")
    support_line_x           = [mysys.DateToIndex(range_prices,'2025-06-02'),mysys.DateToIndex(range_prices,'2025-08-21')]
    support_line_y           = [range_prices.loc['2025-06-02']['Close'],range_prices.loc['2025-08-21']['Open'] + ((range_prices.loc['2025-08-21']['Close'] - range_prices.loc['2025-08-21']['Open'])/2)]
    slope,intercept          = np.polyfit(support_line_x,support_line_y,1)
    support_line_start_price = slope * mysys.DateToIndex(range_prices,support_line_start_date) + intercept
    support_line_end_price   = slope * mysys.DateToIndex(range_prices,support_line_end_date) + intercept
    seq_of_seq_of_points.append([(support_line_start_date,support_line_start_price),(support_line_end_date,support_line_end_price)])
    linewidths.append(1.5)
    colors.append('xkcd:red')

    colors.append('xkcd:white')
    
    return seq_of_seq_of_points,linewidths,colors
```

```python
# 盤前：支撐線繪製
image = mysys.DrawOnKlineChart('TAIEX','2025-02-27','2025-08-21',line_drawing_callback)
display(image)
```

```python
# 盤後：支撐線繪製
image = mysys.DrawOnKlineChart('TAIEX','2025-02-27','2025-08-22',line_drawing_callback)
display(image)
```
