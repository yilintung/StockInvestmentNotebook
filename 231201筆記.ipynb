### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線偏多，留意背離  
昨天台股與國際股市大多是小紅小黑，整體變化不大，唯一波動比較劇烈的是晚上的美股道瓊指數，大漲520點，創波段新高，不過夜盤期指並沒有對應表現，反而小跌48點。  
雖然整體架構都還維持高檔整理，除了之前提到的持續關注日本股市(尚未有效過高)和投信籌碼(持續偏空)之外，今天想深入討論一個重要的現象-高檔背離，因為背離趨勢反轉的領先工具，很容易演變成多頭力竭的現象。  
背離的風險性分成幾種層級，第一種最常見風險也最小，就是單指標一次背離，第二種反轉風險會上升一些，是單指標二次背離，第三種風險係數又更高，是多指標一次背離，最後一種最危險，是多指標二次背離。  
目前加權指數只有價量關係出現二次背離，其他指標還算正常，但KD指標有可能先出現一次背離。櫃買指數則是僅有價量關係的一次背離。必須持續追蹤背離現象，是否有往高風險的方向前進。  
背離發生後只要沒有出現轉折現象，趨勢就會持續走多，因為背離只是警訊而非賣訊，而且背離可能2次、甚至3次，只是背離越多次，後續風險越高。這是近期的觀察重點，如果背離風險有增加，或是出現轉折現象，會適時提醒大家。 (出處：K線捕手楊忠憲粉絲團)

如果常常看我解盤，會看到持股資金比重需要調節，有時候是3成、有時候是5成，有時候大多頭行情又沒風險，就可以滿持股。但必須先說，這不是我的發明，而是法人、大戶也都是如此，他們會依照外資風險高低，透過控制資金比重調節，達到風險控管。  
於是有同學問了一個有趣的問題，可否不需要資金調節，永遠滿持股？(以下指的是作價差，長期投資那是另一個宇宙，不在我們討論範圍)我說當然可以，假如你可以完全的紀律，沒有絲毫的人性弱點跟猶豫。  
例如：某人可用資金一千萬，一年365天永遠持股一千萬，某天金融市場突發性系統性風險出現，有辦法一天砍光持股，一千萬幾天內瞬間虧100-200萬也承受得起(這是最保守的假設)，那這樣的情況可以不做調節，或是你資金本來就低，比如低於50萬以下，那也就沒有甚麼好調節或是分割。  
因為50萬再做分割，獲利金額和能買的商品就相對有限，而50萬即使盤勢突然遽變，瞬間慘賠50%，也不過25萬，雖然心情還是很沉重，但個人或家庭經濟不至於受到太大影響，還可以承受得起。
