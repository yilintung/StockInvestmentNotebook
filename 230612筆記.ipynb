【[整數關卡，不是實質支撐或壓力](https://tinyurl.com/2fpnsn27)】  
當指數行進到16999，就會有同學開始問：17000點會不會有壓力？  
其實整數關卡並非實質的壓力或支撐，只是一個市場上的共同心理，否則如果要照這個邏輯推演，那從12629漲上來，13000點、14000點、15000點、16000點…，或是分割更細一點13500、14500、15500、16500，也都要有支撐或壓力，那這樣盯盤會很累。  
所以，除非這個整數關卡，剛好也是技術面的支撐或壓力，比如前波高低點、大量區間、頭部或底部頸線、黃金分割率、中長天期均線…。你是不是又想問，那17000有沒有剛好重疊的支撐或壓力，其實還真的有，但我說過因為指數位階太高，會出現鐘擺效應，即便真有支撐或壓力，也要放寬到百點以上的誤差。  
所以這樣回頭來說，在這個位階和時間點，整數關卡就更加沒有太大的意義，所以不預設高點，守好原先規劃的轉折現象，也就周末影音直播提到的，支撐跌破法會優於目標滿足法，這樣可能反而比較單純。
