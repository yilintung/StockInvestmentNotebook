### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】加權盤整，櫃買創高  
昨天盤前提到，短天期狹幅盤整，除非震盪區間擴大，天數不會太久，近日就有機會出現較明朗走勢，隨即出現大漲168點，脫離狹幅盤整，來到17084，成交量3320億，約等於10日均量的3326億，算是溫和放大，但又還沒大幅高過10日均量，所以量能只能算是及格，所以僅就這個條件，就判斷短線翻多，還是言之過早，因為有可能從狹幅進入擴箱，所謂擴箱就是盤整區間擴大，從狹幅盤整轉為箱型整理。  
所以不能只難一天上漲，就判斷趨勢改變，還需要搭配其他工具做綜合研判。先觀察多方條件，單日價漲量增轉強之外，加權指數站上20日線且上揚，加上扣抵仍低，所以中期支撐穩健，而創高的多方類股有紡織纖維、電機機械、鋼鐵、電腦周邊，共有4大類，周KD也維持在高檔鈍化區，尚未死亡交叉，6月8日前低支撐也守穩，並沒有跌破，歐美股市屬於偏多方的整理。  
而空方條件尚未站上10日線，且10日線仍下彎，而空的類股有塑膠、電器電纜、化學、生技、醫療、橡膠、觀光、油電燃氣、網通、電子通路，相較之下明顯比多方類股要多出許多，共有10類。外資未平倉多單3467口，雖然仍然偏多，但相較6月多單驟減。台幣自從6月20日正式貶破30.89支撐，中期持續趨貶。相較之下，空方條件比多方條件，明顯多出許多。  
因此，雖然短線技術面在20日線出現轉強，但是其他條件尚未同步跟上，要立刻挑戰前高17346壓力，條件還不夠充分，上述空方條件尚未改善之前，維持在17346壓力與月線之間做整理，是對多方比較有利的走勢。不過有一點比較特別，就是加權指數雖然還在區間整理，但昨天盤前提到，櫃買指數表現相對強勢，昨天也已經創高，不少中小型個股仍然表現活潑，電機機械、營建、電腦周邊、網通、電子零組件，都維持創高續漲。 (出處：K線捕手楊忠憲粉絲團)
