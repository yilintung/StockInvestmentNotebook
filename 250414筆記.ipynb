```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import mplfinance as mpf
import numpy as np
import myutils
import sqlite3
import requests

from dotenv import load_dotenv, find_dotenv
from FinMind.data import DataLoader
```

```python
load_dotenv(find_dotenv())
token = os.environ.get('FINMIND_TOKEN')
api = DataLoader()
api.login_by_token(api_token=token)
```

```python
# 台股總覽 TaiwanStockInfo
df = api.taiwan_stock_info()
display(df)
```

```python
# 股價日成交資訊 TaiwanStockPrice：一次拿特定日期，所有資料(只限 backer、sponsor 使用)
df = api.taiwan_stock_daily(
    start_date='2025-04-13',
)
display(df)
print(type(df))
print(df.empty)
```

```python
# 取得今天資料
df = api.taiwan_stock_daily(
    start_date='2025-04-14',
)
display(df)
```

```python
### 參照「230203筆記」進行轉換、存檔與驗證 ###
df_daily_price = df.drop(columns=['Trading_money','spread','Trading_turnover'])
df_daily_price = df_daily_price.rename(columns={'date':'Date','stock_id':'StockID','Trading_Volume':'Volume','open':'Open','max':'High','min':'Low','close':'Close'})
# 保存格式：日期、股票代碼、開盤價、最高價、最低價、收盤價與成交量
df_daily_price = df_daily_price[['Date', 'StockID', 'Open', 'High', 'Low', 'Close', 'Volume']]
display(df_daily_price.head())
```

```python
# 檢查：DataFrame轉換前後的資料確認
is_correct = True
for index in range(0,len(df_daily_price)) :
    if df_daily_price.iloc[index]['Date'] != df.iloc[index]['date'] :
        is_correct = False
        break
    if df_daily_price.iloc[index]['StockID'] != df.iloc[index]['stock_id'] :
        is_correct = False
        break
    if df_daily_price.iloc[index]['Open'] != df.iloc[index]['open'] :
        is_correct = False
        break
    if df_daily_price.iloc[index]['High'] != df.iloc[index]['max'] :
        is_correct = False
        break
    if df_daily_price.iloc[index]['Low'] != df.iloc[index]['min'] :
        is_correct = False
        break
    if df_daily_price.iloc[index]['Close'] != df.iloc[index]['close'] :
        is_correct = False
        break
    if df_daily_price.iloc[index]['Volume'] != df.iloc[index]['Trading_Volume'] :
        is_correct = False
        break
# 如果正確則印出True
print(is_correct)
```

```python
# 測試：保存至SQLite資料庫 #

# 建立資料庫
conn = sqlite3.connect('data/TaiwanStocks.db')

# 如果資料表存在，就寫入資料，否則建立資料表
df_daily_price.to_sql('DailyPrice', conn, if_exists='append', index=False)

# 關閉資料庫
conn.close()
```

```python
# 測試：讀取SQLite資料庫 #

# 建立資料庫
conn = sqlite3.connect('data/TaiwanStocks.db')

# 確認指定日期是否存在於資料庫中？
cursor = conn.cursor()
cursor.execute("SELECT * FROM DailyPrice WHERE Date='2025-04-13'")
db_result=cursor.fetchone()
# 如果回傳為None，則代表該日期並不存在於資料庫中
print(db_result is None)
cursor.execute("SELECT * FROM DailyPrice WHERE Date='2025-04-14'")
db_result=cursor.fetchone()
print(db_result is None)

# 依"日期"讀取資料
price_db_df = pd.read_sql("SELECT * FROM DailyPrice WHERE Date='2025-04-14'", conn)
display(price_db_df.head())
# 從該日期的資料中找出對應股票代號的資料
display(price_db_df[price_db_df['StockID'] == '2454'])

# 依"股票代碼"讀取資料
price_db_df = pd.read_sql("SELECT * FROM DailyPrice WHERE StockID='2330'", conn)
display(price_db_df)

# 關閉資料庫
conn.close()
```
