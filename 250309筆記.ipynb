📈 長線投資的風險  
  
🏭 台塑 (1301) 案例分析  
自 2020 年 9 月以來，台塑股價一路從 121 元下跌，走勢進入空頭市場。這種情況下，長期投資仍然是正確的選擇嗎？假設某投資者從高點開始，逢低攤平，是否為正確策略？最終，台塑股價最低跌至 31.25 元，即使這樣，其平均持有成本仍高達 75 元。即使目前價格回升至 38.65 元，投資者仍然處於虧損狀態。  
  
🔍 值得思考的是：  
  
📉 38.65 元的低點會維持多久？  
🛠️ 投資者是否有足夠的勇氣在下跌時持續買進？  
這些問題都值得深思。  
  
🔄 攤平 vs. 🚀 反彈操作  
有同學詢問：「攤平與反彈操作有何不同？」  
  
這兩種方式本質上都是逆勢交易，但風險與策略大不相同。  
  
📌 攤平：  
  
當股價下跌時持續買進。  
❌ 若股價持續下跌，資金部位將加重，導致更難解套。  
⚠️ 這種策略在空頭市場中通常是錯誤且不可行的。  
📌 反彈操作：  
  
在空頭市場中，股價偶爾會出現短線小幅反彈。  
⏳ 若未及時撤出，將面臨更大的風險。  
🚀 反彈並非市場反轉的訊號，因此應短線操作。  
⛔ 停損策略的重要性  
有位同學持有 88 元買進的台塑股票，不確定是否該停損。我曾經說過：當股票虧損超過 20% 時應果斷處理，因為超過這個幅度後，解套的難度將大幅增加。  
  
例如，當股價從 90 元跌至 72 元時，投資者應該選擇停損還是繼續持有？  
  
🔹 這是一個兩難的抉擇，因為：  
  
📉 股價已經大幅下跌，可能隨時反彈。  
🛑 但若不果斷停損，長期來看可能會跌得更低。  
🏃‍♂️ 長期持有 vs. 🦤 鴕鳥心態  
是否應該對持股置之不理？  
  
長期持有但不處理，實際上只是 「鴕鳥心態」。  
  
⚠️ 如果持有十年、二十年甚至三十年，股價真的會回升嗎？  
⚠️ 若股價跌破 31.25 元仍不停損，最終會如何？  
⚠️ 當市場反彈時，應該在哪個點位賣出才算是「逃命」？  
  
許多股票的走勢會讓投資者陷入進退兩難的境地。如果選擇停損，會擔心是否砍在最低點；如果想要減少虧損，卻又無法確定何時賣出才是最佳時機。  
  
📊 「九張加一張」策略解析  
有同學提出 「九張加一張」 的策略：  
  
📌 假設投資者在 88 元套牢 1 張，然後 在 36 元低點一次性回補 9 張。  
📌 這樣一來，股價只要回升至 41.2 元，即可實現盈虧平衡。  
  
❗ 但這種策略的前提是：  
  
🔢 投資者 必須精確計算持股數量與潛在風險。  
⚠️ 若股價持續破底或長期盤整，仍然可能陷入虧損。  
本質上，這仍然是一種 「攤平」，若市場趨勢不變，風險依舊存在。  
  
✅ 最佳策略：小虧換大賺  
⛔ 不設定停損機制，最終往往會導致投資悲劇。  
  
💡 最好的策略是：  
  
💰 犧牲小額虧損，避免未來大賠。  
📈 在股價反彈至壓力區時獲利了結，將資金轉移至其他更具潛力的標的。  
❓ 然而，每位投資者是否能做到這樣的操作？  
❓ 如果股價運行過程中發生劇變，能否及時調整策略？  
❓ 是否能克服人性的弱點，避免盲目攤平？  
  
這些都值得投資人深思。  
  
🔚 總結  
✅ 系統性風險無可避免，長期投資者應具備風險應對策略。  
✅ 攤平策略風險極高，可能導致資金部位加重，最終難以解套。  
✅ 停損機制極為重要，超過 20% 虧損後的補救措施成本高昂。  
✅ 九張加一張策略風險仍在，若無法精確計算持股數量與風險，可能仍陷入長期虧損。  
✅ 最佳策略是適時認賠，並將資金投入更有成長潛力的標的。  
  
📢 投資市場風險與機會並存，唯有理性決策，才能在市場波動中站穩腳步！ 🚀

```python
_,_,max_min = chart_pattern.DetectTurningPoints('close',order=10,smoothing=2)
pd.set_option("display.max_rows", None)
print("轉折點總筆數 ＝ {:d}".format(max_min.shape[0]))
display(max_min)

# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定均線與轉折點    
turning_points_len = len(np.array(df_k_line['Close']))
turning_points = np.array([np.nan]*turning_points_len)
for point in max_min.iterrows() :
    turning_points[point[0]] = point[1]['Price']    
apds = [
    mpf.make_addplot(turning_points,type='scatter',marker='o',markersize=50,color='xkcd:sky blue'),
    mpf.make_addplot(df_sma['SMA5'],width=0.5,color='xkcd:maroon'),
    mpf.make_addplot(df_sma['SMA20'],width=1.0,color='xkcd:violet'),
    mpf.make_addplot(df_sma['SMA60'],width=0.5,color='xkcd:dark orange')
]

# 繪出K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), addplot=apds, warn_too_much_data=1100)
mpf.plot(df_k_line,**kwargs)
```

```python
patterns = chart_pattern.RecognitionBottom(turning_point_args={'mode':'close', 'order':10, 'smoothing':2})
```

```python
pattern = patterns[0]['bottom_pattern']

neckline_start_date  = pattern['neckline_start_date']
neckline_start_price = pattern['neckline_start_price']
neckline_end_date    = pattern['neckline_end_date']
neckline_end_price   = pattern['neckline_end_price']
bottom_price_date    = pattern['bottom_price_date']
bottom_price         = pattern['bottom_price']
on_neckline_price    = pattern['on_neckline_price']
target_price         = pattern['target_price']
    
# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定頸線
seq_of_seq_of_points=[
    [(neckline_start_date,neckline_start_price),(neckline_end_date,neckline_end_price)],
    [(bottom_price_date,bottom_price),(bottom_price_date,on_neckline_price)],
    [(neckline_end_date,neckline_end_price),(neckline_end_date,target_price)]
                     ]

# 設定均線
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.5,color='xkcd:maroon'),
    "SMA20": mpf.make_addplot(df_sma['SMA20'],width=1.0,color='xkcd:violet'),
    "SMA60": mpf.make_addplot(df_sma['SMA60'],width=0.5,color='xkcd:dark orange')
}

# 繪製K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), volume=True,addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=(1.2,10,10), colors='xkcd:orange yellow', alpha=0.6), datetime_format='%Y-%m-%d', warn_too_much_data=2000)
mpf.plot(df_k_line,**kwargs)
```
