### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】中期偏空，醞釀反彈  
從上周四出現接近400點的跌勢之後，近幾個交易日都是偏空發展，昨天雖然開高走高，最大漲點來到134點，可惜量能嚴重不足，只有2668億，10點半過後隨即拉回，收盤甚至小跌11點作收。籌碼一樣沒有改善，法人賣、主力指標轉弱、融資增、融券減，籌碼持續偏空。更值得注意的是原先最強的櫃買指數，跌破重要支撐，先前變成假突破，短線轉弱，反而有醞釀頭部的雛型。  
先前提到中期空方的3個要件，依序是周KD值是否滑落80以下、20日線是否跌破下彎、盤整區間是否能夠守住？而這3個空方條件都接連出現，技術面已經從中期盤整轉為出現下降軌道，轉為高點越來越低，低點越來越低的空頭結構，走空趨勢應該已經確認，投資人比較想問的應該是多方還有沒有機會？  
雖然走勢偏弱，不過依照黃金分割率以及迷你頭部跌幅目標價，第一階段跌幅滿足應該會落在16370-16400，再者加上統計過投資人隨機的提問個股，雖然很多個股出現迷你頭部，但是因為只是迷你頭部，所以其實這些個股很快就會跌到支撐，所以殺傷力還不大，大盤不至於出現急跌。  
所以指數有賣壓，但還不至於有系統性風險的出現，而且伴隨昨天夜盤期指表現，以及剛好到下降軌道支撐，今天甚至多方有機會表現，關鍵是在未來修正過程中，中期支撐會不會有守，而能否有守的關鍵正是原先的迷你頭部個股，會不會進一步擴大成中期頭部個股，以及原本已經是中期頭部會不會開始起跌。  
而除了個股變化之外，加權指數到短線支撐稍微止穩之後，接下來更為關鍵的支撐，應該是浮動的中長期的上升趨勢線，會是跌幅是否繼續擴大的關鍵，在修正期間要密切觀察：外資未平倉口數、台幣升貶、類股結構、月KD變化，會是未來是否跌破中長期上升趨勢線的關鍵。最後，多頭看壓力不看支撐，所以多頭要轉強，一樣要站回16900再說。 (出處：K線捕手楊忠憲粉絲團)
