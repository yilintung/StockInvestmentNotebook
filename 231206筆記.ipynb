### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】二次背離，短線轉弱  
前兩日提到，國際股市和台股雖然持續偏多，但多方有5個瑕疵。首先是個股表現不如前幾周整齊，除生技和傳產之外，其他有表現的都是次族群、零星個股，且漲跌輪動快速，對一般投資人操作難度相對高。  
其次，投信近三周累計站在賣方，外資雖然買進，但買盤卻不強，融資持續穩定攀升，籌碼面中性偏空。第三，台幣升值到壓力區，最高來到31.23，短線壓力不容易有效突破，維持區間盤整的機率較高，台股的資金動能相對有限。  
第四，原先全球最強的指標-日本股市，也出現短線轉弱，短線只能維持狹幅盤整無法創高，這也是未來的潛在風險，要多加留意。最後，近期持續出現的高檔背離現象，台股指數創新高，但成交量沒有創新高的價量背離，在高檔位置出現價量背離，很容易變成多頭力竭的現象。  
加權指數周一開高走低，留上影線之後，昨天持續小跌修正93點，成交量只有3205億賣壓不重，不過KD指標死亡交叉，跌破5日線且5日線下彎，技術面短線轉折出現，價量關係2次背離，KD指標一次背離。  
外資站在賣方，投信買盤不強，融資小幅減碼，籌碼面也持續偏空。不過雖然加權指數短線轉弱，但中期只是回到大箱型區間，而20日線扣抵至少還需要2周的時間才會扣抵到高檔區，所以不至於出現連續性下跌。  
不過儘管如此，已經大大的影響台股推升的動能，加上櫃買指數到上升軌道壓力區，以及乖離過大，因此不論大型股或中小型股，都不容易表現，甚至漲多個股容易出現賣壓。  
由於指數是緩跌而非急跌，做多會受到壓抑，但沒有出現恐慌性下跌，我自己目前持有兩檔個股，維持上周的策略，採用「漸進式減碼」，也就是並非恐慌性的一次將持股全部出清，而是依照個別股票，出現賣出訊號再依序離場。 (出處：K線捕手楊忠憲粉絲團)
