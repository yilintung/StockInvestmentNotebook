### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線續漲，留意反壓  
加權指數短線連續反彈，回到中期盤整區間，但中長期變數會增加，未來季線會走平會對多方有利，但2周換半年線可能下彎，又轉為對空方有利。中性看待是高檔箱型，保守看待有擴頂現象。  
台幣走勢短線升值，中期維持箱型整理；類股指數只有3個類股偏多，其餘盤整或反彈；從均線多空比觀察，多方個股只有17%，空方個股在5成左右，空方個股明顯比多方個股多；綜合籌碼，短線偏多。  
國際股市和台股在關鍵支撐短線轉強，但中長期都有擴頂現象，趨勢尚未真正翻多。短線上有5%逆勢上漲股可以做多，優先挑選低位階、有底型、剛起漲的個股，但資金不宜太高，因為選股難度仍高。  
上周五夜盤期指上漲102點，今天估計會有小幅補漲，勿過度樂觀，本周留意反彈壓力區，是否有止漲現象。另外近3個月上漲都是低量，但是到轉折高點反而爆量，留意是否出現類似現象。 (出處：K線捕手楊忠憲粉絲團)
