從技術面的觀察，一檔穩健的強勢股，尤其是個股走勢，通常會出現階梯式上漲的現象，也就是漲勢停歇並沒有拉回，而是用橫盤的方式，然後越墊越高，橫盤時間越盤越短，直至最後在市場最樂觀的時候，高檔伴隨K棒的噴出。相對接近的是2021年5月中旬~2021年7月初的航運指數，在末升段有類似的走勢，加權指數目前已經有相似的現象。  
在5月底的時候，外資未平倉多單口數，在短短幾天內急遽攀升到接近35000口，之後每當指數越往上漲，期貨多單口數反而越減，相當符合外資過去在高檔區的操作。  
未來加權指數可以同步採用估算滿足點的方式，搭配設定移動式支撐，從旗型的漲幅規劃，加權指數滿足區可以設定在17250，而從頭肩底的規劃，漲幅較為樂觀，可估算到17400，換言之，從技術圖形可以得到漲幅滿足在17250-17400的區間。但為避免有超漲現象，可以同步觀察觀察移動式，追蹤單日從最高點拉回200點以上，或3日累計300點以上，也就是K棒出現爆量長黑或長上影線。  
雖然不論加權指數是偏多續漲，但投資人可能會覺得選股仍然很難，這是十分正常的，因為截至6月13日，雖然加權指數創波段新高，但加權指數28大類股中，只有3個類創新高，這意味著其他類股要不是盤整，不然就是走弱中，換言之，隨機要挑中會漲的個股，只有10%，甚至5%的機率。 (出處：K線捕手楊忠憲粉絲團)

【自己的買賣點，只有你自己知道】  
我常常說過每個人條件不同，打個最簡單的比喻，假如某股票目前100，你買在80那可以很悠哉的抱著，但如果你買在95，現在每天可能都要相當警戒，又比如追買在99，那99一跌破，可能就要考慮減碼或停損。  
這還只單就一個買進價格做假設的情境做假設，倘若還加上每個人究竟是試單階段、加碼階段，還是買足階段，3種價格乘上3種情境，就會有9種，但更可怕的是每個人的風險控管、張數上限、資金水位、交易周期、做多做空、股齡、專業程度都不同...這樣推演下來，會有數百種排列組合。  
因此我解盤極少提到如何交易，只能客觀就我看到的多空趨勢、漲跌強弱、轉折訊號、位階、型態...，至於如何交易或是做多做空，就要靠各位的經驗跟智慧了，不是故意要打啞謎，是真的有困難，而且這當中的買賣感覺、交易操略，都必須自己親身參與，才會學到東西。  
也因為上述的條件不同，在每天說明跟教學過程，一直都會出現這樣的情境，其實我寫的內容只有一個版本，可是到了不同人眼中，就會出現各種不同的組合跟變化。莎士比亞的名劇哈姆雷特：一千個人眼中，有一千個哈姆雷特。同一套劇本傳了幾百年，永遠只有一個版本，但是一千個人眼中會有一千種體會，和各自解讀的方式。  
  
我們用反向ETF做個示範，大家就會比較有感。有些人資金不多，但可能到16000點就全壓了；有人可能更辛苦一點，資金部位大，但一樣16000點就全壓；有些人目前只用了1/3的錢，他還有2/3的資金，可以在未來半年甚至一年，平均分批進場；有些人這段時間都持續個股作多，想等盤勢正式出現轉折，反向ETF再規劃進場。  
當然如剛剛說的再加上每個人的風險控管、交易周期、股齡、專業程度…，那最後產生的變化跟結果，就會有很大的差異，會出現無限多種的可能，所以只有你可以幫自己做決定。

```python
##### 使用2022年9月19日到本日(2023年6月14日)收盤之日線圖資料 #####
# 載入從「好神通PLUS」輸出的Excel檔
df = pd.read_excel('./data/SINO_2610_230614.xlsx')

# 保存K棒的基本資訊(開、高、低、收、量)
df_k_line = df.drop(columns=['SMA5','SMA10','SMA20','SMA60','SMA120','SMA240','MA5','MA10'])

# 保存價的均線
df_sma = df.drop(columns=['開盤價','最高價','最低價','收盤價','成交量','MA5','MA10']) 

# 將K棒的Columns的名稱由中文改為英文
df_k_line = df_k_line.rename(columns={'時間':'Date','開盤價':'Open','最高價':'High','最低價':'Low','收盤價':'Close','成交量':'Volume'})

# 將價的均線的Columns的名稱由中文改為英文
df_sma = df_sma.rename(columns={'時間':'Date'})

# 將K棒的Date設為Index
df_k_line.set_index(df_k_line['Date'],inplace=True)
df_k_line = df_k_line.drop(columns=['Date'])

# 將價的均線的Date設為Index
df_sma.set_index(df_sma['Date'],inplace=True)
df_sma = df_sma.drop(columns=['Date'])
```

```python
# 手動設定頸線
neckline_price = df_k_line.loc['2023-05-29']['Close']
neckline_start_date = '2022-09-22'
neckline_end_date = '2023-06-14'
print('頸線價格 ： {}元'.format(neckline_price))

# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定頸線
seq_of_seq_of_points=[
    [(neckline_start_date,neckline_price),(neckline_end_date,neckline_price)],
                     ]

# 設定均線
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.4,color='xkcd:royal purple'),
    "SMA10": mpf.make_addplot(df_sma['SMA10'],width=0.4,color='xkcd:cyan')
            }

# 繪出K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), addplot=list(added_plots.values()),alines=dict(alines=seq_of_seq_of_points, linewidths=2.5, colors=('xkcd:yellow orange')), volume=True)
mpf.plot(df_k_line,**kwargs)
```
