「[老師剛剛說](https://www.facebook.com/kevinyang20180801/posts/pfbid02ciQdCzox2i6kgnSsZrreiGyiFugedXAoUmJrFkFSvBggJqKNRrDZ9AZ4gK8RQj6nl)，今天空方壓力測試到這邊應該暫到一段落，多方可以稍微喘息一下，不過還是要留意國際股市表現。說完之後真的差不多就是最低點，老師是怎麼判斷的呢？」  
前2周欠大家的一個問題，但中間很忙(其實都是盤中跑出去玩)，一直忘記回覆，同學提醒之後，趕緊抓時間補寫。故事是這樣的，某一天指數下跌，我說跌到某個位置，應該就會止跌，之後到了該位置，真的就立刻止穩反彈了，盤中要如何判斷上漲的壓力跟支撐呢？  
首先還是必須要說，判斷支撐跟壓力不難，但是否當下反彈或拉回，時間會有誤差，發文的當下，我也不知道說完，就會立刻反彈，所以說完立刻就反彈和拉回，那是有運氣成分。  
接著是如何規劃支撐和壓力，通常我會從日K線的趨勢、位階、型態、轉折，去找到支撐和壓力，但如果想要在盤中看得更細，那就需要切換到5分鐘K線，可以看得更仔細，不過真的是盤中很無聊，我才會看這麼短，否則一般做交易，看日K線就足夠。  
第三是必須同步觀察亞股走勢和當天的領漲或領跌類股，可以提早輔助判斷是否出現止穩和反彈，比如某一天，因為國際股市偏弱和電子指數領跌，通常國際指數和電子股會領先止穩，這時候就可以判斷台股指數應該也很快就會反彈。  
第四是預估量的觀察，可以判斷賣壓的大小，以及市場恐慌的氣氛，不過預估量9點半過後會比較穩定和準確。最後是下跌的類股結構，如果當天指數大跌，但是下跌的都是漲多股拉回，或是弱勢股破底，但續漲股相對有撐，這時候就也可以知道，跌到支撐區容易會有守。  
以上這些現象，必須用動態說明，還有圖文對照講解，才會比較清楚，文字說明只能點到為止，最重要的是希望大家知道，看盤是有所依據，不是用感覺和猜想，細節部分有機會上課再完整說明。
