### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線續強，留意壓力  
台股昨天開高大漲，雖然10點過後稍有壓回，但加權指數仍上漲192點，櫃買指數上漲0.32%的漲幅，相同的是兩個指數量能都還是偏低，始終在10日均量之下。  
未來短線挑戰20883前高的要件包含：類股結構要改善，多方類股至少要增加到10個；價量關係要在10日均量之上，增加到4500億是最低標準；期貨空單口數降低到1萬口以下；現貨籌碼，法人買超增幅持續擴大；台幣升值，20日線下彎。  
目前上述條件，只有法人現貨買超幅度擴大，其餘條件還是沒有明顯改善，尤其是反彈過程沒有量，都是可以接受的，但是如果要攻高，量能就很關鍵，未來指數越往上漲，要越加留意箱型區間上緣壓力。  
中期技術面觀察季線支撐和扣抵值，長線觀察月KD指標。選股和操作不難，3-5成資金偏多操作。原先最強的三大族群，營建還持續多方，但重電和電纜已經短中期翻空甚至做頭。  
但有其他族群接棒演出，營建、橡膠、觀光餐旅、金融、貿易百貨、水泥，選股和操作難度不高，不過中期趨勢沒有明朗之前，保守操作者資金維持3成，積極操作者資金也要控制在5成以下。 (出處：K線捕手楊忠憲粉絲團)
