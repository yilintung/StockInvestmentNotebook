```python
def StockRankEvaluate(prices_mean,prices_std,price):
    rank = ''
    price_std_range = (price - prices_mean) / prices_std
    if price_std_range < -2.0 :
        rank = '極低'
    elif price_std_range > -2.0 and price_std_range < -1.0 :
        rank = '低'
    elif price_std_range > -1.0 and price_std_range <= 0 :
        rank = '稍低'
    elif price_std_range > 0 and price_std_range < 1.0 :
        rank = '稍高'
    elif price_std_range > 1.0 and price_std_range < 2.0 :
        rank = '高'
    elif price_std_range > 2.0 :
        rank = '極高'
    print("價格 ＝ {:.2f}元（標準差範圍 ＝ {:.2f}） ， 位階評價：{}".format(price,price_std_range,rank))
    return rank
```

```python
# 抽取收盤價
prices_close=np.array(df_k_line['Close'])

# 求2022年2月25日到2023年3月17日之收盤價平均值與標準差
prices_close_mean = np.mean(prices_close)
prices_close_std = np.std(prices_close)
print("收盤價區間：平均值 = {:.2f} ， 標準差 ＝ {:.2f} ， 最高 ＝ {:.2f} ， 最低 ＝ {:.2f}".format(prices_close_mean,prices_close_std,np.max(prices_close),np.min(prices_close)))
```

```python
# 評價買入時(36.30元)的位階
price = 36.30
rank = StockRankEvaluate(prices_close_mean,prices_close_std,price)

# 評價2023年6月29日收盤價的位階
price = prices_close[-1]
rank = StockRankEvaluate(prices_close_mean,prices_close_std,price)
```
