10年前的今天，自己寫的筆記，人的記憶最不可靠的，寫下來才會清楚。當時已經慢慢進入成熟的階段，所以筆記做的比較簡略。  
  
  
我是2009年開始買股票，但2010之前都還是亂買亂賣，然後慢慢覺得受不了忽賺忽賠，所以開始認真讀書學習跟上課，2011-2012是最密集的學習，邊做邊練習，到2013年算是很關鍵的一年，那一年之後操作才慢慢變的穩定，看了自己寫的筆記，會切割得很清楚。  
   
有同學問我，為什麼技術面跟籌碼面就已經轉弱，甚至可以說是翻空，但我們現在接收到的媒體，都還是樂觀看多？  
我10多年前遇過一個業內的前輩，他民國75年入行，他說以前進公司時，老闆上班跟她說的第一件事情，甚麼都可以寫，就是不能寫空。  
因為寫空就沒人看，報紙雜誌就會賣不好，所以就算有一天已經累計跌幾千點了，你還是會看到逢低承接、長線看多、基本面沒問題…，所以投資人必須要有自己的判斷和思考，不能人云亦云。  
但你不要怪媒體，永遠看多寫多，是媒體的天職。  
  
同學問：老師，之前上課有提到，在多頭走勢中，如果有回檔狀況，以盤代跌是比真下跌好。那在目前的空頭走勢中，這種以盤代彈的方式，是不是比真反彈還要糟糕的意思？因為老師上課時對空頭琢磨較少，想跟老師確認一下。請老師有空再回覆就好，不要影響到您的旅程，祝老師闔家旅途愉快短中期翻空  
我回答：其實盤整本來就有兩層意義，從多方思考來說，空頭下跌中，以盤代跌自然是化解壓力的方法，但是在空頭整理中，盤整卻也是最弱勢的表現，所以這兩件事情並沒有衝突，只是從不同角度切入，表徵意義不同。
