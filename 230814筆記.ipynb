【[空方來襲時，不同屬性的投資策略](https://www.facebook.com/kevinyang20180801/posts/pfbid03CS1td4U36rhWf4sToRfEnYb3AMzpgqFbG5MJYYtsUk7g3SP4c1yxDWmWcDeVzJpl)】  
由於國際股市和台股都偏空方弱勢，短中期進入空方，以下依照不同投資人的屬性，做了5種策略規劃，提供大家參考。  
首先是第一種最保守的作法，就是原先在高檔已經做預防性減碼，持股比重低甚至已經空手的投資人，那就靜待指數完全落底，再重新進場。  
第二種則是先前尚未做減碼，多方部位資金比重仍高，務必緊盯個別股的離場條件，並且紀律執行，即使無法做到紀律停利停損，切勿急著往下逢低攤平承接。  
第三種則是積極型，不僅做多部位出清，更已經反手做空，不過一定要衡量自身條件和交易能力，並做好風險控管。  
第四種是保守型，但仍想參與指數下跌，可以適度配置反向的ETF，雖然利潤空間有限，但一樣可以參與下跌的獲利機會。  
最後一種，尋找仍在多方或是底型的個股操作，但由於對於國際股市和台股的外在條件而言，屬於逆勢操作，資金比重務必壓到最低，只求有賺錢不求賺大錢。  
以上5種，大家可以看看你是屬於哪一種？可以參考對應的策略和注意事項。
