【[關於存股你該注意的一些事情](https://tinyurl.com/35fae65x)】  
其實早期已經寫過蠻多對存股(或者要稱為價值型投資或長線投資)的看法，當中討論很多存股的盲點和風險，這邊就不一一贅述，但要先聲明我不是反對存股、長期投資或價值型投資。  
只是很多人好像都以為，只要買完之後，擺著、放著，甚麼事都不要做，有一天就會變成有錢人，但這真的是很危險的，總結幾個重要觀點，給想存股的同學，應該有的一些觀念：  
1.是否有願意用年以上為單位，逢低有持續以及有計畫性，分批買進的資金和勇氣？  
2.持有一段時間後，股價上漲了，該一直加碼？還是逢高減碼和獲利調節？還是像坐電梯抱上又抱下？而調節的具體依據和工具又是甚麼？  
3.萬一長期逢低持續買，買了多年以後，平均成本是100元，但後來股價跌到50元甚至更低，才發現股價可能再也不回來了，那時候該怎麼辦？  
4.持續買了很多年之後，已經投入巨額資金，後來公司基本面才轉差，而且配股配息也變差，甚至未來可能不配股配息，那時又該怎麼辦？  
5.投資系統中有沒有停損機制？願意執行停損機制嗎？  
如果你沒有想清楚以上的問題和風險，以及具體的因應措施和解決方案，那就不是存股、不是長期投資、不是價值型投資，而是孤注一擲的豪賭，恐怕有一天會出大事的。  
  
【[把它當小說看就好](https://tinyurl.com/57bdar7m)】  
今天盤勢應該變化不大，來聊聊我學習投資的啟蒙，不知道大家幾歲的時候，開始思考職涯規劃，當然也包含退休，更進一步的說就是現在流行的財富自由。聊這個話題之前，先講一個很衝突的名詞「樂觀進取」，為什麼說是衝突，因為這個詞的組合本身就有問題，樂觀的人怎麼會進取呢？樂觀的人會活得很自由自在，每天都很快樂，船到橋頭自然直，等到問題出現再來說，所以樂觀的人很難進取，所以這只是心靈雞湯鼓勵你的話，千萬不要當真。  
因此，真實的情況，悲觀的人才會進取，因為懂得煩惱，我就是屬於這一種，真的是很悲觀、很會煩惱，所以1996年剛剛考上大學，就精神錯亂的開始會想這些問題，為什麼會開始想？如果對生活稍微有危機感，把月薪的中位數乘法12個月，再乘上40年的職場生命，一輩能賺多少錢，計算機一算幾秒鐘就知道了，基本上已經固定，扣除基本開銷，房租、買屋、結婚、生子、教育、退休、養老…，其實根本一出社會就宣告破產了，人生是從負債開始。  
但是當我跟身邊同儕聊這個想法時，但身邊的同學都笑說，我腦子有問題，想那麼多做甚麼，上大學就是要努力玩，雖然我也是哈哈大笑，不過心裡已經悄悄埋下種子，必須比別人更積極努力，因為如果不做點跟別人不同的事情，先不說追求甚麼大富大貴，光是想要爬上貧窮線，像人一樣的活著就是問題了。這件事情在1996年就已經很衝擊，當然這個情況經過20多年，又更加惡化了，上面說的問題現在看來，更是像針一樣刺進每個在職場上的人心理，只能像白老鼠在迴圈中，中期一生無法停歇。所以我18歲讀大學時，就已經開始準備了…
