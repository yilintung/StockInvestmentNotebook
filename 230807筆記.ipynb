【關於499這個神秘數字】  
投資人在單日或波段大漲大跌時，會出現499的大單，一般會判讀這是大戶的進出，只是很特別的是，好像最多看到的就是499，很少看到499以上的大單，在觀察交易明細時，不難發現常常出現單筆499張的大單出現，通常被視為大戶敲單動作，但為何是499呢？  
因為我也沒敲過這種單XD，所以請教了一下券商的朋友。其實主要原因是因為只要超過500張，主要的原因就是超過500張就會認定是鉅額交易，必須有申報的程序。而且為了避免這樣的麻煩，下單系統軟體通常會自動拆單，所以單筆委託上限就是499張。   
不過鉅額交易不只以500張為基準，只要要單一證券金額達1,500萬元以上，或是股票組合5種股票以上且總金額達1,500萬元以上，就符合鉅額交易的門檻。不過這對一般人算是冷知識了，純粹是訊息分享，真的要遇到的人應該比較少，對交易和投資影響不大。

講到第5集，終於要聊到青少年投資理財的具體項目，以我自己的作法，基本上是從幼稚園大班開始規劃，首先是小孩收入的來源，我不會每周或每月固定給小孩零用錢，而是當他有好的表現，舉凡參加比賽、考試優異、被師長誇讚、遵守生活常規、幫忙大人…，會在一個專用本子，上面蓋一個印章，而每個印章可以兌換10元。  
這樣子做有兩個目的，其一，如果直接固定給零用金，容易養成不勞而獲的習慣，好像甚麼都不用作，每個月自動就會有錢提撥下來，要讓他知道必須透過勞務，或者是努力付出才能得到，自己想要的東西必須靠。其二，為什麼是給印章，而不是直接給10元？因為當時還太小對實體貨幣沒概念，而用一個專用本子紀錄，可以有類似存摺的意義，這是他自己收入的來源。  
接著就是如何分配這些收入，基本上可以先初步分為3個區塊，儲蓄、消費、投資，但是投資對10歲以前的小孩，太過虛無飄渺，所以我會直接把儲蓄和投資先納為一塊，假如一整年累積了200個印章，當中的一半也就是100個印章必須存起來，如果一整年100個都沒有花掉，我會多給他5個，除了培養儲蓄的習慣之外，也是建立存錢在銀行有利息的觀念(雖然這利息比銀行高很多)。  
而另外100個印章，會讓他自由分配，可以買想要的東西，也就是自己支配消費，但我會跟他說這100個印章不能只有買玩具跟遊戲，部分要買學習和知識的書籍，通常會聽他說完想買的東西，會分析他當中的優缺點，例如需要跟想要的差異、遞延消費欲望、不要隨便浪費、同商品不同網站的價差、運費的成本…，也會提醒他要幫助弱勢，所以他曾經用印章的錢，換了身障人士的玉蘭花，還有流浪狗的飼料。  
小六之後就幫他開了一個郵局戶頭，正式有了自己的存摺，讓他有私人財產的觀念，國一開始會要求記帳，會提供一個簡易的記帳格式範本，例如：上期結餘、本期收入、本期支出、收入和支出明細、本期結餘，不需要天天更新，大概一個月或是有較大的異動填寫一次就可以。我覺得國中之前，能建立和落實這些觀念就足夠，至於到高中之後，還需要增加甚麼投資的知識？留待下一期再聊，最後感謝您的按讚以及轉傳分享~
