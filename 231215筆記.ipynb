### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】突破創高，守穩支撐

因此加上今天，漲點這個條件已經符合原先的規畫要求，而一直缺少的量能在突破關鍵壓力區也補出來，昨天成交量來到3898億，超過3500億，最後就是時間了，今天是突破的第二天。  
所以短線加權指數從盤整回到多頭軌道，因為是多頭續漲，其實可以用旗型估算漲幅滿足，不過因為現在位階高，用旗型估算怕過於樂觀，心態不容易調整，所以改成守移動支撐。  
多頭看支撐不看壓力，只要支撐沒跌破，趨勢就沒改變，目前支撐可以設定在17550，但因為是移動支撐，每天都會隨著指數漲跌而需要挪移，近期會盡量每天都更新支撐位置，讓大家參考。  
也因為創了新高，也出了量表態，接下來走勢就只能前進不能後退，即使小幅後退，支撐也必須要守住。目前唯一的瑕疵，就是多方接棒的個股還不夠多，幾乎都是原先已經突破或是上漲，所以如果續航力道要紮實，可能還需要有更多生力軍加入。  
另外，外資期貨空單維持在萬口第二天，屬於偏高的現象，超過5千口已經不是單純的避險空單，這也是需要密切觀察。最後是操作策略，目前持股只剩一檔，今天有可能會新增1-2檔。
