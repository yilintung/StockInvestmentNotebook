【真正的決戰，永遠在空頭】  
先不要回溯到2009，畢竟那有點太遠，從2022年開始回顧就好。短一點的觀察，這一波起漲大約是從2022年10月開始，接下來2023年，中間雖然有很多的波段進入盤整，但整體來說指數是越墊越高，因此整個趨勢都是偏多方發展的。  
因為長達1年又3季的大多頭，因此過去投資人即使有短套，只要基本面不要太糟，死抱活抱也會解套，甚至套牢幅度也可能不會太大，不管是研究基本面、籌碼面、技術面，甚至聽消息、不紀律、亂買、亂賣...，2022年底到2024年第二季都可能會獲利。

投資人務必檢視自己的操作方式，究竟是僥倖賺到錢，還是用嚴謹、有系統、正確的方式賺到錢。多頭全部都是贏家，大多數的人都賺錢，但是最後的決戰永遠在空頭市場，究竟鹿死誰手，誰高誰低，那個時後才見真章。
