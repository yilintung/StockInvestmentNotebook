但既然國際股市和台股都在多方，個股自然要順勢作多，只是作多過程中要同步留意一些訊號。  
首先是台幣的升貶，短線上台幣仍在31.21-31.65區間波動，外資仍會持續站在多方，但倘若貶破31.65，容易形成新一波貶勢，外資持續買進權值股的動能將會降低，指數推升動能會大幅降低。其次是外資期貨未平倉口數變化，不論多空口數在5千口以下，基本上意義不大，但倘若期貨空單攀升到萬口之上，要留意外資現貨買超力道的竭盡。
