[對散戶而言，空頭來臨時，持有現金就是最好的避險。](https://www.facebook.com/kevinyang20180801/posts/pfbid02cXKWNYb25mQXoujbnVsLQALckjHHos7c4gic6tjnuYcDAjPBDmRQuvKZnKhLNoPRl)  
【保留現金就是最好的避險】  
投資人一定要記得，不論做多或是做空就是為了賺價差，絕對沒有所謂的避險操作。因為坊間盛行一種說法，假如有400萬的股票部位，但是覺得自己的股票基本面都很好，即使出現技術面賣訊，仍不想賣出，於是做空一口期貨，做為避險。  
但是這是相當危險的事情，投資人只要想，假如像規畫中的，指數真的下跌，現貨部分必然跟著下跌，而放空的期貨會獲利，但是下跌之後，接下來到底要賣現貨，還是賣期貨呢？恐怕就是一個很難的問題，因此這樣的操作充其量只是讓虧損鎖住不放大，但是並沒有真正解決問題。  
再另一種更複雜的情況，假如指數不跌反漲，指數做空部位虧錢，結果持有的現股並未跟漲，反而出現下跌現象，豈不是兩邊都虧損。既然如此何不在一開始有風險或是出現賣訊，直接出清持股，來的簡單些呢！  
但為什麼會有這種說法的產生呢？因為這是法人使用的方法，由於法人現貨部位太多，比方台積電持股超過7成，根本不可能一次完全出脫，所以才需要避險操作。但是散戶假如有5千萬、1億已經是大戶等級了，但是如果持有中大型股票股，一天就就可以賣光，而且對股價也不會有波動跟影響。  
所以對散戶而言，保留現金就是最好的避險，要清楚不論做多或是做空就是為了價差，千萬不要為了避險而一手做多，一手做空，這樣只會讓問題更加複雜，陷入更為難的境地。
