【解盤練習應用技巧】  
感覺最近越來越多同學，加入寫解盤試卷的行列，當然一開始寫的牛頭不對馬嘴或是窒礙難行，都是很正常的，凡事都是由淺入深熟能生巧，上課時也都有一步一步帶大家練習，這邊稍微補充一下，寫解盤練習的應用技巧。  
其實寫解盤練習，主要目的就是訓練自己，在不依賴任何人的狀況，假定個股滿分10分，能挑出積分8分、甚至9分的股票，換言之解盤練習的目的，就是訓練自己選股的能力，所以鉅細靡遺的解盤內容，只要寫一次即可。  
假如你寫解盤練習，真的還在模擬階段，只是為了累積經驗值，而非一定要下單，那可以比較悠閒一點，每天收盤甚至間歇性的，觀察股票是不是如最初所規劃的，完成底型、旗型、黃金分割，或是依照支撐跌破法完成離場的條件。  
但假如寫完解盤練習，真的挑中想要買進的個股，那就要依照選股表格中最後欄位擬定的策略進行實戰操作，倘若股價依照規劃行進，那不用每天重新寫一張解盤試卷，這樣工程太大了，也太累人了。  
而是當股價走勢偏離原本解盤規劃，甚至出現可能需要停損，要讓自己歸零思考，再重新完整檢視時，那樣的情境出現時，再重新寫一張新的即可。最後，請給自己10檔的練習，只要堅持寫10檔，會看到全新不同的自己。
