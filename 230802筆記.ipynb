### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】金融偏多，電子偏空  
加權指數小漲67點，成交量3815億，低於10日均量的4024億，屬於價漲量縮的反彈格局。法人小幅回補，融資續增，融券減碼，主力指標轉弱，整體籌碼中性偏空。台幣續貶創新低，來到31.52。類股結構上一樣電子弱、金融強，也呼應周末提到的部分。  
不過當多數人都關心加權指數，更要謹慎小心的是櫃買指數，已經第6次測試頸線，有明顯頭部雛形，所幸下方有季線支撐，短時間還不至於立即有危險。但從籌碼面觀察，法人續賣、融資增加、主力指標轉弱，已經持續一個月左右。未來先發生風險的地方可能是櫃買指數，當然如果你買的是中小型股居多，接連兩日應該已經明顯感到賣壓。  
AI概念股還在尋求支撐測試，尚未全面止跌轉強，一定有當初沒有跟上車的投資人，正在思考要不要拉回找買點，但要注意是「拉回找買點」，不是拉回就買，所以是拉回時出現買訊，且搭配多頭趨勢尚未改變。未來或許還有機會出現上漲，但已經過了無腦操作階段，真正的高手還會有機會，但操作難度很高，功力不夠看戲就好。  
趨勢翻空會有領先訊號、同時訊號、確認訊號、落後訊號，目前領先訊號算是出現，但是同時訊號只出現一半，所以在沒有跌破關鍵支撐之前，盤勢仍可以視為漲多拉回，或者中性一點的說，還是大箱型區間整理，而確認訊號比較簡單的方法也可以觀察昨天提到的，中期和長期兩條上升趨勢線。 (出處：K線捕手楊忠憲粉絲團)
