```python
import mysys
import markdown
import base64
from IPython.core.display import HTML
```

```python
def stock_analysis( analysis, stock_id, buttom_pattern = False) :
    results,images = analysis.analysis(stock_id)
    if buttom_pattern is True and '底部型態' in images :
        print('底型反轉交易策略：')
        display(images['底部型態'])
    print('解盤內容：')
    if '整體評價' in images :
        display(images['整體評價'])
    result_md   = results.to_markdown(tablefmt="grid")
    result_html = markdown.markdown(result_md, extensions=['markdown_grid_tables:GridTableExtension'])
    display(HTML(result_html))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import mplfinance as mpf
import numpy as np
import sqlite3
```

```python
stock_id         = 'TAIEX'
daily_start_date = '2024-06-28'
daily_end_date   = '2025-07-02'
```

```python
# 連線資料庫
conn = sqlite3.connect('data/stock.db')
```

```python
# 從資料庫中載入「台股總覽 TaiwanStockInfo」
df_stock_info = pd.read_sql("SELECT * FROM StockInfo", conn)

# 將載入的「台股總覽 TaiwanStockInfo」進行格式轉換
df_stock_info.set_index(df_stock_info['StockID'],inplace=True)
df_stock_info = df_stock_info.drop(columns=['StockID'])
```

```python
# 取得該股票代碼的產業分類
industry_category = df_stock_info.loc[stock_id]['IndustryCategory']
```

```python
# 讀取日Ｋ價格資料
sql_cmd = "SELECT * FROM DailyPrice WHERE StockID='{}' AND (Date BETWEEN '{}' AND '{}')".format(stock_id,daily_start_date,daily_end_date)
daily_price_df = pd.read_sql( sql_cmd, conn)
```

```python
# 格式轉換：日期格式、成交量(成交值)
daily_price_df           = daily_price_df.drop(columns=['SerialNo','StockID'])
daily_price_df['Date']   = daily_price_df['Date'].astype('datetime64[ns]')
daily_price_df.set_index(daily_price_df['Date'],inplace=True)
daily_price_df.drop(columns=['Date'],inplace=True)
if industry_category == 'Index' or industry_category == '大盤' :
    daily_price_df           = daily_price_df.drop(columns=['Volume'])
    daily_price_df           = daily_price_df.rename(columns={'Value':'Volume'})
    daily_price_df['Volume'] = daily_price_df['Volume'].div(100000000.00)
    daily_price_df['Volume'] = daily_price_df['Volume'].round(2)
else :
    daily_price_df           = daily_price_df.drop(columns=['Value'])
    daily_price_df['Volume'] = daily_price_df['Volume'].div(1000)
    daily_price_df['Volume'] = daily_price_df['Volume'].round()
    daily_price_df['Volume'] = daily_price_df['Volume'].astype('int64')
```

```python
display(daily_price_df)
```

```python
def date_to_index(df_in,date_in) :
    number_array = df_in.index == date_in
    idx = 0
    for number_index in number_array :
        if number_index == True :
            break
        idx = idx + 1
    return idx
```

```python
# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定下降壓力線
resistance_line_start_date  = '2024-07-05'
resistance_line_end_date    = daily_price_df.iloc[-1].name.strftime("%Y-%m-%d")
resistance_line_x           = [date_to_index(daily_price_df,'2024-07-11'),date_to_index(daily_price_df,'2025-01-08')]
resistance_line_y           = [daily_price_df.loc['2024-07-11']['Low'],daily_price_df.loc['2025-01-08']['High']]
slope,intercept             = np.polyfit(resistance_line_x,resistance_line_y,1)
resistance_line_start_price = slope * date_to_index(daily_price_df,resistance_line_start_date) + intercept
resistance_line_end_price   = slope * date_to_index(daily_price_df,resistance_line_end_date) + intercept

# 設定盤整區間
line1_start_date = '2024-09-23'
line1_end_date   = '2025-03-26'
line1_price      = daily_price_df.loc['2025-01-07']['Open']
line2_start_date = line1_start_date
line2_end_date   = line1_end_date
line2_price      = daily_price_df.loc['2024-12-02']['Open']

seq_of_seq_of_points=[
    [(resistance_line_start_date,resistance_line_start_price),(resistance_line_end_date,resistance_line_end_price)],
    [(line1_start_date,line1_price),(line1_end_date,line1_price)],
    [(line2_start_date,line2_price),(line2_end_date,line2_price)]
]

# 繪製K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), alines=dict(alines=seq_of_seq_of_points, linewidths=(1.5,1.0,1.0), colors=('xkcd:red','xkcd:bright blue','xkcd:bright blue'), alpha=0.6), volume=True, datetime_format='%Y-%m-%d')
mpf.plot(daily_price_df,**kwargs)
```
