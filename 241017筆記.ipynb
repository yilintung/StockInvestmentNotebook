[子曰：「今吾於人也，聽其言而觀其行。」](https://www.facebook.com/kevinyang20180801/posts/pfbid0kT9691whJnW6SSb8kBDZiPHAKq3kU8vYEN6sF17efMz8HrdadCrk4CSvffTFfG2Yl)  
基本分析是「聽其言」，聽公司派怎麼說、聽法人怎麼說、聽媒體怎麼說、聽小道消息怎麼說...，你要怎麼聽都沒關係，但不能只靠耳朵做股票。如果你說的都是真的，那股價就應該要表現給我看，所以技術分析正是「觀其行」。  
倘若基本面和未來產業趨勢講得天花亂墜，但股價卻跌得一蹋糊塗，那就是言行不一，這種股票還是避而遠之比較好。  
假如基本面跟股價背道而馳，你會相信基本面，還是相信股價？我絕對會相信股價，因為決定我賺賠的是股價，不是營收、獲利那些財報數據。你一定想說，基本面好，股價以後一定會漲回來，但很多時候都是股價先跌了一大段、跌慘，甚至跌死，媒體才跟你說，很抱歉基本面轉壞了。
