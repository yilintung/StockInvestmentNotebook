### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】量縮整理，狹幅區間  
昨天再度出現開高走低，符合盤前說的「狹幅盤整，忽多忽空」，進入第7天的狹幅整理，短天期持續盤整，明顯有拉大出小的現象，也就是表面上加權指數穩定，但實質上櫃買指數走弱。從資金面的觀察，台幣貶勢趨緩，但尚未回到升值軌道；籌碼部分，外資買超幅度不算大，不過至少讓多方喘口氣；技術面高點沒過低點也沒跌破，只是目前空方條件仍比多方條件多，而且如果多方再不表現，下半周國際股市，當然也包含台股，會有更多指數出現季線下彎。  
台股的類股結構變化，高檔箱型：電腦及周邊設備、網通、電子零組件、電子、營建，中期翻空：光電、油電燃氣、綠能環保、數位雲端、運動休閒、居家生活、塑膠、紡纖、玻璃陶瓷、航運、金融保險、化工，中期頭部：電子通路、資訊服務、其他電子、水泥、食品、電機機械、電器電纜、化學、生技醫療、造紙、鋼鐵、橡膠、汽車、觀光餐飲、貿易百貨、半導體，因此空方類股仍多，空方結構紮實。  
所以總體來說，空方條件是否有改善？從資金面、籌碼面、技術面、國際股市、類股結構，多方現象都還沒有出現，只是符合周末影音解盤說的，短線有支撐，但止穩轉強還有待努力，而短線盤整最好的策略，不急著重壓多空，等待是最好的策略。中長期觀察除了台幣升貶、外資期貨、月KD指標之外，還可以觀察一個現象，多頭是否結束，看強勢股是否止漲，反過來說，空頭是否結束，看弱勢股是否止跌，而這一波國際股市中，最弱勢的就是陸港股，可以列入條件之一。  
最後是萬眾矚目的輝達利多公布財報，但我先前曾提到兩個觀點，第一是不管公布的是利多還是利空，都是要修正的，第二有出現指標跟價量關係，3個高檔2次背離的現象，昨天開創高後走低，也符合了多指標二次背離，未來留意短中期空方轉折是否依序出現。 (出處：K線捕手楊忠憲粉絲團)
