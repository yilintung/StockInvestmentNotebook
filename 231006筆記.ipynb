### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】止穩有餘，尚未翻多  
昨天盤前提到幾個重點有利於「醞釀短彈」，包含：成交量2426億賣壓不重；台幣續創新低後，由貶轉升；來到16202前低支撐，多方會有抵抗；結構上賣壓沒有集中在特定族群。  
開盤後受到美債殖利率與美元因油價出現創高回檔，美科技股走強，加上政府基金撥款消息激勵，加權指數開高走高，終場上漲180點，成交量2400億，小幅站上16450壓力，收在16453點，雖然預期會上漲，但漲幅還是相對驚人。  
不過即便單日出現大漲，成交量仍然偏低，結構上屬於盤整量縮，籌碼面也還沒有明顯改善，外資尚未明顯跟進。而櫃買指數也是同步上漲，但還在大三角收斂區間，趨勢尚未扭轉。  
各類股當然同步上漲，結構上30大類股都是反彈，只有電子通路偏多，化工和生技醫療轉為中性偏多，因此只有個別股表現，多頭類族群仍不足，而且即將迎來4天雙十連假，變數增多市場會偏觀望居多。  
尤其昨天提到的國際股市中，除那斯達克勉強屬於瀕臨做頭，德國、道瓊、費半、陸股、港股、日本、韓國都已經做頭，空方力量並沒有改變，昨天雖然台股短線表現強勢，但止穩有餘，上攻力道不足。 (出處：K線捕手楊忠憲粉絲團)

如果常常看我解盤和上課的同學，很容易發現我解盤，甚至於交易，有一套固定模組，只是重複的運用而已，但分享會時偶而會聽到同學問：「老師，你怎麼不看某某技術指標？」、「老師，你怎麼沒追蹤某某財經數據？」、「老師，你怎麼沒有看分點券商籌碼？」、「老師，你怎麼沒有提到某個媒體發布的產業新聞？」  
其實蒐集、研究和分析眾多資訊，不外乎是為了得做出結論，以及研擬操作策略，比如要做多還是做空？資金比重要高還是要低？要選指數型商品還是個股？如果目前既有的工具就足以得到這樣的效果，再繼續追蹤下去，只是變成畫蛇添足，增加自己的工作負擔，單就技術分析工具，坊間能找到的可能就有近百種，怎麼可能全數納入。  
其實，《技術分析精論》一書40多年前早就說過，技術分析工具10-12種是最剛好的，如果一次使用太多，要不是重複，要不然就是沒發揮個別工具的功能，只是一直堆疊，像駢體文無止盡的堆砌文字，看似很華麗，但根本沒有實質內容。而會出現這種無止盡追尋的無限迴圈，通常是因為無法得出結論，或是想找出支持自己論點的訊息，但這兩種都不是很正確的心態，解盤的精神就是理性和客觀的均衡拆解，透過自己的判斷，然後做出結論，不管最終是對是錯，那才是屬於自己的東西。
