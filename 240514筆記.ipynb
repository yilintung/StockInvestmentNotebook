### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤後快訊】小幅過高，策略不變  
加權指數出現過高的上漲，接下來呢？要區分成盤中跟盤後的觀察。昨天加權指數雖然過高，但從上漲與下跌檔數觀察，加權指數下跌檔數580檔，高於上漲的535檔，櫃買指數的比例更加懸殊，上漲僅有377檔，下跌高達603檔。  
因此指數雖然過高，但資金過度集中在航運、橡膠、半導體、貿易百貨，多方氣勢稍嫌不足，投資人也沒那麼有感。  
接著是盤後條件，技術面突破三要件：漲幅、量能、時間，由於原本指數已經高達20883，如果以傳統的條件，要漲超過3%才算是足夠，必須高達600點，所以稍微放寬，先以450點為目標，昨天上漲148點，所以還有300點需要努力。  
昨天是指數突破的第一天，至少觀察到周末，再來做確認也不遲，最重要的是成交量只有4344億，僅小幅突破10日均量4164億，還是稍嫌不足，而且補量是好事，但也意味著只能前進不能後退，否則突破量會變成套牢量。  
研究指數多空強弱，只是為了操作策略的研擬，所以即便昨天指數小創新高，但操作策略不變，維持3-5成資金偏多操作，靜待指數完全站穩，再放大資金。 (出處：K線捕手楊忠憲粉絲團)
