### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】前波支撐，短線有守  
昨天加權指數10點半之前，還維持平盤上下震盪，10點半過後急轉直下，快速出現修正的走勢，成交量3738億，終場下跌148點，收在16001點。櫃買指數有小跌破盤整區間的現象，有迷你頭部的雛型。  
上課曾經教過，最弱勢的反彈型態就是橫盤，而最短的反彈天數就是3-5天，基本上加權指數走勢符合這兩個現象，只橫盤整理3個交易日又再度破底。  
雖然今天盤中修正力道很猛烈，但也因為如此，在低迷多日之後，昨天盤後終於可以跟大家說一些好消息，「在周末就已經規畫，第一個小支撐在15950，基本上就現在位置，所以多方稍微可以喘息一下，雖然還不是終點站，但至少再來不至於立刻連續下跌。」  
加上下跌有量，會逼出一些不穩定的買盤，有利於短線籌碼穩定，從夜盤期指也稍微出現止穩現象，甚至還能小漲64點，未來弱勢整理區間應該會落在16000-16075，稍微強勢一點則是16000-16200。  
短線高手當然可以搶一下多方個股的價差，但因為外資期貨空單已經攀升到15297口，已經轉為空方思維，而且投信雖然持續買超，但投信期貨也有避險空單，而技術面的120日線一周後會下彎，長期趨勢容易助跌。  
因此指數只是短線區間整理，頭部 (藍色區)明顯大於支撐(紅色區)，之後仍有續跌壓力，大膽一點(不一定準)的假設可能就在一周之後。而且更大的重點，可能要放在櫃買指數，因為昨天小破迷你頭部支撐，支撐相對不穩定。 (出處：K線捕手楊忠憲粉絲團)
