【[休息是為了走更長遠的路](https://www.facebook.com/kevinyang20180801/posts/pfbid0LpaysZ4DZ68aHadZgR3sVfo8Yh2aAC8ujZ17m9NRHsDi6kYZFMu9YMsavVSu4GR3l)】  
多數散戶會習慣每天都進行交易，通常是因為分不清楚行情好壞，然後在連連虧損之後，最後發出嘆息甚至抱怨：最近好難做！當行情不好時，本來就會增加操作的挫敗感，長期以往信心會被消磨殆盡，我的經驗中，比較常出現行情不好的原因有以下三種：  
1.指數趨勢走空：當加權指數或櫃買指數趨勢翻空時，以20日線為趨勢工具，當20日線跌破下彎，空方個股比重會高於8成，而即使主動式選股能力很強，失敗率都還是很高，何況是多數散戶都是隨機選股、感覺選股，當然怎麼買怎麼錯。  
2.個股輪動快速：指數位階偏高、趨勢走空前或是盤整的初期，由於投資人大多數都是獲利，容易抱不住股票，產生個股漲跌輪動快速，波段上漲的股票相對少，昨天漲今天跌，或是昨天弱，今天又轉強，一天多，一天空，短線交易能力不夠強，容易追高殺低。  
3.沒有主流族群：這邊指的主流必須是市場聚焦的大族群，不是小族群或是次族群，一個比較紮實的多方結構至少必須有4個以上的主流族群，這樣市場才容易聚焦，選股也會比較容易。  
目前屬於第二種和第三種，雖然趨勢尚未翻空，但個股輪動已經很快，而且缺少連續性波段上漲個股。在股市中養成休息的習慣，不只為了走更長遠的路，更是一種美德，適時地停下腳步，不只可以避開風險，更可以讓自己對行情判斷更加理性客觀。  
備註：隨機選股就是沒有嚴謹的選股濾網，看到、聽到、感覺好的股票，就想要買，這樣基本是就是在賭運氣。但以我個人為例，股票是從200檔個股中，先初步篩選出50檔，再從50檔精選到剩下10-15檔，最後盤中從這15檔去挑選有技術面發動現象的個股，自然勝率就會高。
