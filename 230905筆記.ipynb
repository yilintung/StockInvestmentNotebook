【解盤與猜盤】
其實一般人在解盤，常常犯的一個問題就是做了太多的假設和分析，但其實這些假設和分析，只會引發更多的假設跟分析，變成過度億測，而導致距離希望的研究結果，越來越遠。如果你仔細看我解盤，我根本不會在這些事情上鑽牛角尖，因為根本不會有答案，即使有答案，也對你的交易沒有絲毫的幫助。  
但為什麼一般散戶會養成這樣的習慣呢？因為在過去的人生經驗中告訴你，不打破砂鍋問到底，就沒有辦法有結論或是找到解決問題的方法。可惜事實上股市中並非如此，因為你要的結果，早早已經呈現在指數或股價和技術分析之中，你只要知道這樣的現象你該做甚麼交易就可以。  
看趨勢、抓轉折、賺價差，這才是投資人該專注研究的方向和事情。而不是花時間在猜盤。一直以來，你以為你在「解盤」，其實你只是在「猜盤」。其實從事股票教學，技術分析的本身是最好教的，因為KD不會，講完你就會了，乖離率不會，我教完你也會了。  
一次不會兩次不會，第三次總是會學會，但最難的是心性和觀念，也就是認知，是最難教，也最難學的的。因為這些根源每個人的成長經驗、人格特質、年齡、性別、是非判斷、風險控管...，都是透過很長的時間才養成，改變這些對教和學的人，都是很大的考驗。所以前輩們才重複地說「股票最後考驗的是人性」，還有「股市是對人性的總檢討」。  
  
【不要成為股市中的工具人】
媒體原本的天職應該要報導客觀的真相，但是在金融市場中，常常會出現有心人刻意透過媒體，故意操弄輿論，蓄意散布加工的、特定的，甚至不實的資訊，而多數的投資人也在這樣過程中，被操弄和愚弄而不自知，只會盲目地相信，甚至不自主地受控，爭先恐後進場買賣股票，而這樣的瘋狂的現象，每天都在金融世界上演。  
遙記多年前的3D列印，後來的元宇宙，乃至於到近幾年的天國一輝、航運鋼鐵、防疫概念股、航空貨運、IC設計、解封內需，到最近熱門的AI概念股和超導體。市場永遠有源源不絕的題材  
比如當大家在看法人或媒體發布的觀點，要有一些基本的分析能力，首先單一個法人發布訊息，那只代表單一法人看法，不是所有法人看法，更不是整個市場的看法。其次，法人看法不一定是真的或是對，你可以回溯和統計，該外資過去發訊息的準確度，來評估未來的有效性。  
第三點，即使法人發布消息是真的，也不代表股價必然會漲或是會跌。最後你還要回去看釋放利多的目的，利多發佈時，股價位階是高是低，如果高檔放利多，而利多不漲，那可能反而要小心。所以外資的訊息，只是眾多參考資訊中的一個而已，不要過度神化，要理性客觀的看待。  
基本面跟產業面，多時候只是主力作手說服你買賣股票的工具，如果全部都照單全收，那就會成為股市裡的工具人，今天很得意抱住股票的原因，日後就會成為賣不掉的癥結。如果只因為基本面跟產業面很好，就盲目買進股票，很可能就會成為被主力說服，甚至出貨的對象其實不論任何從管道得知的資訊，投資人本來就應該理性和客觀分析，而不是人云亦云、隨波逐流，全部照單全收，所有市場上流傳的利多和利空，都要學會獨立思考和判斷，再決定交易行為，這些資訊都是選股的參考，但究竟要怎麼買跟賣，那才是賺賠的關鍵，不要一再重演這些悲劇，成為被有心人操弄的工具人。
