【[其實崩不崩盤，一點意義也沒有](https://www.facebook.com/kevinyang20180801)】  
當單日跌幅較大時，很多同學會很緊張的問，股市會不會崩盤？其實這是個很難回答的問題。因為光是崩盤二字的定義，本身就是很大的問題，會因為交易周期跟商品，會產生很大的歧異。  
對於做期貨和當沖的人而言，指數一天跌300點，可能就已經算崩盤了，因為已經賠到連爸爸媽媽都不認得了。對於做現股，可能要跌1500點以上，才會認為這是崩盤，如果用國際股市和指數的定義長線走空，那可是要跌20%以上，換成台股現在就是要跌4000點。但如果對於真的做長線或是價值的人而言，可能指數必須腰斬，跌破一萬點，才認為是崩盤。  
所以你不大會看到我解盤時，用崩盤這樣的名詞，因為這對不同類型的投資人，解釋的方式會有很大的不同，對應的策略差異更大。  
不過有一個很容易判斷的事情，也很重要的事情，倘若你是做價差的投資人，其實每天股價的漲跌，都是對自己的壓力測試，如果某一天指數跌300點，你就承受不了，那表示持有的商品，可能早就已經起跌，甚至出現停損或賣出條件，心理面提醒你該賣出了，這時就該紀律執行，即使後面指數再跌1000點、3000點、5000點，手上根本就沒有持股，自然也不會受傷，所以崩盤事實上跟你也沒關係。 (出處：K線捕手楊忠憲粉絲團)
