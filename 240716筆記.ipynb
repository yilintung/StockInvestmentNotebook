### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線盤整，市場觀望  
昨天盤前提到川普公開演講發生槍擊事件，不過「政治新聞」不等於「政策新聞」，所以對整體股市影響應該不至於太大，加權指數和櫃買指數昨天波動都不大，陷入低量盤整的格局，觀望氣氛濃厚，原本漲多股幾乎都休息，低位階個股相對表現，接下來近幾日可能都是類似格局，市場可能都在等待台積電法說會。  
只是法說會大家反而是要留意的潛在風險，因為大家普遍都是一片看好，而股價也已經反映到相當高的位階，如果沒有更多的新亮點，只是符合預期，反而要留意賣壓，事實上謹慎一點的投資人，不管好壞肯定先跑一趟再說，所以不少台積電相關題材的個股，上周五就先陸續修正了。  
再者，市場對這個超級資優生充滿高度期待的時候，是不允許有瑕疵的，先不要說有甚麼大的利空，萬一公布了一絲一毫的保守言論，就會變成重大利空，那時候就要更小心市場的劇烈反應。  
就如同是場普遍都把降息當成利多，但問題來了，假如降息是利多，一直以來的升息不就是利空嗎？那為什麼近兩年來，全球卻在在這個重大利空之下屢創新高，因為在眾多上漲的因子中，早就已經在反應預期降息的利多，所以一旦真的降息，反而可能變成利多出盡。  
利多利空的轉換都在一線之間，現在屬於內資軋空和軋空手的行情，不論對象是外資或散戶，但這樣的情況不會長期持續下去，強勢軋空結束後，隨之而來的是猛烈的修正，短時間只要支撐和轉折沒有出現(可以直接觀察10日線)，暫時還不用過度恐慌，只是接下來一兩周警覺性要高一些。  
延續昨天策略，我沒有做空的習慣，目前也不會急著立刻全部出清持股，會採取漸進式減碼策略，也就是剩餘5成資金部位，個股如果出現賣訊，就依照條件做離場(事實上昨天有些持股已經出現離場訊號)，離場後的部位暫時先保持現金，不會立刻積極再回到市場。 (出處：K線捕手楊忠憲粉絲團)
