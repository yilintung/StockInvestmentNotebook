### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】止穩有餘，仍在空方  
加權指數跟國際指數周一急跌重挫，但也因為跌勢過快，多數都已經到波段跌幅滿足區和支撐區，雖然空頭看壓力不看支撐，但要止穩還是非得要先有支撐不可，昨天提到「短線因為急跌，已經到了前低、上升趨勢線、黃金分割率0.618的複合式支撐」，所以初步止跌條件是具備的。  
但趨勢要由空翻多，至少必須站回短線壓力20800，才代表短線轉強，昨天收盤為20501，所以暫時沒有改變短線空方趨勢，而且即使未來站回20800還要伴隨台幣升值、外資期貨空單降至2萬口以下、4個多頭類股翻多領軍、成交量放大，否則上攻還是容易曇花一現。  
解盤不能只看一天漲跌，要看趨勢、位階、型態、支撐、壓力、轉折，這樣才能看到盤勢的全貌。短線止跌有餘，短中期空方警報並沒有解除，往上套牢反壓仍大(藍色)，甚至還是有可能往長期空翻空的方向前進，尤其月KD指標已經滑落鈍化區，加上國際指數多數都在歷史高位階區間，所以投資人切勿因為一天止穩就掉以輕心。  
原本空手的人昨天進場才叫做搶反彈，但原本滿手股票套在高檔的人，昨天進場叫做攤平。同樣的交易行為，在不同的個體條件，會產生不同的意義。搶反彈是抓低點轉折，買到成本最低的時候，未來漲多漲少都是賺，隨時可以離場。攤平交易是高檔沒有減碼，持有滿手股票，低檔持續買進試圖降低平均持有成本，倘若後續彈幅不大，未到解套成本，那反而會越套越深。  
急跌後的短彈是最難操作的，不論做多做空都一樣，做多，不知道能彈到哪？轉折沒抓準，一個不小心就續跌。做空，真的急跌太深了，所以彈幅也可能更大。最好是多看少做，讓她多走幾天，會更明朗一些。 (出處：K線捕手楊忠憲粉絲團)
