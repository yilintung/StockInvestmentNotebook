【真正的決戰，永遠是在空頭】  
大多頭時，任何投資策略都會賺錢，當沖、追強勢股、做波段、定期定額、長期投資、算籌碼、看基本面、看產業，但真正的決戰永遠是在空頭，多頭是分不出勝負的。  
股票或指數下跌時，你究竟知不知道該怎麼辦？有沒有停損機制？有沒有在第一時間停損？有沒有正確的停損方法？還是持續逢低攤平定期定額？你有這麼多錢嗎？還是已經ALL IN套在高點？真的沿路崩跌你敢買嗎？  
你是不是又想說，有一天空頭會變多頭，當然是沒錯，但最初的高點可能10年、20年，甚至更久都不會再回來了。這時你又該怎麼辦？如果你對你相信的投資策略和交易系統沒有徹底的理解，甚至不知道該方法的缺點、風險和致命傷在哪邊，這是很危險的事情。  
買了股票之後，股票上漲當然很快樂，但萬一下跌，很容易就會從原本信心滿滿，變成恐慌和焦慮。而投資前本來就先預想好最不好的情境，一旦發生你該怎麼辦？而不是盲從，甚至催眠自己，只要買著、放著不管她，有一天就會變成有錢人。
