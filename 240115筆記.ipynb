2月5日為最後一個交易日，2月封關前僅剩下3個交易日，2月15日才會重新開市，因此與國際股市會有9日脫鉤的空窗期，因此投資人更為關心的是究竟要不要抱股過年？倘若是做長線的人，其實就不用考慮，因為本來就是繼續持有！但假如是做波段價差的投資人，能預知休市期間，國際上會發生甚麼重大政經事件？  
台灣會不會像沙烏地阿拉伯，突然發現價值155兆黃金礦脈？還是疫情又重新爆發？這樣想就容易理解了，結論是根本沒有人能夠知道。簡單的說，抱過過年，偏向賭徒行為，只是賭年後的多空。但真有必要去賭這件事情嗎？會漲的股票也不會只漲一天，會跌的股票，也不會只跌一天，封關過年後回來，再進行回補即可。明明可以握著現金，很安穩的吃年夜飯，跟家人團聚，為什麼一定要把自己逼到心驚膽跳的狀態呢？
