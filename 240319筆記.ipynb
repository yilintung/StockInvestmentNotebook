加權指數是台股所有個股的加總的表現，假如撐起指數的只有特定的幾檔權值股，那意味著資金沒有平均分配到1200檔，不是每檔在加權指數掛牌的個股都受惠。投資人倘若沒有層層篩選的嚴謹濾網，只是採隨機選股的方式，那要選到續漲個股的機率可能是百分之5以下。所以在這種時候，只要跟緊這些相關控制指數強勢權值股，或是與指數連動的ETF，其實會相對輕鬆。
