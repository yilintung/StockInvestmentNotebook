### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】單日上漲，反彈架構

不過解盤不能只看一天的漲跌，而是必須把趨勢、位階、型態，支撐、壓力、轉折，都納入觀察，才能客觀且完整的判斷整個局勢變化。  
既然上漲是可預期的，那就不用過度興奮，或是認為趨勢已經逆轉，尤其昨天加權指數和櫃買指數雖然都同步上漲，但是成交量能明顯萎縮，屬於「價漲量縮」的反彈格局。  
其次是短線關鍵壓力區，加權指數為22867，而櫃買指數在頸線位置，也都還沒有有效攻克，所以從技術面來說，趨勢仍然沒有改變。不過籌碼面稍微有改善，倒是一件好事。  
從類股結構和均線多空比觀察，空方個股的比重也一樣，遠高於多方個股，所以並沒有因為一天的上漲，而改變結構上的問題。航運類股和橡膠類股連日上漲，不過已經接近中長期前高，接下來短線容易出現震盪。  
所以整體的風險係數和操作難度，並沒有太大的改變，如果有出現比較明確的多方轉折或次翻多條件，先賣個關子，倘若出現會再告訴大家，但至少昨天還肯定還不算。 (出處：K線捕手楊忠憲粉絲團)
