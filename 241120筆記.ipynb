```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import mplfinance as mpf
import numpy as np
```

```python
##### 使用2023年7月7日到今日(2024年11月20日)收盤之日線圖資料 #####

# 載入從「好神通PLUS」輸出的Excel檔
df = pd.read_excel('./data/SINO_2332_241120.xlsx')

# 保存K棒的基本資訊(開、高、低、收、量)
df_k_line = df.drop(columns=['SMA5','SMA10','SMA20','SMA60','SMA120','SMA240','MA5','MA10'])

# 保存價的均線
df_sma = df.drop(columns=['開盤價','最高價','最低價','收盤價','成交量','MA5','MA10']) 

# 將K棒的Columns的名稱由中文改為英文
df_k_line = df_k_line.rename(columns={'時間':'Date','開盤價':'Open','最高價':'High','最低價':'Low','收盤價':'Close','成交量':'Volume'})

# 將價的均線的Columns的名稱由中文改為英文
df_sma = df_sma.rename(columns={'時間':'Date'})

# 將K棒的Date設為Index
df_k_line.set_index(df_k_line['Date'],inplace=True)
df_k_line = df_k_line.drop(columns=['Date'])

# 將價的均線的Date設為Index
df_sma.set_index(df_sma['Date'],inplace=True)
df_sma = df_sma.drop(columns=['Date'])
```

```python
def date_to_index(df_in,date_in) :
    number_array = df_in.index == date_in
    idx = 0
    for number_index in number_array :
        if number_index == True :
            break
        idx = idx + 1
    return idx
```

```python
# 設定頸線
neckline_start_date = '2023-09-15'
neckline_end_date = df_k_line.iloc[-1].name.strftime("%Y-%m-%d")
neckline_x = [date_to_index(df_k_line,'2023-12-04'),date_to_index(df_k_line,'2024-11-08')]
neckline_y = [df_k_line.loc['2023-12-04']['Close'],df_k_line.loc['2024-11-08']['Open']]
slope,intercept = np.polyfit(neckline_x,neckline_y,1)
neckline_start_price = slope * date_to_index(df_k_line,neckline_start_date) + intercept
neckline_end_price = slope * date_to_index(df_k_line,neckline_end_date) + intercept
breakout_date = '2024-11-19'
breakout_price = slope * date_to_index(df_k_line,breakout_date) + intercept
print('頸線開始價格 ＝ {:.2f}元 ， 頸線突破價格 ＝ {:.2f}元 （ 今日收盤價 ＝ {:.2f}元 ） ， 頸線結束價格 ＝ {:.2f}元'.format(neckline_start_price,breakout_price,df_k_line.iloc[-1]['Close'],neckline_end_price))

# 估算目標價
bottom_price_date = '2024-08-07'
bottom_price = df_k_line.loc[bottom_price_date]['Open']
on_neckline_price = slope * date_to_index(df_k_line,bottom_price_date) + intercept
target_price = (on_neckline_price - bottom_price) + on_neckline_price
print('底部最低價格 ＝ {:.2f}元 ； 估算目標價 ＝ {:.2f}元'.format(bottom_price,target_price))
```

```python
# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定頸線
seq_of_seq_of_points=[
    [(neckline_start_date,neckline_start_price),(neckline_end_date,neckline_end_price)],
    [(bottom_price_date,bottom_price),(bottom_price_date,on_neckline_price)],
    [(breakout_date,breakout_price),(breakout_date,target_price)]
                     ]

# 設定均線
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.5,color='xkcd:maroon'),
    "SMA20": mpf.make_addplot(df_sma['SMA20'],width=1.0,color='xkcd:violet'),
    "SMA60": mpf.make_addplot(df_sma['SMA60'],width=0.5,color='xkcd:dark orange')
    
            }

# 繪製K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), volume=True,addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=(1.0,10,10), colors='xkcd:orange yellow', alpha=0.6), warn_too_much_data=1100)
mpf.plot(df_k_line,**kwargs)
```

```python
### 轉折點驗證 ###

import myutils

# 尋找轉折點
_,_,max_min = myutils.FindingTurningPoints(df_k_line,'close',real_body=True,order=10)
pd.set_option("display.max_rows", None)
display(max_min)

# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定均線與轉折點    
turning_points_len = len(np.array(df_k_line['Close']))
turning_points = np.array([np.nan]*turning_points_len)
for point in max_min.iterrows() :
    turning_points[point[0]] = point[1]['Price']    
apds = [
    mpf.make_addplot(turning_points,type='scatter',marker='o',markersize=50,color='xkcd:sky blue'),
    mpf.make_addplot(df_sma['SMA5'],width=0.5,color='xkcd:maroon'),
    mpf.make_addplot(df_sma['SMA20'],width=1.0,color='xkcd:violet'),
    mpf.make_addplot(df_sma['SMA60'],width=0.5,color='xkcd:dark orange')
]

# 繪出K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), addplot=apds, warn_too_much_data=1100)
mpf.plot(df_k_line,**kwargs)
```

```python
### ZigZag驗證 ###

# 尋找ZigZag
_,_,max_min = myutils.FindingDirectionalChangePoints(df_k_line, mode='close')
display(max_min)

# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定均線與轉折點    
turning_points_len = len(np.array(df_k_line['Close']))
turning_points = np.array([np.nan]*turning_points_len)
for point in max_min.iterrows() :
    turning_points[point[0]] = point[1]['Price']    
apds = [
    mpf.make_addplot(turning_points,type='scatter',marker='o',markersize=50,color='xkcd:sky blue'),
    mpf.make_addplot(df_sma['SMA5'],width=0.5,color='xkcd:maroon'),
    mpf.make_addplot(df_sma['SMA20'],width=1.0,color='xkcd:violet'),
    mpf.make_addplot(df_sma['SMA60'],width=0.5,color='xkcd:dark orange')
]

# 繪出K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), addplot=apds, warn_too_much_data=1100)
mpf.plot(df_k_line,**kwargs)
```

```python
### 頸線驗證 ： 使用自動趨勢線 ###

# 以自動趨勢線做為頸線
trendlines,_  = myutils.FitTrendlines(df_k_line['2023-10-31':'2024-11-18'])
trendlines_start_index = date_to_index(df_k_line,'2023-10-31')
neckline_slope = trendlines[1][0]
neckline_intercept = trendlines[1][1]

neckline_start_price = (neckline_slope * (date_to_index(df_k_line,neckline_start_date)-trendlines_start_index)) + neckline_intercept
neckline_end_date = df_k_line.iloc[-1].name.strftime("%Y-%m-%d")
neckline_end_price = (neckline_slope * (date_to_index(df_k_line,neckline_end_date)-trendlines_start_index)) + neckline_intercept

# 估算目標價
bottom_on_neckline_price = neckline_slope * (date_to_index(df_k_line,bottom_price_date)-trendlines_start_index) + neckline_intercept
target_price = (bottom_on_neckline_price - bottom_price) + bottom_on_neckline_price
breakout_on_neckline_price = neckline_slope * (date_to_index(df_k_line,breakout_date)-trendlines_start_index) + neckline_intercept
print('突破時頸線價格 ＝ {:.2f}元 （ 今日收盤價 ＝ {:.2f}元 ）， 底部最低價格 ＝ {:.2f}元 ； 估算目標價 ＝ {:.2f}元'.format(breakout_on_neckline_price,df_k_line.iloc[-1]['Close'],bottom_price,target_price))

# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定頸線
seq_of_seq_of_points=[
    [(neckline_start_date,neckline_start_price),(neckline_end_date,neckline_end_price)],
    [(bottom_price_date,bottom_price),(bottom_price_date,bottom_on_neckline_price)],
    [(breakout_date,breakout_on_neckline_price),(breakout_date,target_price)]
                     ]

# 設定均線
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.5,color='xkcd:maroon'),
    "SMA20": mpf.make_addplot(df_sma['SMA20'],width=1.0,color='xkcd:violet'),
    "SMA60": mpf.make_addplot(df_sma['SMA60'],width=0.5,color='xkcd:dark orange')
    
            }

# 繪製K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), volume=True,addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=(1.2,10,10), colors='xkcd:orange yellow', alpha=0.6), warn_too_much_data=1000)
mpf.plot(df_k_line,**kwargs)                            
```
