### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】指數盤整，個股操作  
國際股市目前多數維持高檔盤整，而德國創歷史新高，目前是全世界中的強勢指標，可以做為未來趨勢的觀察，究竟是成為領頭羊，其他指數跟進創高？或者僅只有德國過高，最後遇壓拉回？整體來說，國際面、技術面、籌碼面、資金面，觀望氛圍濃厚，尚未決定多空。  
整體籌碼中性偏空，但成交量要沒有放大到3500億以上，所以仍然沒有脫離11月21日以來，長達13天的狹幅盤整，中期則是以16800上下500點的箱型區間，趨勢尚未明朗。多數類股指數也同樣進入盤整，多方族群集中在IC設計、鋼鐵、紡織較為集中。  
先前提到的背離只是波段過熱，但背離現象無法確認，後續究竟是盤整或是修正，需觀察短中期轉折是否出現。指數上有鍋蓋、下有鐵板，尚未分出趨勢，但悶鍋的蓋子一旦打開，或是鐵板一旦跌破，多空趨勢會走得相當快。本周是多方表現最好的時間點，因為再過5天20日線就可能走平，就會出現下彎風險。  
投資分成幾個種類，當該商品趨勢由空翻多，反轉向上時，順勢作多，這是多數人都知道的。第二種是當趨勢盤整或是轉弱，將作多部位減碼。第三種是當空頭趨勢確立，反轉向下時，出清做多持股，或尋找對應的反向ETF或是衍生性商品，順勢作空。現在屬於第二種，只是做多保守，維持中低持股或減碼，但並未完全空手，或是轉為做空。 (出處：K線捕手楊忠憲粉絲團)

[【過高賣壓】追強勢股留意風險](https://www.facebook.com/kevinyang20180801/posts/pfbid0b4LLG1Tj4etMTPRUiAWTSLEw86ni8a9hrHpHhqfg6y8VU43DTxheXqRjWKmKxP5zl)  
當指數或是個股，在相對高位階，突破長天期或是關鍵壓力，甚至創歷史新高時，往往會出現來回震盪測試支撐的現象，這是甚麼原因呢？
因為位階高檔，原先有許多追高短套的投資人，一旦解套就會急著想要趕快停利，而原先成本低的投資人，由於獲利頗豐也會想賣。所以解套的人想賣，和獲利豐碩的投資人也想賣，這一來一往之間，就會出現股價或是指數的震盪。
但這種現象是很正常的，尤其是創歷史新高的股票，必須等到震盪整理，但前提是整理期間，不能出現爆量長黑或長上影線，當然關鍵支撐更不能跌破，否則又會變成高檔的價突破，反而是危險的現象，等到籌碼良性換手之後，才會蓄積能量再續攻。
因此追強勢和高位階的股票並沒有錯，多頭做多漲追漲不追跌，但必須謹慎留意單一K棒轉折、量能變化，以及重要支撐。
