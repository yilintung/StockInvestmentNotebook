### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】夜盤續弱，加權補跌  
前天盤中就提醒「已經接近保守版本的漲幅滿足，容易有漲多拉回的現象」，以及「持股中高水位的投資人，接下來應該多思考怎麼賣，不要過度積極買」，加權指數連續兩天轉弱，昨晚夜盤期指續跌174點，小跌破5日線，直接測試10日線，今天開盤可能就要先來測試這個支撐。  
指數漲多出現震盪，同學的提問難免就多了起來，但不外乎是「老師，您說保守的漲幅滿足到了，那台股漲勢會結束嗎？」、「還是有可能往樂觀的漲幅前進？空間還有多少？」  
首先，保守滿足大約是22000，樂觀滿足大約是22950，為什麼加上大約呢？因為現在指數位階高，已經超過2萬點，鐘擺效應會更加明顯，所以誤差50點，甚至100點都在合理範圍，所以沒有像算個股滿足區那樣的錙銖必較。  
以周一高點21937來說，基本上跟22000意思是一樣的，先前提到過關鍵轉折基本上就是觀察5日線和10日線死亡交叉，所以即使今天補跌測10日線，應該也還沒出現上述轉折條件。  
轉折條件沒出現之前，都還算是漲多回檔，不是短線翻空或是主力出貨的現象，未來還是有機會往樂觀滿足區前進，但昨天早盤還是提醒「持股比重高的真的投資人要預防性的漸進式調節，不要滿持股撐到最後一天」  
多頭小回測支撐是很好的機會，一來可以測試持股強弱，二來可以看看自己心理壓力會不會很大，如果指數只是這樣小回就覺得受不了，那表示可能追高或是買太多了，應該要提早有些動作。 (出處：K線捕手楊忠憲粉絲團)
