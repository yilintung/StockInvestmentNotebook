### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】加權盤整，櫃買表現  
昨天盤前提到「有機會再度挑戰16650壓力，甚至上週四收盤位置16770」，但同時也提醒「短線有機會轉強，但中期趨勢還沒這麼快改變，沒意外的話，偏多方發展，但建議多觀察一兩天」  
加權指數開高走高，但是由於量能只有2497億，10點過後出現遇壓拉回的現象，最後上漲96點，收在16719點，剛好在16650和16770之間。接下來由於依照扣抵計算，季線有20個交易日會持續下彎，而月線在過5天反而有機會上揚。  
中期均線一上一下，代表中期盤整，因此短線不容易立刻脫離箱型格局，變化不會太大，反而應該把重心放在櫃買指數，因為已經出現短底，短期趨勢轉強，未來3-5天中小型股容易有表現，從櫃買指數的類股分布也可以看的出來，昨天普遍是收漲居多。  
歐美股市表現普遍穩定，雖然中期空方趨勢尚未化解，至少維持高檔箱型整理，短線風險暫緩，至於最弱勢的陸股跟港股，都還沒有止穩轉強的現象，中期頭部現象和風險仍在。  
昨天早盤只挑了5檔觀察股，就有4檔漲停，雖然只是次主群，但選股難度已經降低，也符合昨天提到的策略：積極的投資人可以少量資金搶短，但保守的投資人可以繼續追蹤即可，不用急著進場，可以靜待中期趨勢明朗。 (出處：K線捕手楊忠憲粉絲團)
