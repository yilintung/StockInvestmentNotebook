【學習要無所求】

無所求並不是甚麼都不追求，而是在學習過程中，我們都不會在意過程中的賺賠，因為不論賺賠，都很平常心看待，因為那都是必要的過程。只要知道自己是往對的方向前進，每天都有進步，讀小學讀完讀國中，國中讀完讀高中，最後自然就會上大學，重點是你在每一階段學習的時候，有沒有完全專注，有沒有做到我說的每一件事情，上課有沒有認真，課後有沒有複習。  
我們都清楚的知道，等方法學會系統建立，把事情做對，自然就會賺錢。但不是一開始要逼自己要賺錢，那不是可能的事情，甚至有時把自己逼急了，反而事半功倍。當學生時代讀書學習的時候，你會在意你要賺多少錢嗎？會要求百萬年薪？只知道要按部就班完成學業，以後就會有機會，不會在意短期間的賺賠，因為我們看的是更遠大的未來  
從在網路上分享資訊開始，到現在快要8年了，就先以7年做計算，一年粗估240天，解盤到現在7年左右，有1680個交易日，只有2天沒有解盤。個股建診一周一次，一個月粗估4.5次，7年下來約403次，只有1次放假。一次保守估計100檔，所以總共看了4萬多檔個股的K線圖。我用自己的生活作息，證明給各位看持之以恆和紀律的重要性，股市是人性的總檢討，我只是把在生活中的態度跟習慣，轉換到股市投資和交易。  
其實有時候同學會問我，怎麼選股？怎麼找買賣點？事實上這個問題就像學習英文，同學問：老師，我甚麼時候可以跟外國人對答如流，快速的閱讀文章，寫出完整的英文作文。其實，表面上只是一個問了一個問題，但其實已經是英文學習的全部。要能夠做到上面的狀態，首先要學會英文ABC大小寫、背單字從50個、100個、500個甚至超過1千個，接著英文文法該理解的要理解該記得要記，然後還要想辦法多看英文電影，接著是無數次聽力練習，還有實際寫作，找機會和外國人說話聊天。  
其實學股票投資也是一樣，最最基礎的技術分析和籌碼工具，你必須要會，不只要懂，還要背起來，熟才有機會生巧，如果只是一知半解，有了基礎之後再往進階的單元前進，最後將所學融會貫通，進入實戰演練，從小賺大賠變成小賺小賠，最後做到大賺小賠。  
所以，同學的問題就是，你們可能還在小學階段，卻問了大學的問題，但你們基本功不紮實的時候，我講了你也聽不懂，即使聽的懂一次，那也是一次的。就像你聽不懂某一句外國人說的話，你來問我幫你翻譯，但下一句你還是不會，因為你單字背的不夠多、文法不熟練...  
最後，請給自己多一點時間，如果學會這個技能，能夠讓你一生衣食無憂，你願意花多久的時間學習，以前說當學徒要「三年四個月」，但是我規劃有系統的學習，只要你願意用功，絕對可以大大縮短時間，以及減少犯錯走冤枉路的時間，我看過一個學生1年2個月密集學習，就辭掉科技夜300萬年新的工作，取得人生的自主權，應該已經退休快5年了。  
要不是被你們逼到了，我實在不想講這種故事，因為有人學很久，還是沒學起來，關鍵是你的求富之心夠不夠很強烈？有沒有下定決心。不是賠錢就叫做苦，真的要苦學，苦讀...把她當成考證照、考國考、考聯考...那種態度，我在那個同學身上看到10多年前的自己，所以我知道他會成功。  
你們現在的條件，跟我以前比起來，都還在快樂學習狀態，不知苦為何物學，我以前上了20個老師的課，只有3個比較正常，而那3個老師教的濃縮再一起，還不如我教你們3堂專題課程的豐富跟完整，在金融市場賺錢，絕對不是投機或是不勞而獲，他跟任何專業技能一樣，都需要紮紮實實的硬功夫。與各位共勉之，還有我不一直教下去的，我還是想留點時間給自己和家人，有幫自己規畫教學退休的時間，你真的想學真的要好好把握時間。  
雖然有點長，但是當你學習遇到瓶頸，花幾分鐘看一下，會有幫助的~

當解盤的時候，希望每個時間點能夠盡如人意，能夠完全掌握未來走勢，但以我的經驗，實務上大概分成幾種層次和比重，我依照不同情況，用不同顏色做標示。  
第一種，是非常有把握，用紅色標示，甚至可以知道會漲到哪邊？或是跌到哪邊？這種時間在一整年的時間，大約佔75%的時間，可以依照判斷的趨勢做積極交易。(紅色區間)  
第二種，對多空有大概的方向，但並不是很有把握，用橘色標示，這種時間一年之間，大約佔15%。第三種是對多空方向完全沒把握，也沒有頭緒，用藍色標示，這種時間在一整年大約佔10%。  
目前這段區間屬於第二種橘色，心理大概有方向，但把握度只有70%，還沒有掉到第三階段藍色。
