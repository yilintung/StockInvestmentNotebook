### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線轉折，回檔修正  
加權指數上周二來到波段高點17421，與7月31日的前高17463，只有一步之遙，僅有42點的微小差距，但由於17463是波段前高、下降壓力線、價量累計圖的複合壓力區，加上正乖離率放大到4.59%，我自己的「漲停板指標」也出現。  
而且自11月1日以來，雖然指數短線上表現不錯，但結構上一直都有瑕疵，沒有出現市場的集體共識和主流，代表類股都是短多而且輪動快速，大大增加選股和操作難度，如此一來不容易持續性有效的讓投資市場聚焦。  
台股技術面昨天短線正式出現止漲現象，反彈的短多暫時宣告結束，意味著指數會進入修正，做多難度就會變高，整體架構又回到以16800為中軸，上下500點震盪，第一階段有效支撐，可以觀察月線以及箱型中軸16800。  
此外，櫃買指數上週四曾出現爆量長上影線之後，昨天也出現短線空方轉折，除了因為防疫題材，生技股獨強之外，多數的中小型股也出現賣壓同步轉弱。  
最關鍵的是投信的籌碼，從11月15日開始，罕見的反手為賣，之後開始出現連續小幅賣超，雖然還沒有大幅賣超，但很明顯的連續3個半月積極買超，已轉為觀望，要留意未來是否出現連續性大量調節，密切追蹤投信籌碼變化。  
而且全球原先最強的日本指數，也同步出現短線轉折開始拉回，未來投信籌碼以及日本走勢，會是相當重要的參考，直接影響台股指數修正的時間和空間。 (出處：K線捕手楊忠憲粉絲團)
