### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線強彈，中期箱型  
加權指數6月15日來到波段高點17346，出現長達15個交易日的波段修正。但是修正期間，外資雖然現貨積極減碼，但未平倉口數空單增幅不大，甚至後來還能維持少量多單，加上技術面的月KD指標，還能守在80以上的鈍化區，因此雖然趨勢短線轉空，但空方力量不強。  
緊接著7月11日起，國際股市轉強，加上台幣強升的助攻之下，法人買盤積極回補，成交量也放大到10日均量之上短短4個交易日，台股再度續攻，急漲超過600點，距離6月15日的前高僅相距63點，頗有挑戰前高的態勢。  
不過又要再度重新挑戰價量累計圖中的大量區間，先前6月15日就曾遇到這個大壓力區，無法有效站穩且突破，所以拉回整理。對多方有利的條件是國際股市偏多，尤其是美股4大指數，除了道瓊指數之外，都出現突破中期盤整區，短期續漲創新高的表現。而台股本身的籌碼面，三大法人短線積極回補，最後則是台幣強勢升值。但7月14日融資增幅過大，單日增加高達48億，要留意散戶過度積極，造成籌碼凌亂，而波段上漲最大的動能，來自台幣急升，但台幣也已經到中長期壓力區，要再持續強升有其難度。  
接著空方條件，第一是國際股市結構，雖然美股偏多，但原先最強勢的歐股和日股，明顯有利多鈍化漲不動的現象，通常在末升段會有強勢族群不漲，弱勢群補漲的現象，可以同步觀察陸港股是否出現補漲。先前波段上漲，挑戰高點失利，有一個很的因素就是指數的紮實度，雖然指數表面上強漲，但是事實上，類股指數並沒有跟上。維持多方續漲的類股只有電子、電腦及周邊設備、電子零組件，當然這之中主體其實就是最熱門的AI概念股，總計也只有3類。而進入盤整的類股有8類。空方包含反彈的類股有22個類股。  
要真正創高，讓中期脫離箱型，挑戰6月15日的17346，以及2022年3月30日的17770點，還需要3個關鍵條件加持。第一是依據道氏理論，真突破必須具備漲幅、量能、時間， 其次，是OBV量能指標，在指數創高的過程就算無法領先創高，或至少要做到同步創高，否則形成高的離的現象。最後的則是強勢的族群能否維持續漲 AI概念股，在指數攻高過程，是最核心的動能，也是人氣之所在，所以必須同步持續上攻，不能有高檔出貨的技術面現象，例如爆量長黑或長上影線。 (出處：K線捕手楊忠憲粉絲團)
