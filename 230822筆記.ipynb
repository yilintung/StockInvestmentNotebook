櫃買指數上周一由於20日負乖離率，已經來到6.15%，而且接近前波支撐和黃金分割率的0.382，短線容易有支撐，但從價量累計圖觀察，往上仍然反壓重重，204-210區間狹幅整理的機率較高。周末影音解盤有提到，國際股市短線有支撐，所以昨晚已經出現反彈，不過中期空方趨勢仍未改變。
