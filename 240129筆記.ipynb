### 【K線捕手盤前快訊】再過6日，封關過年  
上週國際股市表現，日本、費半、那斯達克短線轉弱，中期支撐有守，道瓊和標普500持續多方架構，陸港股止跌反彈，國際指數整體來說中性偏多。  
加權指數持續創高，櫃買維持箱型整理，外資期現貨做多，綜合條件逐漸改善，往多方前進。結構上，短線空方個股減少，但中期空方個股仍偏多。偏多族群只有有半導體、電子、電子通路、綠能環保，AI短線休息，但多數類股仍屬反彈或盤整。  
指數中性偏多看待，但考驗投資人的選股功力，二月五日最後交易日，包含今天再過6個交易日，即將封關過年，究竟要不要抱股過年？短線投資人短進短出，長線投資人也繼續持有，當然不用考慮這個問題，只有一種投資人需要思考，那就是操作波段價差的投資人。  
顧近6年農曆年後走勢：2018年、2023年箱型整理；2019年、2021年上漲；2020年、2022年下跌。上漲、下跌、盤整等3種走勢都出現過。投資人永遠在少賺與多賠之間取捨，所以不論多空，我始終不抱股過年，卻沒因此減少獲利，因為隨時可再回補，倒是因為這樣，2020年和2022年避開了下跌的風險。  
但是要清空持股及完全空手，要有很強大的專業和心理素質，以及配套方法，畢竟抱股過年的有其優缺點，反過來，不抱股過年也有其優缺點，必須依照自身情況跟條件，規劃適合自己的策略。 (出處：K線捕手楊忠憲粉絲團)
