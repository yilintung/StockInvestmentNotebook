短一點的觀察，這一波起漲大約是從2022年10月開始。接下來2023年，中間雖然有很多的波段進入盤整，但整體來說指數是越墊越高，因此整個趨勢都是偏多方發展的。只是這樣的走勢，到2023年第四季開始出現變化，大家應該明顯感受到選股和操作的不易，因為股價是不會漲到天上去的，終究有一天會有修正回檔，甚至走空的時後。  
但是因為長達1年多的大多頭，因此過去投資人即使有短套，只要基本面不要太糟，死抱活抱也會解套，甚至套牢幅度也可能不會太大，不管是研究基本面、籌碼面、技術面，甚至聽消息、不紀律、亂買、亂賣...，2022年底到2023年第三季都可能會獲利。投資人也因此會失去戒心，更不會想檢視自己的操作方法，究竟是靠專業和交易技巧賺到錢，還是只是僥倖跟到多頭的順風車。
