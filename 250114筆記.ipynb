同學問：「老師，這幾個星期你明明挑到好股票，怎麼忍得住不買呢？」  
其實，這要看你的投資心態進化到哪個層級：  
第一階段：懵懂無知，亂買亂賠  
什麼趨勢、什麼風險？完全不懂，只管衝進市場。這是許多散戶初入股市時的常態。  
第二階段：看得懂趨勢，但心癢手癢  
明知道空頭市場風險高，但不夠信任自己的判斷，硬是要下場試水溫，結果往往是賠了又後悔。  
第三階段：吸取教訓，學會忍耐  
吃過虧後，開始尊重趨勢和風險評估，懂得按兵不動，即使看到好股票也能忍住不買。  
第四階段：無欲無求，不再糾結  
不僅能判斷趨勢與風險，更因經驗累積，連買的慾望都沒有了。這已經不是「忍住」，而是徹底明白空頭市場不適合做多，自然不會心動。  
我就是第四階段，根本沒有忍的過程，因為壓根不想買。  
當你理解市場規律，尊重風險，投資就不再是一場慾望與理性的拉鋸戰，而是一場心境的升華。
