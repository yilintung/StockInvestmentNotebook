台股昨天比想像中弱勢，原先走勢尚稱穩定，還能維持在平盤上下整理，但早盤10點過後，加權指數和櫃買指數就同步出現修正，一路走低收盤幾乎收在最低，，期現貨的籌碼也都偏空。
盤前提到「台積電535箱型支撐」，昨天剛好收在這個點位，未來持續追蹤這個位置能否有守，緯創維持平盤附近，技嘉跌深反彈，但廣達昨天小創新低，整體AI族群仍然偏弱。  
反彈指的是空頭結構中的上漲段落，還可細分為短線反彈、中期反彈、長期反彈，目前屬於短線反彈。而反彈的基本條件，就是不會突破前高或是壓力區，而且量能萎縮。  
多頭看支撐不看壓力，空頭看壓力不看支撐，只要壓力沒站上，所有的漲勢都是反彈。所以不是上漲就是好，要看趨勢、位階、型態。反彈後會如何呢？除非出現翻多的要件，否則通常會續跌，如果強一點的走勢，就是以盤代跌，化解空方壓力。  
昨天夜盤期指持續偏弱，不過在支撐區跌勢稍有收斂，估計今天開也是偏弱，呈現緩跌態勢，雖然下跌時投資人壓力很大，但不怕「大漲大跌，最怕不漲不跌」，希望能快點走出趨勢，不然感覺身邊的投資人都快被盤死，很磨人的盤勢。  
雖然在箱型整理區間，但已經逼近中期支撐，尤其費半指數小幅跌破頭部頸線，台股也要觀察的比較仔細，提防空方條件否增加，最後再複習一下近期的觀察重點。
