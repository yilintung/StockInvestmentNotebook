### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線止穩，中期箱型  
周末影音解盤提到，「加權指數和櫃買指數短線有機會止跌」，昨天加權指數和櫃買指數同步開高，加權指數更是大漲594點，止跌回穩不意外，不過漲勢確實挺驚人的，稍嫌美中不足的，是成交量只有3611億。  
櫃買指數漲幅相對小，只有0.71%，而且同樣量能不足，單日強勢類股有半導體、電子、光電、造紙、電機機械、電腦及周邊設備、數位雲端。台幣由貶轉升、歐美股市轉強、投信連續買盤回歸、4個以上類股回到多方軌道。  
因此雖然台股單日轉強，但上述四個條件尚未明顯會聚，所以只能視為短線轉強，中期趨勢還是維持不變，箱型整理中性看待。不過至少原先持股比重高的投資人，可以暫時鬆一口氣，至於已經降低持股的投資人是否需要積極回補？因為只是短線止跌轉強，但中期趨勢不變，所以採取漸進式回補策略即可。  
尤其個股短中期多空結構，其實還沒有明顯改善，空方個股比例還是偏高，所以不用急著大量回補，真的挑選到符合選股濾網的個股，慢慢布局即可，而且仍要持續追蹤國際指數和台股指數變化。昨天夜盤期指續漲211點，今天加權指數有機會跟進補漲，不過即使如此這個漲幅也還不會改變中期箱型整理結構，以及突破長天期壓力區。 (出處：K線捕手楊忠憲粉絲團)

【[老師，你的反指標朋友們，現在好嗎？](https://www.facebook.com/kevinyang20180801/posts/pfbid099bdf8MB2dFf6Ed1KwXxwb4FZMcYjzWTfErfmHA6U6JpjukuiYxA5Dq7RVSsyGSYl)】  
老同學都知道，我有6個反指標朋友，但已經有好長一段時間，我都沒有提過他們，這是因為反指標們沒有明確的交易策略，只是聽消息做股票，所以資金幾乎都套死在市場，主要來自於兩個波段。  
第一波是2022年的俄烏戰爭起跌(藍色)，這一波反指標朋友們幾乎主要是套在傳產和原物料相關類股，還有今年2024年7月的急跌(綠色)，主要是套在重電、電纜、AI、營建。  
投資最重要的就是資金的流動性，如果你錢卡在市場裡，沒有設好停利停損，不知道如何脫身，換言之就是沒有養成獨立交易的技巧跟習慣，只是聽消息做股票，即使有再多的機會跟潛力股，也都與你無關。  
投資人很容易跟著媒體一頭熱，以為自己在研究產業面、基本面、題材面、公司治理…，殊不知這些全部都叫消息。我當初給他們的觀點不外乎以下幾點：  
1.千萬不要攤平，會越攤越平。  
2.財報永遠都是落後指標，股價先跌一大波，基本面才會變差。  
3.錢多沒有用，方法不對只會賠更多。  
4.沒有本多終勝這件事情，那只是賭運氣。  
5.資產倍增可以有無數次，但只要沒有做好資金控管和調節，破產只要一次。  
所以目前的舊反指標朋友們，都非常安靜，不過倒是有一些新的反指標朋友，慢慢冒了出來，不過還不夠多，等到樣本數夠多，再跟大家分享。
