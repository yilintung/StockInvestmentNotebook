1.先前偷跑提早做空指數，可能要有停損機制，或許換成空方個股做空可能會好做一些。  
2.主動式選股能力不強的投資人，要考慮參與一下權值股或是連動高的ETF(2個月前就有說過)  
3.主動式選股能力不強，又不想買權值股跟ETF，那資金比重應該降低，保守經營，甚至空手休息一下。  
4.主動式選股能力強，操作又靈活的投資人，資金比重可以放大。
