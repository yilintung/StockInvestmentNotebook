【沒有投資和投機之別，只有賺跟賠】  
相信大多數投資人都看到這則新聞：「輝達（NVIDIA）近日提交美國證交會（SEC）文件指出，執行長黃仁勳在9月初一共賣出將近9萬股輝達股票。」相信很多人很驚訝，但其實我之前分享會跟上課都說過，其實群組也常分享，首先第一件事情，雖然說過N次，但還是要念念經：你以為在研究基本面跟產業面，其實只是在聽消息做股票。  
基本面跟產業面很多時候，只是主力作手說服你買賣股票的工具，如果全部都照單全收，那就會成為股市裡的工具人，今天很得意抱住股票的原因，日後就會成為賣不掉的癥結。如果只因為基本面跟產業面很好，就盲目買進股票，很可能就會成為被主力說服，甚至出貨的對象。  
第二，一間企業的經營者絕對是投資者，但是股價上漲賣掉股票賺取價差，這是投資還是投機？開公司不也是為了賺錢嗎？股票低買高賣也是賺錢，所以本質上根本沒有投機跟投資之別，只有賺跟賠。倘若我賣掉股票獲取的價差，是80年的股息，或是賣掉公司資金，可以再重開10家一樣的公司，那請問你要不要賣？我肯定半夜就賣光光了，只是不好意思全部賣而已。  
第三，其實不論任何從管道得知的資訊，投資人本來就應該理性和客觀分析，而不是人云亦云、隨波逐流，全部照單全收，所有市場上流傳的利多和利空，都要學會獨立思考和判斷，再決定交易行為，這些資訊都是選股的參考，但究竟要怎麼買跟賣，那才是賺賠的關鍵，不要一再重演這些悲劇，成為被有心人操弄的工具人。
