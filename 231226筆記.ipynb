【自己的買賣點，只有你自己知道】  
我常常說過每個人條件不同，打個最簡單的比喻，假如某股票目前100，你買在80那可以很悠哉的抱著，但如果你買在95，現在每天可能都要相當警戒，又比如追買在99，那99一跌破，可能就要考慮減碼或停損。  
這還只單就一個買進價格做假設的情境做假設，倘若還加上每個人究竟是試單階段、加碼階段，還是買足階段，3種價格乘上3種情境，就會有9種，但更可怕的是每個人的風險控管、張數上限、資金水位、交易周期、做多做空、股齡、專業程度都不同...這樣推演下來，會有數百種排列組合。  
因此我解盤極少提到如何交易，只能客觀就我看到的多空趨勢、漲跌強弱、轉折訊號、位階、型態...，至於如何交易或是做多做空，就要靠各位的經驗跟智慧了，不是故意要打啞謎，是真的有困難，而且這當中的買賣感覺、交易操略，都必須自己親身參與，才會學到東西。  
也因為上述的條件不同，在每天說明跟教學過程，一直都會出現這樣的情境，其實我寫的內容只有一個版本，可是到了不同人眼中，就會出現各種不同的組合跟變化。莎士比亞的名劇哈姆雷特：一千個人眼中，有一千個哈姆雷特。同一套劇本傳了幾百年，永遠只有一個版本，但是一千個人眼中會有一千種體會，和各自解讀的方式。
