```python
def stock_analysis( analysis, stock_id, buttom_pattern = False) :
    results,images = analysis.analysis(stock_id)
    if buttom_pattern is True and '底部型態' in images :
        print('底型反轉交易策略：')
        display(images['底部型態'])
    print('解盤內容：')
    if '整體評價' in images :
        display(images['整體評價'])
    result_md   = results.to_markdown(tablefmt="grid")
    result_html = markdown.markdown(result_md, extensions=['markdown_grid_tables:GridTableExtension'])
    display(HTML(result_html))
```

* 加權指數  

> 盤前：[十秒鐘看台股](https://www.facebook.com/kevinyang20180801/posts/pfbid05FtXZJ2XYQ9WnHeMScv73Q7BxzYFBk2rjycohhnZZriNhzF9CUzcocSSXh4tqqU2l)  
> 加權指數中期壓力解除回到多方，短線高點未過維持盤整，法人散戶同買，籌碼中性偏多。  
>   

```python
# 支撐線
def line_drawing_callback( range_prices) :

    seq_of_seq_of_points = []
    linewidths           = []
    colors               = []
    
    # 設定上升趨勢線
    support_line_start_date  = '2025-04-01'
    support_line_end_date    = range_prices.iloc[-1].name.strftime("%Y-%m-%d")
    support_line_x           = [mysys.DateToIndex(range_prices,'2025-06-20'),mysys.DateToIndex(range_prices,'2025-08-04')]
    support_line_y           = [range_prices.loc['2025-06-20']['Low'],range_prices.loc['2025-08-04']['Low']]
    slope,intercept          = np.polyfit(support_line_x,support_line_y,1)
    support_line_start_price = slope * mysys.DateToIndex(range_prices,support_line_start_date) + intercept
    support_line_end_price   = slope * mysys.DateToIndex(range_prices,support_line_end_date) + intercept
    seq_of_seq_of_points.append([(support_line_start_date,support_line_start_price),(support_line_end_date,support_line_end_price)])
    linewidths.append(1.0)
    colors.append('xkcd:red')

    # 設定盤整區壓力線
    resistance_line_start_date  = '2025-08-07'
    resistance_line_end_date    = range_prices.iloc[-1].name.strftime("%Y-%m-%d")
    resistance_line_price       = range_prices.loc['2025-08-19']['Open']
    seq_of_seq_of_points.append([(resistance_line_start_date,resistance_line_price),(resistance_line_end_date,resistance_line_price)])
    linewidths.append(1.0)
    colors.append('xkcd:azure')

    # 設定盤整區支撐線
    support_line_start_date  = '2025-08-07'
    support_line_end_date    = range_prices.iloc[-1].name.strftime("%Y-%m-%d")
    support_line_price       = range_prices.loc['2025-08-07']['Open']
    seq_of_seq_of_points.append([(support_line_start_date,support_line_price),(support_line_end_date,support_line_price)])
    linewidths.append(1.0)
    colors.append('xkcd:azure')
    
    colors.append('xkcd:white')
    
    return seq_of_seq_of_points,linewidths,colors
```

```python
# 盤前：支撐線繪製
image = mysys.DrawOnKlineChart('TAIEX','2025-03-31','2025-08-25',line_drawing_callback)
display(image)
```

```python
def line_drawing_callback( range_prices) :
    # 設定頸線
    neckline_start_date = '2025-03-26'
    neckline_end_date   = range_prices.iloc[-1].name.strftime("%Y-%m-%d")
    neckline_price      = range_prices.loc['2025-05-15']['Close']

    # 設定底部到頸線差距
    bottom_date  = '2025-04-10'
    bottom_price = range_prices.loc['2025-04-10']['Close']

    # 估算目標價
    target_price = (neckline_price - bottom_price) + neckline_price

    print('估算目標價為{:.2f}元'.format(target_price))

    # 設定突破日期
    breakout_date = '2025-08-15'

    seq_of_seq_of_points=[
        [(neckline_start_date,neckline_price),(neckline_end_date,neckline_price)],
        [(bottom_date,bottom_price),(bottom_date,neckline_price)],
        [(breakout_date,neckline_price),(breakout_date,target_price)]
    ]

    linewidths=[2.0,15.0,15.0]
    
    colors=['xkcd:orange yellow','xkcd:orange yellow','xkcd:orange yellow']

    return seq_of_seq_of_points,linewidths,colors
```

```python
image = mysys.DrawOnKlineChart('2464','2024-12-25','2025-08-26',line_drawing_callback)
display(image)
```
