### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線反彈，中期偏空  
加權指數自7月31日創高17463後起跌，呈現高點越來越低，低點越來越低的空頭走勢，昨天雖然小漲61點，但成交量急縮至2986億，籌碼面變化不大，外資、融資、主力指標累計均站在賣方，比較值得注意的是投信連續兩周，持續逢低承接。  
加權指數未來的跌幅滿足，可以從兩個面向做觀察，第一個方式是頭部的跌幅滿足，大約是在16000點左右，也剛好是第一季和第二季的箱型上緣。第二個方式則是採用黃金分割率，回檔0.382的位置在15543，取整數可以記為15550，回檔0.5的位置在14986，取整數可以記為14950。  
不過在台股修正過程中，除了觀察上述16000、1555、14950三個支撐區之外，並不代表修正到該位置就會出現轉強，仍需觀察是否出現止跌訊號，再來判斷是否轉強。尤其本波空方走勢，並非台股獨有現象，國際股市也普遍同步弱勢，甚至不少國際指數隱約都有頭部雛形，加上季線扣抵未來1-2周都可能扣到高檔，季線如果下彎，季線的支撐力道不足，一旦頸線跌破，就會開啟新一波跌勢。  
中期空方趨勢基本上已經確認，尚未到支撐區以及出現明確的止跌訊號前，或有機會小漲或反彈，但不會影響空方行進趨勢。持股比重高的投資人，一定還是會想問，是否有立刻由空翻多的機會？如果以目前做觀察，基本上最少要有兩個條件，類股結構是否能出現4個以上的主流類股，以及16650能否站回，可以自行做觀察，不過16650是浮動條件，倘若指數續跌，這個標準會再下修。 (出處：K線捕手楊忠憲粉絲團)
