### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】反彈遇壓，留意轉折  
加權指數從上周強勢反彈，短短3天累計超過500點，從漲幅跟時間來看，算是符合轉強要求。而且上攻到了季線與120日線之間，已經屬於強勢反彈。趨勢的行進過程中，會有許多大大小小的壓力和支撐，越短期的支撐和壓力，容易常常突破跟跌破，所以意義相對小。  
目前走到最樂觀的位置，也是中期空方最後的防線，再跌破就要退守到長期防線，所以走到這個位置，算是中期多空的關鍵，倘若這個壓力有效突破，中期趨勢以及後續觀點，可能要做大幅修正。只是要突破這個壓力，需要多方條件匯聚，例如：國際股市、類股結構、法人籌碼、台幣，都要持續往多方發展，更重要的是往上續攻立刻就會遇到6-7月的前波大量的頭部壓力(藍色)，因此10日均量必須要放大到4千億，否則不容易挑戰成功。  
多頭是緩漲急跌，空頭是急漲緩跌，目前正是急漲緩跌，目前顯然是空頭中的反彈。所以究竟有沒有由空翻多，問投資人自己最清楚，前幾個月買的股票解套了嗎？大部分應該沒有，除非是10月31日開始搶反彈，短多才有可能獲利。周末影音解盤提到，上半周就可能反彈結束，留意反彈壓力區，是否有止漲現象。搶反彈也要留意短線過度追高的風險 (出處：K線捕手楊忠憲粉絲團)
