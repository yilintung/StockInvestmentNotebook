### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】櫃買偏空，小型股弱  
聯準會7月利率決策升息1碼，利率區間提高至5.25%~5.5%，由於符合市場預期，歐美股以小漲小跌表現，包含夜盤台期指小漲0.25%，整體變化不大。不論聯準會是鴿是鷹，股市中這一秒的新聞，下秒就是歷史，所以多空訊息立刻反應在漲跌變化，今天就不用再花時間過度解釋。  
回到台股本身，昨天加權指數雖然只小跌36點，但是投資人卻感覺跌了300點，不少個股都接連湧現賣壓。這來自兩個原因，首先是加權指數的內含，金融指數上漲1.38%，但電子期下跌0.88%(電子零組件類股下跌1.22%)，所以金融指數平衡了電子指數的跌點，讓加權指數看起來很平穩。  
再者，真正重災區是櫃買指數，單日表現開低走低，午盤到終場更是下跌1.48%，儼然有中期頭部雛形(紅色是樂觀版頸線，藍色是悲觀版頸線)，因此不少中小型股，特別是AI概念股，都有不小的賣壓。雖然加權指數看似小跌，但台股結構上是偏空且弱勢的。  
其實這也是典型的拉大出小策略，利用指數強勢表現(週二)或是波動不大(週三)，掩護手上的中小型股出貨，不過目前為止賣壓，集中偏二線的AI概念股，昨天一線AI概念股還沒有明顯出貨現象，一方面因為出一線的AI股容易市場被發現，一方面因為有些仍是處置股，大量出貨會變成跌停(其實有幾檔中小型AI就是如此)。  
散戶小台增加，融資沒有出現大幅減碼，顯見散戶仍看多做多，反觀法人6月中旬至今，卻是積極減碼，籌碼面也是偏空。不過因為並非全面性的個股走弱翻空，只要沒有過度追高AI概念股，應該也不至於太恐慌。  
維持周末以來的策略，指數沒止穩之前，盯緊持股的支撐即可，離場後可以暫時保留現金，採取漸進式的減碼，修正過程的股票不要急著低接，20日線下彎或是做頭的個股更要避開，個別股出現離場訊號，就算做不到減碼或是停損，也不要急著低接或是搶短。 (出處：K線捕手楊忠憲粉絲團)
