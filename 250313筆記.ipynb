```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import mplfinance as mpf
import numpy as np
import myutils
```

```python
def date_to_index(df_in,date_in) :
    number_array = df_in.index == date_in
    idx = 0
    for number_index in number_array :
        if number_index == True :
            break
        idx = idx + 1
    return idx
```

```python
##### 使用2024年8月21日到今日(2025年3月13日)收盤之日線圖資料 #####

# 載入從「好神通PLUS」輸出的Excel檔
df = pd.read_excel('./data/SINO_TSE_250313.xlsx')

# 保存K棒的基本資訊(開、高、低、收、量)
df_k_line = df.drop(columns=['SMA5','SMA10','SMA20','SMA60','SMA120','SMA240','MA5','MA10'])

# 保存價的均線
df_sma = df.drop(columns=['開盤價','最高價','最低價','收盤價','成交量','MA5','MA10']) 

# 將K棒的Columns的名稱由中文改為英文
df_k_line = df_k_line.rename(columns={'時間':'Date','開盤價':'Open','最高價':'High','最低價':'Low','收盤價':'Close','成交量':'Volume'})

# 將價的均線的Columns的名稱由中文改為英文
df_sma = df_sma.rename(columns={'時間':'Date'})

# 將K棒的Date設為Index
df_k_line.set_index(df_k_line['Date'],inplace=True)
df_k_line = df_k_line.drop(columns=['Date'])

# 將價的均線的Date設為Index
df_sma.set_index(df_sma['Date'],inplace=True)
df_sma = df_sma.drop(columns=['Date'])
```

```python
# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定頸線（參照「隱形頸線」）
neckline1_start_date  = '2024-09-16'
neckline1_end_date    = df_k_line.iloc[-1].name.strftime("%Y-%m-%d")
neckline1_price       = df_k_line.loc['2024-11-29']['Close']
neckline2_start_date  = '2024-09-12'
neckline2_end_date    = neckline1_end_date
neckline2_x           = [date_to_index(df_k_line,'2024-09-19'),date_to_index(df_k_line,'2025-03-10')]
neckline2_y           = [df_k_line.loc['2024-09-19']['Close'],df_k_line.loc['2025-03-10']['Close']]
slope,intercept       = np.polyfit(neckline2_x,neckline2_y,1)
neckline2_start_price = slope * date_to_index(df_k_line,neckline2_start_date) + intercept
neckline2_end_price   = slope * date_to_index(df_k_line,neckline2_end_date) + intercept

seq_of_seq_of_points=[
    [(neckline1_start_date,neckline1_price),(neckline1_end_date,neckline1_price)],
    [(neckline2_start_date,neckline2_start_price),(neckline2_end_date,neckline2_end_price)],
]

# 設定移動平均線
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.6,color='xkcd:maroon'),
    "SMA10": mpf.make_addplot(df_sma['SMA10'],width=0.6,color='xkcd:cyan'),
    "SMA20": mpf.make_addplot(df_sma['SMA20'],width=0.6,color='xkcd:violet'),
    "SMA60": mpf.make_addplot(df_sma['SMA60'],width=0.6,color='xkcd:salmon'),
    "SMA120": mpf.make_addplot(df_sma['SMA120'],width=0.6,color='xkcd:blue grey'),
    "SMA240": mpf.make_addplot(df_sma['SMA240'],width=0.6,color='xkcd:sky blue')
            }

# 繪製K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=2.0, colors=('xkcd:red','xkcd:bright blue'), alpha=0.6), volume=True, datetime_format='%Y-%m-%d')
mpf.plot(df_k_line,**kwargs)
```
