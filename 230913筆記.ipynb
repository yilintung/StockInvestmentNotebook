【[成交量太低，會不會賣不掉？](https://www.facebook.com/kevinyang20180801/posts/pfbid0Rujxgw7opDwjNrkPDxSVwDPSF8SNxg4pnfHVLNyXUuLJD7ZKrAEQr6npCMYcH4xvl)】  
常常有同學問我：老師，這檔股票成交量很低，會不會賣不出去？這時候我會反問他，你要買50張還是100張嗎？他說3張。  
成交量低的個股，確實會有流動性風險，但如果你不是主力大戶，一買要買上百張，甚至上千張，我想你不用太擔心這個問題。  
但成交量過低的個股，風險反而不是你賣不掉，而是在外籌碼少，被特定人控盤的機率很高，所以股性活潑，操作難度高，風險係數也高。  
所以我買一檔股票，其實不會刻意在意成交量大小，而且盤整通常無量，上漲自然就會有量了，我比較在意的是過去的股性，會不會太活潑，也就是會頻繁出現長紅、長黑、長上影線、長下影線、跳空缺口，或是異常急漲急跌，那樣我反而會避開。  
你還是很在意成交量太低？那就挑均量有1千張以上的，基本上應該就會比較安全。  
  
[加權指數到今天盤了24天](https://www.facebook.com/kevinyang20180801/posts/pfbid0n7ZnBZRZDpZjDi785QdSMGW7Me97eMkmiDy2k8vtdXS2KDREELkNaGf1mMTaA7Ubl)，你一定覺得很久對不對？但根據過去慣性，這樣的整理區間，盤整時間最長可以到2個月的交易日…XD，希望不要盤那麼久。  
所以道氏真的是神人，近100年前就已經跟你說，多頭做多，空頭做空，盤整不要做，盤整區間少量參與可以，急著重壓多空，會有時間跟趨勢的風險。  
時間風險就是不知道要熬多久，趨勢風險就是盤完之後要上或下，也說不準，所以盤整時，耐心等待真的是最好的策略。  

  
【[聊一個應景話題](https://www.facebook.com/kevinyang20180801/posts/pfbid033P3JVZv51QHn1yQRbGtKsBqJpSPmekHEa9Yci9uvW4fCvGCwHY6Y4WRBhUJa2KbTl)】關於中秋變盤  
最近不時有同學問我，中秋節後究竟會如何呢？為什麼電視常常會說變盤？  
先解決第一個問題，所謂變盤的定義是甚麼呢？就是當中秋節後，加權指數會由多頭變成空頭，或由空頭變成多頭，又或者是突破盤整格局，也就是中秋節後，出現趨勢改變，而中秋節變成明確的多空轉折日。  
但股票是實戰的科學，只要稍作回測，統計近10年，究竟有沒有85%以上的「變盤」現象，就能知道是真是假，答案是會變盤的機率不到60%，投資統計中沒有75%以上的機率，等同沒甚麼意義。  
更何況如果稍微懂統計學的人，就知道統計學的精神是樣本數必須夠多，才能推演未來的機率。比方過去發生過1000次，當中有800次都是會變盤，這樣才具備統計意義，能夠推演第1001次的發生的有效性，但是倘若樣本只有10次、20次，對於未來推斷的有效性，其實是沒有意義。  
即使過去10次都發生變盤，也都還不足以支持第11次的走勢，何況真的變盤次數還不到60%，這等於是無效統計，所以這個變盤說，事實上不具備實質上的意義，包含各種節氣變盤也是如此：春節變盤、中秋變盤、端午變盤...，很容易不攻自破。  
這樣解讀應該會更好一些，中秋變盤是一種現象，中秋節可能剛好變盤，於是稱為中秋變盤，但是沒有證據或是高機率顯示中秋一定會變盤。但為什麼媒體和電視老愛說這個呢？那就要怪散戶，因為散戶聽到「變盤」二字，就會很像觸電，眼睛聚焦收視率就會提升，屢試不爽，媒體當然反覆操作樂此不疲。盯好趨勢、位階、型態就好，人云亦云的事情，不要隨之起舞。 (出處：K線捕手楊忠憲粉絲團)  
備註：嚴格說來稱的上中秋變盤，由多翻空也只有一次，為抑制股市投機風氣，1988年9月26日中秋節，當時的財政部長郭婉容宣布復徵證所稅，造成台股連跌19天，也可以算是中秋變盤說的由來。關心中秋節沒有連假，可能比較實際，今年2023年，中秋連假為9/29（五）~10/1(日）連休三天。
