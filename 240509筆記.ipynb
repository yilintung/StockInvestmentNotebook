【關於存股你該注意的一些事情】  
其實早期已經寫過蠻多對存股(或者要稱為價值型投資或長線投資)的看法，當中討論很多存股的盲點和風險，這邊就不一一贅述，但要先聲明我不是反對存股、長期投資或價值型投資。  
只是很多人好像都以為，只要買完之後，擺著、放著，甚麼事都不要做，有一天就會變成有錢人，但這真的是很危險的，總結幾個重要觀點，給想存股的同學，應該有的一些觀念：  
1.是否有願意用年以上為單位，逢低有持續以及有計畫性，分批買進的資金和勇氣？  
2.持有一段時間後，股價上漲了，該一直加碼？還是逢高減碼和獲利調節？還是像坐電梯抱上又抱下？而調節的具體依據和工具又是甚麼？  
3.萬一長期逢低持續買，買了多年以後，平均成本是100元，但後來股價跌到50元甚至更低，才發現股價可能再也不回來了，那時候該怎麼辦？  
4.持續買了很多年之後，已經投入巨額資金，後來公司基本面才轉差，而且配股配息也變差，甚至未來可能不配股配息，那時又該怎麼辦？  
5.投資系統中有沒有停損機制？願意執行停損機制嗎？  
如果你沒有想清楚以上的問題和風險，以及具體的因應措施和解決方案，那就不是存股、不是長期投資、不是價值型投資，而是孤注一擲的豪賭，恐怕有一天會出大事的。只有上帝才會知道3年、5年，甚至10年之後會發生甚麼事，所以長期投資，本質上是上帝視角投資法。
