### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】急跌出量，醞釀反彈  
昨天盤前日本出現重挫，並且觸發融斷機制，盤前就提醒「融斷機制可以避免該國系統性風險，但賣壓會外溢到其他市場，要更加小心」，加權指數開低走低，成交量6679億，終場下跌1807點，上週史上最大跌點，沒幾天就再度被刷新。  
常常會聽到這樣的話，我活麼麼久了都沒看過，所以代表不會發生。韓國的網路流行「活久見」，指的不是不會發生，是活得還不夠久。比如曾幾何時看過，貢獻大盤指數漲點的欄位會有從缺，因為幾乎所有股票都重挫，昨天就看到了這樣的天文奇觀。  
跌不停比跌停更可怕，因為跌停只是一天的跌勢，但跌不停是趨勢的形成。所以跌不停是「恐慌」，而跌停是「絕望」，雖然兩個都不是好事情，如果原先已經先下跌過一段，之後融斷機制出現，代表市場趨近於絕望，其實也是短線趕底的時候。  
目前加權指數月線和季線都跌破且下彎，半年線下彎但年線還上揚，雖然中期空頭確立，但長期趨勢則還有轉圜餘地，但有一件事情是確定的，整理期肯定會拉長。短線則因為急跌，已經到了前低、上升趨勢線、黃金分割率0.618的複合式支撐，隨時有機會出現強勢反彈，第一關反壓就在20800。  
現在的急跌走勢跟新冠疫情那一波很像， 如果用地震比喻，融斷機制是主震，後面還會有幾天出現餘震，餘震可大可小，沒有出現完全的止跌和轉折現象之前，千萬不要猜低點，靜待止跌訊號出現即可。  
投資人最關心的問題，是何時可以開始做多？未來即使有上漲，只要短壓20800沒突破，都視為短線搶反彈操作，但前提是高檔有減碼保留現金，現在低檔才有資格談搶反彈，否則都只是等解套，或是變成攤平交易。 (出處：K線捕手楊忠憲粉絲團)

雖然有很高的機率，預期今天會強彈，但真的要在昨天進場買股票，還是要有相當的勇氣。還有，原本空手的人昨天進場才叫做搶反彈，原本滿手股票的人昨天進場叫做攤平。同樣的交易行為，在不同的個體條件，會產生不同的意義。  
搶反彈是抓低點轉折，買到成本最低的時候，未來漲多漲少都是賺，隨時可以離場。攤平交易是高檔沒有減碼，持有滿手股票，低檔持續買進降低平均持有成本，倘若後續彈幅不大，未到解套成本，那反而會越套越深。
