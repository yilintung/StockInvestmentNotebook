這是很好的問題，但說真的我不知其他專職投資人的配置，因為這比較私人，所以我只能以我的經驗狀況做說明。首先先講獲利目標，以一般人年薪50萬為基礎，如果你本金很高純粹靠配息，那要把目標設在120%，也就是要抓60萬，因為配息會有誤差，每年也會不同，所以要有個安全值(當然你也可以抓得越高越好)。所以說真的要完全靠配息退休，本金必須1500萬以上，但這對一般人有難度。  
另一種則是靠賺取價差，這不像股息那麼穩定，會有高、有低、淡季、旺季，把自己想成開一間公司，公司營收獲利本來就有高低，所以你的目標值要抓更高，可能要150%甚至200%，也就是一年要賺75-100萬為目標。比如我2023年和2022年的總獲利，還不到2021年的1/3，所以有大賺的年度，一定記得要守的住，以備未來不可知的風險。  
但因為是賺價差，所以你本金可以低很多，比如如果可以年化報酬率20%，你本金500萬以內就可以達到了，像我上課教的ETF波段操作法，很容易做到，反過來，當真的空頭時，積極操作人也可以改做反向ETF，這樣多空都有獲利。  
再來就是結算獲利的時間週期，盡量以季為單位，甚至是半年，不然你會很緊繃，更何況本來操作不論買點、賣點、等待期，都需要時間，一旦有時間壓力，你會很難做。相同的，這時候你的安全準備金就很重要，我建議至少要6個月以上，但一樣多多益善，安全準備金越多，更能以時間換取空間。  
最後是靠退休的準備期，我稱為實驗退休，我當初給自己是3年，盡可能在職的時候就同步規劃和練習，盡量不要突然裸辭，然後開始悶著頭做交易，這樣你壓力會很大，尤其是上有老下有小，人一旦壓力大，操作就會不客觀不理性，甚至生理上會生病。  
至於我的企管顧問或是餐飲業投資，其實占比已經相對低，不到我總收入的1成，幾乎都是剩下人情，推不掉的案子才接，不然接太多也會影響交易。所以本質上我還是以金融投資為本業。
