### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線過熱，狹幅整理  
昨天加權指數開高上漲，漲點最多來到368點，資金分布也相當平均，80%以上的個股都上漲，但是9點25分曾提醒：盤中的20日正乖離超過5%，短線過熱，有可能因為正乖離過大，稍微休息一下。9點45分隨即開始走弱，最後漲點收斂至157點。櫃買指數沒有這個問題，走勢相對穩定，維持0.8%左右的漲幅。  
我個人累計觀察股依照過去統計，10檔以內漲停代表正常，盤勢很穩健多方，11-15檔表示已經過熱，16-20檔以上漲停，風險係數就會攀升，容易回檔整理。昨天把漲停和接近漲停納入約有13檔，屬於過熱但還沒到危險，也呼應正乖離過大的表現。(並不是台股全部1800多檔中的10檔、15檔、20檔漲停，只有我的觀察股才算數，過去指標準確率超過90%。)  
整理網友的隨機提問的過程，對於個股表現會更加明朗，雖然加權指數續創新高，但真正多方續漲，大概只有佔30%，40%維持盤整，高達40%竟然有頭部雛形，甚至已經做頭。所以很明顯的，目前加權指數不能完全表徵個股的結構，反而櫃買指數比較貼近事實真相。  
不過比較正面的是，如果選股能力有一定基礎，這30%的多方個股很容易掌握，族群也相對集中，基本上都是這兩周有提到的族群。最後，雖然加權指數短線過熱，漲勢暫時休息機率相當高，不管依照支撐跌破法還是目標滿足法，目前波段多頭都還沒結束，所以周末再做完整教學說明。 (出處：K線捕手楊忠憲粉絲團)
