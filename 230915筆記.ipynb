### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】亞股轉強，股匯雙漲  
加權指數上漲226點，成交量溫和放大到2878億，外資期現貨同步翻多，但尚未突破季線反壓16840，很多投資人看到指數大漲，都開始興奮起來，不過其實加權指數結構上仍是小箱型區間。  
重心還是在櫃買指數，昨天漲幅1.68%，而且挑戰季線反壓，相對強勢，因此很多中小型股，都紛紛表態，所以盤勢走到這邊，要多認真觀察，盤整格局感覺好像要走出趨勢，檢視的面向就要多一些。  
國際面、資金面、籌碼面、結構面、技術面、成交量…，都要一併分析，不過早上盤前把這些內容都寫完，工程真的太大，字數也會太多，而且今天應該是小紅小黑居多，不差這一天，所以完整的論述，周末影音解盤再說說明，這邊先簡單說結論。  
從單一日觀察，當然是多方沒有問題，不過加權指數仍維持在箱型區間，櫃買指數再續攻就會進入大壓力區，所以短多不難，但續航力道可能還有待商榷，中期還是先視為箱型整理。  
而且指數大漲還是沒有改變選股難度，雖然雨露均霑多數個股也都跟漲，但從昨天漲幅較大的強勢股觀察，10檔之中可能只有1檔聽過，而且由於原先指標股和主流股尚未真正過高。  
所以指數雖然大漲，一般投資人相對無感，短多類股有油電燃氣、綠能環保、運動休閒、居家生活、塑膠、橡膠、航運、電子通路、金融保險。 (出處：K線捕手楊忠憲粉絲團)
