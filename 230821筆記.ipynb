【訊息量要適中，過猶不及】  
曾經有一個同學，給我看他的加入的社團臉書和群組，真的讓我嘆為觀止，密密麻麻的，少說70個以上，我很訝異的問他，這麼多訊息你怎麼看？他坦白的說，還真的有點招架不住，後來真的都亂看，重點是訊息量太龐雜，反而不知道應該先看甚麼？  
然後我也給他看我的手機，別人建立的line群組，我真的一個都沒有，我開玩笑說我有網路社恐，然後他就反問我，這樣甚麼訊息都沒有，怎麼做股票？我反問他說，那你有這麼多訊息，有做得比較順嗎？他說更累了，每天光是要消化完這多群組的訊息，就忙不過來了。  
我想這是很多投資人的縮影，因為手機的便利性，主動和被動的資訊，鋪天蓋地的襲來，反而會讓人覺得不知道所措，獨立思考還有養成自己的一套工具。國際上發生甚麼大事當然是要關心，但適度就好不宜過量，而且關鍵要懂得建立判斷跟分類機制，還有自主搜尋資訊的sop，太多訊息而超過負荷時，等於沒有訊息了。
