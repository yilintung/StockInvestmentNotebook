### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】測試完畢，恢復盤整  
昨天盤前提到由於上周五歐美股市走弱，以及夜盤期指下跌179點，「沒意外的話會面臨補跌，當然個股也要面臨壓力測試」，開盤下跌約155點，之後開低走低，幾乎收在最低，終場下跌222點，成交量2855億。  
跌點主要來自前五大權值股中的台積電和廣達，貢獻70%以上的跌點，波動的節奏幾乎和這兩檔總和相仿，個別股反而是小紅小黑居多，從櫃買指數的表現也可以看的出來，只小跌0.46%連5日線都沒跌破，不過再來也面臨4月18日前波高點壓力，以及6月5日至8月1日的頭部壓力，要出現連續性大漲有一定難度。  
近期三大法人也是忽買忽賣，陷入觀望沒有連續性的買盤，融資則是持續增加，主力指標轉弱，整體籌碼偏空。不過即便昨天加權指數出現大跌，其實仍然沒有改變中期趨勢，一樣以16850為中軸，上下400點箱型區間的結構，只是維持中軸之下，中性偏空表現。  
加權指數直到9月18日共盤整28天，但根據過去慣性，這樣的整理區間，盤整時間最長可以延長到40天的交易日，當然你我都希望不要盤那麼久，而櫃買指數短線也應該會維持在月線和季線之間進行整理。  
因此「多頭做多，空頭做空，盤整不要做」，盤整區間少量參與打發時間，無傷大雅是沒問題的，如果急著重壓多空，會有時間跟趨勢的風險。時間風險就是不知道要熬多久，趨勢風險就是盤整完之後要上或下，暫時也說不準，所以盤整時，耐心等待真的是最好的策略。
