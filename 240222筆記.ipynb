### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】漲多小回，多頭不變  
昨天加權和櫃買指數都有漲多整理的現象，不過都還守穩短中長期支撐，維持多方架構，重要的轉折現象也還沒有出現，所以看法和昨天差不多，但有一個新增的觀察，就是道瓊指數其實有點上升楔形的雛形，其實上周就有看到，不過因為沒有錄影音解盤，就沒特別說，不過只是雛形，正式跌破支撐楔形才成立。  
另外，相信大家都看到這幾天，都看到陸港股強勁的表現，但事實上在農曆封關前的影音解盤，就已經提到跌幅滿足也到重要支撐區，所以出現止跌上漲是正常現象，也是可以預期的，只是短線急漲到這邊，也到大壓力區，再來漲勢可能會趨緩，甚至進入整理。  
陸港股的止跌上漲，對全球股市有好有壞，假如把全球當一個國家，而主要指數就像是個股，當最弱勢的個股止穩轉強，不會再拖累整體表現，健康輪漲，對整體多方是好事。可是這也正是缺點所在，一個中長期多頭走勢，末升段往往是最強勢的個股漲不動(道瓊、日本)，換二線的個股輪漲，等到二線的個股再漲不動，連最弱勢的個股都漲了，那代表波段行情即將進入尾聲。  
當然這也呼應到這兩周對於時間跟漲幅空間的推算，但以時間推算至少到下周結束前，應該還不用太緊張，等到進入3月後，同步留意轉折訊號的發生，屆時再積極減碼即可，道瓊指數看20日線、加權指數看10日線、櫃買指數看旗形滿足+5日線。當然如果轉折現象提早出現，減碼速度就要快一點。  (出處：K線捕手楊忠憲粉絲團)
