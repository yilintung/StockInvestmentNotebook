另一個重要新聞則是蘋果新機，市場則是反應冷淡，普遍認為沒有太多新亮點，面對手機市場漸趨飽和，以及大陸政策以及華為新機影響，今年全球整體出貨估下滑6%創10年新低，預期銷量恐怕不如預期，發表會後蘋果股價下跌1.71%。  
歐美股市漲跌互見，但都是小紅小黑，道瓊、德國、法國小跌，標普500、費半、那斯達克先漲後跌，但收盤維持小漲，台指期夜盤小漲64點，今天台股有機會開出小紅，不過仍未脫離盤整區間。  
加權指數直到昨天共盤整24天，但根據過去慣性，這樣的整理區間，盤整時間最長可以到2個月的交易日，當然大家都希望不要盤那麼久。所以「多頭做多，空頭做空，盤整不要做」，盤整區間少量參與打發時間，無傷大雅是沒問題的。  
但是如果急著重壓多空，會有時間跟趨勢的風險。時間風險就是不知道要熬多久，趨勢風險就是盤完之後要上或下，也說不準，所以盤整時，耐心等待真的是最好的策略。
