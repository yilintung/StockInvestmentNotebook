### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線偏多，留意轉折  
從11月2日開始，連續上漲12個交易日，漲幅逾千點，最高來到17253，距離中期7月31日前高17463僅一步之遙，相距只有210點。過程中11月15日的成交量放大至3855億，雖然與4千億的標準還有些距離，但已經量能算是已經放大，只是也因為連續性急漲，加權指數上週持續上漲，但正乖離率已經到達4%，下半周由於正乖離過大，漲勢趨緩。  
在高點未過、低點未破之前，仍視為中期高檔箱型，以16800為中軸，上下500點震盪，但因為在中軸線16800之上，屬於偏多方的強勢整理。櫃買指數也是相仿的走勢，11月17日最高來到222.75，相當逼近7月6日中期波段前高227.48。7月31日的高點17463，不只是波段前高，同時也是從年初1月5日高點18619以來的長天下降壓力線，以及價量累計圖的大量壓力區，因此上攻會面臨到複合式的套牢壓力，所以戰線要拉長觀察。  
不論是國際環境、技術面、資金面、類股結構，確實都有往多方前進的現象，也符合指數強勢短多表現。只是即將面臨中長期的複合壓力區，各國主要指數都有近關情怯的現象，要正式突破壓力讓中期趨勢翻多，諸多條件仍需再加把勁，否則容易過高壓回，或是出現高檔價量背離的現象，而主要條件是量能必須穩定維持四千億之上，外資投信同步買超，並且出現市場共識的主流群，附加條件則是國際指數同步創高，這樣新一波的多頭才會走的長、走的久。  
操作上延續這幾周的策略，中期的多空尚未分出勝負之前，投入的資金控制在3-5成，檔數3-4檔，等中期趨勢正是出現再做加碼，反之，倘若未來在關鍵壓力前，出現短線空方轉折，那就要漸進式減碼，持有的個股不需要一次性全部減碼，可以等到個別股票依序出現賣訊，再分批離場。 (出處：K線捕手楊忠憲粉絲團)
