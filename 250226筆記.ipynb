許多投資人經常詢問：「📉 老師，我可以搶反彈嗎？」這個問題背後，隱藏著許多投資觀念的誤解。特別是「搶反彈」與「攤平交易」，兩者看似相似，但實際上有本質上的區別。  
🔥 什麼是搶反彈？  
🟢 搶反彈 指的是在市場大幅下跌後，價格可能短期修正反彈時，投資人選擇在低檔小額進場，利用短線上漲的機會獲利離場。這種策略適合 在高檔時已經減碼、保有現金，或者完全空手的投資人，因為他們可以在低點試探性買入，並在反彈時適時獲利了結。  
💡 搶反彈的核心 在於 快進快出，謹慎控制風險，並非盲目地在低點買進，而是基於市場趨勢、技術分析或消息面來判斷短線行情。  
⚠️ 攤平交易：隱藏的風險  
❌ 攤平交易 通常發生在投資人 已經被套牢 的情況下。當股票價格持續下跌，部分投資人會選擇加碼買進，以拉低整體持股成本，期待未來價格回升時能更快回本。然而，這種策略的風險遠遠大於搶反彈，因為：  
🚨 主要風險：  
❗ 1. 資金可能被套得更深📉 當市場持續走空，股價沒有止跌跡象時，攤平等於加大部位暴露在下跌風險中，可能讓虧損擴大。  
❗ 2. 心理壓力增加，錯誤決策機率更高😰 投資人若持續攤平，會將大量資金投入原本已經虧損的部位，心理壓力隨之加重，導致更難做出理性判斷，甚至進一步影響資金管理。  
❗ 3. 錯誤的期待，忽略趨勢🤔 很多投資人認為「股價終究會回來」，但市場趨勢並非以個人意志為轉移。如果標的本身基本面轉差，甚至趨勢反轉，那麼攤平只是在錯誤的道路上越走越遠。  
🎯 該如何正確應對市場波動？  
當市場出現跌勢，與其考慮「攤平」或「搶反彈」，不如先問問自己：  
✔ 是否有明確的進出場計劃？  
✔ 是否有足夠的資金管理與風險控制機制？  
✔ 是否清楚當前市場趨勢？  
💰 如果你在高檔已經減碼，手上有現金，想要利用短線機會操作， 那麼 搶反彈是可行的策略，但必須設定好停損點，並且小額試探，不要貿然重倉。  
❌ 但如果你是一路被套，想靠加碼來攤平成本， 那就要特別警惕，因為這可能會讓你陷入更大的風險。當市場趨勢不利時，適時停損、重新布局，才是更健康的投資策略。

```python
def date_to_index(df_in,date_in) :
    number_array = df_in.index == date_in
    idx = 0
    for number_index in number_array :
        if number_index == True :
            break
        idx = idx + 1
    return idx
```

```python
from collections import defaultdict

# 原型版：初步開發
def chart_patterns_recognition_4point_prototype(turning_points, max_bars = 100):
    # 儲存型態實例
    patterns = defaultdict(list)
    
    #循環迭代轉折點數據
    # TODO : 問題待檢討 ： len(turning_points) → (len(turning_points)+1)
    for i in range(4, (len(turning_points)+1)):
        
        # 在變數window中一次儲存4個局部最小值和局部最大值點
        window = turning_points.iloc[i-4:i]
        
        # 型態必須在max_bars內發揮出來（預設為 100）
        if window.index[-1] - window.index[0] > max_bars:
            continue
        
        # 儲存4個獨特點以檢查條件
        e1      = window.iloc[0]['Price']
        e2      = window.iloc[1]['Price']
        e3      = window.iloc[2]['Price']
        e4      = window.iloc[3]['Price']
        rtop_g1 = np.mean([e1,e3])
        rtop_g2 = np.mean([e2,e4])
        
        if e2 > e1 and e4 > e3 and e3 > e1 and e4 > e2 and (e2 - e1) > (e4 - e3) and abs(((e4 - e3) - (e2 - e1)) / (e2 - e1)) > 0.25 :
            # 上升楔型
            patterns['RISE_WEDGE'].append(window)
        elif e1 > e2 and e3 > e4 and e1 > e3 and e2 > e4 and (e1 - e2) > (e3 - e4) and abs(((e3 - e4) - (e1 - e2)) / (e1 - e2)) > 0.25 :
            # 下降楔型
            patterns['FALL_WEDGE'].append(window)
        elif (e1 > e2) and (abs(e1-rtop_g1)/rtop_g1 < 0.0075) and (abs(e3-rtop_g1)/rtop_g1 < 0.0075) and \
            (abs(e2-rtop_g2)/rtop_g2 < 0.0075) and (abs(e4-rtop_g2)/rtop_g2 < 0.0075) and (min(e1, e3) > max(e2, e4)):
            # 矩形
            patterns['RECT'].append(window)
        elif e1 < e3 and e3 < e2 and e3 < e4 and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.03) and \
            (abs(e2-e4) <= np.mean([e2,e4])*0.05):
            # 雙重頂
            patterns['DOUBLE_TOP'].append(window)
        elif e1 > e3 and e3 > e2 and e3 > e4 and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.03) and \
            (abs(e2-e4) <= np.mean([e2,e4])*0.05):
            # 雙重底
            patterns['DOUBLE_BOTTOM'].append(window)    
        
    return patterns

# 原型版：初步開發
def chart_patterns_recognition_5point_prototype(turning_points, max_bars = 100):
    # 儲存型態實例
    patterns = defaultdict(list)
    
    #循環迭代轉折點數據
    # TODO : 問題待檢討 ： len(turning_points) → (len(turning_points)+1)    
    for i in range(5, len(turning_points)+1):
        
        # 在變數window中一次儲存5個局部最小值和局部最大值點
        window = turning_points.iloc[i-5:i]
        
        # 型態必須在max_bars內發揮出來（預設為 100）
        if window.index[-1] - window.index[0] > max_bars:
            continue
        
        # 儲存5個獨特點以檢查條件
        e1      = window.iloc[0]['Price']
        e2      = window.iloc[1]['Price']
        e3      = window.iloc[2]['Price']
        e4      = window.iloc[3]['Price']
        e5      = window.iloc[4]['Price']
        
        if (e1 > e2) and (e1 > e3) and (e3 > e5) and (e2 < e4):
            # 三角形
            patterns['TRIG'].append(window)
        elif (e1 > e2) and (e2 < e3) and (e4 < e5) and (e2 < e4) and (e3 < e5) and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.10) and \
            (abs(e2-e4) >= np.mean([e2,e4])*0.015) and \
            (abs(e3-e5) >= np.mean([e3,e5])*0.015):
            # 上升旗形(看跌)
            patterns['BEAR_FLAG'].append(window)
        elif (e1 < e2) and (e2 > e3) and (e4 > e5) and (e2 > e4) and (e3 > e5) and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.10) and \
            (abs(e2-e4) >= np.mean([e2,e4])*0.015) and \
            (abs(e3-e5) >= np.mean([e3,e5])*0.015):
            # 下降旗形(看漲)
            patterns['BULL_FLAG'].append(window)
        elif (e1 > e2) and (e2 < e3) and (e4 < e5) and (e2 <= e4) and (e3 > e5) and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.10) and \
            (abs(e3-e5) >= np.mean([e3,e5])*0.015):
            # 看跌三角旗形
            patterns['BEAR_PENNANT'].append(window)
        elif (e1 < e2) and (e2 > e3) and (e4 > e5) and (e2 > e4) and (e3 <= e5) and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.10) and \
            (abs(e2-e4) >= np.mean([e2,e4])*0.015) :
            # 看漲三角旗形
            patterns['BULL_PENNANT'].append(window)
            
    return patterns

# 原型版：初步開發
def chart_patterns_recognition_6point_prototype(turning_points, max_bars = 100):
    # 儲存型態實例
    patterns = defaultdict(list)
    
    #循環迭代轉折點數據
    # TODO : 問題待檢討 ： len(turning_points) → (len(turning_points)+1)    
    for i in range(6, len(turning_points)+1):
        
        # 在變數window中一次儲存6個局部最小值和局部最大值點
        window = turning_points.iloc[i-6:i]
        
        # 型態必須在max_bars內發揮出來（預設為 100）
        if window.index[-1] - window.index[0] > max_bars:
            continue
        
        # 儲存6個獨特點以檢查條件
        e1      = window.iloc[0]['Price']
        e2      = window.iloc[1]['Price']
        e3      = window.iloc[2]['Price']
        e4      = window.iloc[3]['Price']
        e5      = window.iloc[4]['Price']
        e6      = window.iloc[5]['Price']

        if (e3 < e1) and (e5 < e1) and (e2 < e3) and (e4 < e2) and (e4 < e6) and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.03) and \
            (abs(e3-e5) <= np.mean([e3,e5])*0.05) and \
            (abs(e2-e4) >= np.mean([e2,e6])*0.05) and \
            (abs(e4-e6) >= np.mean([e2,e6])*0.05):
            # TODO : 判斷條件待確認，參照：https://ithelp.ithome.com.tw/articles/10317498
            # TODO : 頭肩底與三重底的區分？
            # 頭肩底
            patterns['IHS'].append(window)
        elif (e3 > e1) and (e5 > e1) and (e2 > e3) and (e4 > e2) and (e4 > e6) and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.03) and \
            (abs(e3-e5) <= np.mean([e3,e5])*0.05) and \
            (abs(e2-e4) >= np.mean([e2,e6])*0.05) and \
            (abs(e4-e6) >= np.mean([e2,e6])*0.05):
            # 頭肩頂
            # TODO : 頭肩頂與三重頂的區分？
            patterns['HS'].append(window)
        elif (e3 < e1) and (e5 < e1) and (e2 < e3) and (e4 < e5) and (e6 < e5) and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.03) and \
            (abs(e3-e5) <= np.mean([e3,e5])*0.05) and \
            (abs(e2-e4) <= np.mean([e2,e4,e6])*0.05) and \
            (abs(e4-e6) <= np.mean([e2,e4,e6])*0.10) :
            # 三重底
            # TODO : e4-e6特例？
            # TODO : 頭肩底與三重底的區分？
            patterns['TRIPLE_BOTTOM'].append(window)
        elif (e3 > e1) and (e5 > e1) and (e2 > e3) and (e4 > e5) and (e6 > e5) and \
            (abs(e1-e3) >= np.mean([e1,e3])*0.03) and \
            (abs(e3-e5) <= np.mean([e3,e5])*0.05) and \
            (abs(e2-e4) <= np.mean([e2,e4,e6])*0.05) and \
            (abs(e4-e6) <= np.mean([e2,e4,e6])*0.05) :
            # 三重頂
            # TODO : 頭肩頂與三重頂的區分？
            patterns['TRIPLE_TOP'].append(window)
    
    return patterns
```

```python
patterns = chart_patterns_recognition_5point_prototype(turning_points,max_bars=240)
print(patterns)

pattern = patterns['BULL_FLAG'][-1]

line1_start_date = pattern.iloc[1]['Date'].strftime("%Y-%m-%d")
line1_end_date   = pattern.iloc[-1]['Date'].strftime("%Y-%m-%d")
line1_x = [date_to_index(df_k_line,pattern.iloc[2]['Date']),date_to_index(df_k_line,pattern.iloc[4]['Date'])]
line1_y = [pattern.iloc[2]['Price'],pattern.iloc[4]['Price']]
line1_slope,line1_intercept = np.polyfit(line1_x,line1_y,1)
line1_start_price = line1_slope * date_to_index(df_k_line,line1_start_date) + line1_intercept
line1_end_price   = line1_slope * date_to_index(df_k_line,line1_end_date) + line1_intercept

line2_start_date = line1_start_date
line2_end_date   = line1_end_date
line2_x = [date_to_index(df_k_line,pattern.iloc[1]['Date']),date_to_index(df_k_line,pattern.iloc[3]['Date'])]
line2_y = [pattern.iloc[1]['Price'],pattern.iloc[3]['Price']]
line2_slope,line2_intercept = np.polyfit(line2_x,line2_y,1)
line2_start_price = line2_slope * date_to_index(df_k_line,line2_start_date) + line2_intercept
line2_end_price   = line2_slope * date_to_index(df_k_line,line2_end_date) + line2_intercept

# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定上升楔形
seq_of_seq_of_points=[
    [(line1_start_date,line1_start_price),(line1_end_date,line1_end_price)],
    [(line2_start_date,line2_start_price),(line2_end_date,line2_end_price)],
                     ]

# 設定均線
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.5,color='xkcd:maroon'),
    "SMA20": mpf.make_addplot(df_sma['SMA20'],width=1.0,color='xkcd:violet'),
    "SMA60": mpf.make_addplot(df_sma['SMA60'],width=0.5,color='xkcd:dark orange')
            }

# 繪製K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), volume=True,addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=(1.0,1.0), colors='xkcd:orange yellow', alpha=0.6), warn_too_much_data=1100)
mpf.plot(df_k_line,**kwargs)
```
