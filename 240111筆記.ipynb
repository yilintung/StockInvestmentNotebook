在分享會做過近百次的實驗，我用台積電做示範，我跟現場的同學提問，並且提供獎金100萬，某天台積電收盤價570元，倘若有一個人可以說出台積電當天為何收在70，以及股價漲跌的原因，可以立刻領走100萬獎金。  
但這個答案必須現場80%以上的同學都認同，這才具備意義，將近10年過去了，從來沒有一位同學領走。  
因為當有一個同學提出這是因為本益比，但立刻就會有同學持反對意見，有人說籌碼才是關鍵，但提出籌碼之後，也不見得每個人都接受，因此顯而易見，股價的成因根本不是單一變因，大多時候是複合變因，是由無數的利多利空交雜而產生。  
當然不否認，有時候單一變因非常強大，足以蓋過所有的因素，比如2020年新冠疫情指數12197大跌至8523，但這種情況是少數。只是大多時候，都是包含國際局勢、總體經濟、政府政策、台股多空、籌碼變化、基本面、媒體報導，還有市場的樂觀與恐慌，這只是隨手寫出，倘若有空還可以多寫個10幾個原因。  
但投資人卻異常的執著，股價漲跌只能有單一因素，營收好股價就要漲，營收不好股價就要跌；籌碼好股價就要漲，籌碼不好股價就要跌；媒體發布利多股價就要漲，媒體發布利多股價就要跌，這是非常危險的邏輯的推論，然後用這種非理性的論述，進行投資和交易行為，在股市中當然常常吃虧。  
結論是，專注於價差的投資人，不需要深究背後的原因，因為根本不會有答案，即使真的有答案，對你進行交易行為也毫無幫助，只要知道台積電380元買進480賣出，或是530買進590可以賣出，清楚辨識趨勢、位階、型態，精準掌握支撐、壓力、轉折，清楚交易和進出的位置，這才是最重要的。

【[現在空手的人可以做甚麼？](https://www.facebook.com/kevinyang20180801/posts/pfbid02C1BWZxpA6eyNsWMnaCFFRim3eLG6hC4aY5Y8jkDrSa9bzsNhMLve2SfSz8DRWswkl)】  
因為現在指數偏弱，但倘若很幸運地，已經做了預防性減碼，可以有以下幾種選擇，第一種是小額的勇敢的追強勢股，雖然指數轉弱，但還是會有極少數逆勢上漲的個股，但必須確定短中長期趨勢都是多頭，而非只是單一日的大漲的強勢股，而操作這種短中長期多頭的強勢股隨時有回檔的風險，所以要能夠設定移動式移動支撐，而且紀律執行的人，才能操作這樣的股票。  
第二種是小額預先布局相對穩定的股票，甚麼叫做相對穩定的個股？除了在關鍵支撐位置有守，例如：20日線或是盤整支撐區，當單日指數大跌，但個股只有小跌，當單日指數小漲時，個股卻能大漲。這就屬於穩中透強的股票，等到指數止穩回升，這一類個股都容易有表現。  
當然最後一種，就是繼續放假和追劇，追劇(與財經題材相關的好劇：財閥家的小兒子、半澤直樹、華爾街之狼、藥命效應、大賣空…)，耐心等待趨勢轉強翻多，到時候再重新檢視市場上的個股，擇優進場。不過選擇這種策略雖然是最單純也最安全，但是當趨勢反轉時要能精準判斷積極進場，否則舉棋不定延誤了時間，等到進場時，指數和股票已經又漲一大波了。
