### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】反彈遇壓，慎防拉回  
上周五當市場極度悲觀，很多人甚至認為盤勢會直接持續一瀉千里，市場氛圍非常恐慌的時候，我反而提醒大家不要太悲觀，盤勢應該會會止跌反彈。接著如規劃的連續4天反彈，尤其昨天大漲532點。  
只是在投資人又開始看多的時候，我反而提醒：「今天我想市場又開始樂觀了，但如果有仔細觀察我的解盤內容，跟我們週末預估的其實是一樣的，接下來反而要進入壓力，要多加小心。」  
市場開始出現偏多的觀點，甚至覺得會再度創新高，這時候反而要有些危機意識，並不是漲就是好事情，多方瑕疵其實沒有太大改善，首先是這4天上漲，成交量仍低於10日均量，從價量關係判斷，明顯是反彈結構。  
第二，20日線仍然下彎，未來兩周的時間，扣抵值會扣高加低，20日線會持續下彎。第三則是反彈已經到了紅色壓力區，雖然是強勢反彈，但這邊壓力也相當大。第四，雖然昨天單一日法人同買，但是買超幅度小，反觀融資增幅較大，而且近2周累計，法人站在賣方，融資還是在高水位。  
第五，外資未平倉空單口數，雖然有小幅下降，但高達13333口，所以期貨的警報尚未解除。第六，類股結構多為反彈或盤整，真正創高或是續強不到2成。延續周末3件個觀點，止跌不難，反彈也不意外，也彈到原本規劃的位置，只是今天可能就會遇壓拉回，中期趨勢不明，需要再多觀察。 (出處：K線捕手楊忠憲粉絲團)
