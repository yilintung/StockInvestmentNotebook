【注意ETF的連動商品】  
富邦越南ETF追蹤之標的指數為「富時越南30指數」，該指數是考慮外資可投資性後，由胡志明證券交易所(以下簡稱胡志明證交所)整體掛牌企業中，選取市值排名前30名的股票，綜合反映越南股市的股票績效表現。  
所以直接就是連動越南股市，因此即使台股出現大跌，只要越南沒有跌，那該指數還是會持穩甚至逆勢上漲。但這一類ETF有兩個注意事項，其一，有時候是當時全球國際股市同步下跌，類似系統性風險，雖然不受台灣影響，但他還是會有賣壓。  
其次，有一些標榜連結外國的ETF，本質上只是連結有投資該國的台灣企業，所以本質上跟台股有關係，所以當台股出現劇烈修正，這些號稱連結外國的ETF還是會有賣壓。

【群組限定，私房教學】客觀分析與主觀結論  
不論你是看總體經濟、產業趨勢、基本分析、籌碼分析、技術分析…，在專業能力充足以及正確使用的前提之下，都是客觀的分析工具，因為都是應用這些工具呈現投資人的訊息。  
但是使用各種工具客觀分析之後，得到的結論就會變成是主觀的，因為那是你個人研究後的結果，看多看空、目標漲幅、趨勢強弱…，並不是說主觀是錯誤，而是結論的產生過程，必然會成為主觀，因為是你獨有的。  
說到這邊，大家應該慢慢會意識到問題了，這個主觀的結論，對我而言會因為原先的客觀分析工具本身就是動態，每天都會產生變化，會同步很自然地做更新，但是假如你只是接受了某人的推論結果，卻不知道他是如何推論，當外在條件變化之後，原本提出觀點的人其實早就更新看法，但你還是會一直停留在他的原始觀點。  
一個分析過程是從申論題，縮小範圍到選擇題，再變成是非題，最終得到結論。你只當單向受眾，何時才會改變和更正，必須很依賴的等到提供你觀點的人更正，你才知道要更正。這就是為什麼你必須要自己會投資工具和技能，因為買賣跟交易的答案，來自獨一無二的你，只有你可以幫自己打造出屬於自己的交易模式。  

有時看這個種景氣燈號，除了有時間落後的問題之外，還要留意燈號成分，表現上是藍燈，數值持續墊高，好像是好事，但仔細一看綜合內容，其實9個觀察項目中，除了股價指數改善(紅色框框)，其實實質產業都沒變(藍色框框)。換言之，股市因為資金面，自己漲自己的，但實質景氣和需求沒有改善。
