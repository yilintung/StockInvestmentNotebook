### 【K線捕手盤後快訊】單日大漲，反壓仍多  
今天加權指數開高走高，終場上漲203點，收在16557點，成交量也溫和放大到2867億，台幣也稍微升值，不過究竟有沒有真正由空翻多？不能只看一天的漲跌。  
先從趨勢、位階、型態來做觀察，短線雖然轉強，但中期還是在偏空方的箱型整理，位階屬於中高，型態上仍有盤頭的風險。  
再來是支撐、壓力、轉折，先前提到16400是短線的關鍵轉折，今天雖然收盤暫時站上，但是還沒站穩，必須符合漲幅、時間、量能，也就是上漲需要300點以上，站穩3-5天，成交量在10日均量之上，  
所以目前還言之過早，而且即使真的站上16400，也只是從短中期偏空，轉為以16750箱型整理，尚未完全化解危機，不要因為一天大漲就過度樂觀，慎防一日行情，所以再多觀察幾日，不用急著增加資金比重。  
從價量累計圖觀察，仍可看到層層反壓，籌碼面也只是稍微偏多，但尚未真正轉強，而最關鍵的是個股表現，今天上漲的大多為反彈，或是原先維持多方個股，所以並沒有因此改善選股難度。 (出處：K線捕手楊忠憲粉絲團)  
  

另外有兩個非正規的分析，首先是今天市場普遍認為，是因為美國政府關門危機以及蘋果新機利空，兩大利空問題解除，但如果真是如此，那場只是找題材去拉抬，首先美國政府最後本來就不會「真正關門」，只是政黨角力而已，本質上就是假議題。而蘋果新機散熱問題，也一定會解決，但不會因為這樣就賣的好。  
此外，今天由於亞股普遍休市，所以對台股干擾程度相對低，甚至部分資金會挪移到台股，所以外在條件有利拉抬，但如剛剛解盤說的，慎防一日行情，還是要多觀察幾天表現才算數。
