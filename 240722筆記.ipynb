【我用K線寫日記】  
對於點閱真的是很平常心看待，畢竟10多年前開始在網路上分享，就是把臉書當成個人的交易日記寫，即使沒有人看，照樣可以寫，從最初50個人、100個人開始分享…  
影音解盤是動態的，會比文字清楚很多，而且只有10分鐘，耽誤不了太多時間，但對投資多少會有幫助，就像這一波下跌，即使沒有完全提前或同步紀律減碼，至少不會追在高點。只是很可惜的，如果從臉書跟群組總人數回推，每天願意完整看文字解盤大約只有15%，每周願意花10分鐘看完影音解盤的人低於5%，依照一般隨機的統計數據，股市中賺錢的投資人大約是在10%上下，我想也符合這個比例。  
當然這有一部份原因是影音製作太過於陽春，沒有太精美的美編或是後製，雖然叫做自媒體，其實很多頻道都已經是由團隊經營，雖然你們看到只有短短10分鐘內容，但我準備資料、打字、繪圖、錄製都要接近2個小時，沒有多餘的力氣再去加工，但我想反正是寫給自己看的，單純的內容這樣就足夠了。  
就如同這段文字，我相信會看的人就是會看，不會看的還是不會看，只是正因為這些細微點點滴滴的累積，決定了日後賺賠的差異。我寫這麼多只有一個結論，你絕對可以不看，這是個人自由，我也不能逼你，但千萬不要完全都沒看又跑來問我，已經重複說過很多次的重點，這樣分享的人會很累。
