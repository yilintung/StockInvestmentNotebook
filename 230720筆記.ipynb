### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線續弱，測試支撐
加權指數繼前天週二下跌106點之後，昨天週三10點過後持續走弱，終場再跌111點，收在17116點，成交量4335億，法人連2天減碼，主力指標轉弱，融資續增，短線籌碼偏空，顯見周一提到的價量累計圖，長天期的大量區間，上檔壓力重重。  
周末提到的幾個重點觀察，首先是類股指數，原先多方類股只有3類，但本周也都從高檔轉弱，而盤整與偏空的類股有25類，又更加弱勢，多數也呈現續跌，類股結構沒有改善。  
上漲最大動能的台幣，也到了年線、季線、前高，三個複合式壓力區，由升轉貶。原先一路飆漲的AI概念股也普遍轉弱，有人一路上都看著它漲，所以現在拉回一定想要出手，但是這就要回到前兩周說的，拉回找買點這件事情。  
請注意拉回找買點這個詞，是由兩個行為組成的，一個是找拉回的股票，一個是找買點。閱讀到這邊，應該有看到重點，找買點才是關鍵，不是要拉回就買，如果任何一檔股票拉回就買，那就會變成攤平交易，這是非常危險的。  
所以拉回找買點的是意思是，尋找仍在多方結構的個股，出現漲多拉回的現象，而且在支撐位置出現「多方轉折的買訊」，這個時候才能出手，假如已經出現高檔爆量長黑或長上影線的出貨現象，甚至已經正式形成空頭走勢，每逢拉回就買、拉回就買、拉回就買…，這樣最後會發生悲劇。  
最後，從趨勢型態以及外資期現貨觀察，目前只是短線遇壓轉弱，中期還是維持箱型區間，尚未全面翻空，極短線在17050前後有一個小支撐，不過指數沒止穩之前，手中個股大概就是盯緊支撐，退出後可以暫時保留現金，採取漸進式的減碼。 (出處：K線捕手楊忠憲粉絲團)

【[把它當小說看就好03-小學就財富自由](https://tinyurl.com/bc7j467d)】  
開始看第三集之前，我想再提醒一下各位，其實這個標題只是隨意下的，不過後來也覺得下的蠻好的，我真的希望大家把它當小說看，不要當成太嚴肅的教科書或是學術論文，因為當中的內容只是我的生活體驗，但不代表是嚴謹正確的觀念或想法，或是能百分之百套用在你的身上。  
上期講完賭博電玩累積的一點交易經驗值，這一期想聊一下投資理財的真正開始-儲蓄，時間軸一樣停留在幼稚園跟小學階段，在1980年代的時候，當時多數大人不大懂，小孩更不可能會懂投資理財，不過儲蓄這個觀念倒是普遍都會有，基本上小時候家庭經濟狀況算是小康之上，而且家裡有給零用錢的習慣，然後我真的是很會存錢，有零錢就存，一般的小豬撲滿很快就滿了，所以父親曾經用一段很粗的竹子，做了一個存錢筒，直徑應該有十多公分，長度有90公分，中間一段竹節，開始了我的儲蓄之路。  
反正小孩對消費沒甚麼觀念，基本的生活需求或是玩具，家裡也都會供應，所以很自然地就會把零錢存到竹筒裡，不過時間真的太久遠，我已經忘了到底存多久，最後真的把竹筒存得滿滿的。大家可以想像，那一個存錢筒有多沉，小孩子真的要用抱的才拿得動，最後把竹筒打破結算零錢，印象中應該有2萬左右，然後把那筆錢存到了郵局，所以和很多人一樣，郵局人生中第一個金融戶頭，算是我正式進入投資理財的開始。  
你一定會覺得，小學生就有2萬多算是很有錢，當然是沒錯，對啦！我小學就已經財富自由了，但這邊不得不聊一下通膨的可怕，話說1985年前後，我印象很深刻，一碗陽春麵是10元，滷蛋是3元，可是到了現在一碗陽春麵基本30元起跳，甚至台北市可能會到40元以上，蛋就不用說了，前陣子鬧得沸沸揚揚，大家應該記憶猶新。重點是這意味著甚麼，如果你甚麼都沒做只是放銀行，那個2萬現在的消費力，可能只剩下5-7千，有沒有覺得很可怕。
