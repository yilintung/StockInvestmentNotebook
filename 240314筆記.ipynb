我知道最近ETF很熱，很多同學都問我有甚麼看法，但是老同學就知道我只做股票，加上生性比較懶惰，所以就沒鑽研那麼多商品，因此不做融資券、不做ETF(極少)、不做期貨、不做選擇權、不做權証、不做美股、不做陸港股、不做歐股、不做外匯、不做原物料、不做比特幣…。  
股市中的商品無窮無盡，金融市場的錢永遠賺不完，光是台股快2千檔，只要會選股，已經很夠我買了。既然沒深入鑽研其他商品，就不好意思班門弄斧，反正網路上的分享跟評論很多，隨便搜尋都有幾百篇，至於那些專家說的好、說得明白、說的正確…，大家一定要有自己的判斷力就是了，祝大家投資順利。  
  
「[老師](https://www.facebook.com/kevinyang20180801/posts/pfbid0Ht4eKTyPVXzhfBCx6qRbXBd2EJfKZMjrMkJmjD2fZ4JJctbsifSpmVLWxrALHkcDl)，我上過您的課，看懂您說的技術面由多翻空現象，也大致上能理解，只是心中卻會想著，指數應或許不會修正太多，手上的好公司股價就應該挺得住，所以一路抱到現在，但手上很多股票，說真的有的個股已經跌很多，接下來該怎麼辦呢？還有下次遇到要怎麼克服心理障礙？」  
1.先止血再研究：基本上會看著股票下跌而沒有作為，就是沒有做好明確停損條件的設定，或是設定之後無法紀律執行，但這真的說來話長，真的要寫可以寫1萬字以上，有興趣可以參考專買黑買股1。  
2.誤判行情走勢：趨勢翻空都會有領先訊號和同時訊號，必須學會如何判斷，如果自己不會看，看我每天解盤其實也可以輔助，看不懂推理過程，至少要看結論，領先訊號出現時就應該預防性減碼，同時訊號出現就應該降到最低甚至空手。  
3.仍然心存僥倖：看到指數或股票下跌，過程中的訊號和變化，假設都看得懂，但是心存僥倖，總覺得這一次應該不會這麼慘吧！這是就對這些訊號的信仰還不夠，而信仰來自於經驗值和交易次數，這一次不相信，下一次不相信，多看幾次就相信了，所以做個股交易的投資人，一年至少要有30檔以上的交易案例，否則無法快速累積經驗值，我自己去年10月到現在，就已經交易接近30檔。  
4.為漲跌找理由：當指數或股票下跌時，不願意面對下跌的事實，總是想要在各種新聞、財務數據、小道消息中，尋找上漲的希望，在盤中應該做的是紀律交易，倘若還在為上漲和下跌找理由是最危險的，因為有上過課的同學就知道股票漲跌不是單一變因，而是複合式的變因，即使找到一個上漲的理由，可能有另外99個會下跌的理由沒發現。  
5.永遠有下一次：最後，請敞開心胸容許自己犯錯，這一次虧錢沒關係，誠實的面對自己，把交易紀錄寫下來，不要再犯錯，這樣這一次的虧損就有價值，不然下一次遇到同樣的問題，還是會犯同樣的錯，那這樣這一次的虧損就真的變成沉沒成本了。
