📒【[用「4個獨立帳戶」，簡單又有效的資金管理法](https://www.facebook.com/kevinyang20180801/posts/pfbid0dLu6U5yaoktu43n9Nri5DzvEqouXSUTcVw3w2NH3hvpk4PhkhZc4koxxLEG9enEXl)】  
（不記帳也能控管得一清二楚！）  
💬 很多人問我：「你還會記帳嗎？」  
老實說，當收入和資金到達一定水位後，我就不太做日常流水帳了。  
不是不管帳，而是我改用了 更高效、更清楚的方式來控管資金 —— 就是分成4個帳戶，各司其職。  
🏠 ① 生活帳戶｜每年編列預算，一次轉入  
我會在每年初就設定「全家一年開銷」的總額，像是房貸、水電、保險、育兒、日常開銷等，一筆轉入這個戶頭。  
🔄 所有家庭開支都從這裡出，半年或年底只要看餘額，就知道有沒有超支，管理起來超簡單。  
📈 ② 短期投資帳戶｜本金固定，獲利轉出  
我會設定一筆本金放在股票帳戶中，交易週期以2週～1.5個月為主。  
✅ 每次獲利達標就轉回儲蓄帳戶，本金維持不變，損失也不再加碼，只靠操作回到原點。  
🎯 這樣做的好處是：獲利成果很清楚，本金風險也被鎖定。  
🧭 ③ 長期投資帳戶｜等待大機會，絕不追高  
這個帳戶用來部署半年以上的長期部位。  
目前市場整體位階偏高，所以我暫時都放現金。  
📉 只有等到大修正時（例如台股有3000點，甚至5000點以上的修正時）才會慢慢分批進場。  
🔍 建議把短期與長期帳戶分開，可以讓你看盤時更冷靜，投資思維也不會互相干擾。  
💰 ④ 儲蓄帳戶｜低調沉穩的水庫帳戶  
主要做定存或高流動性資金停泊。  
🌊 功能就像個「資金水庫」，生活開支從這邊撥出，短期獲利也回存這邊，長期投資暫時不用的錢也停在這裡。  
🛟 可攻可守，安心又靈活。  
🎓 給年輕人的提醒  
如果你還在學生階段或剛出社會，📋 建議還是從「記帳」開始，建立資金敏感度與自律消費的基礎，因為 ✅資金管理力 = 投資成功的根本功夫！  
📌 有錢不是靠會賺，而是靠會分配。  
用這4個帳戶，不記帳也能讓你的金錢流向，一目了然！
