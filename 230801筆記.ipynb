上周提到兩個重點，第一是7/24-7/28這周還不至於有風險，但本周風險就會提高，第二則是貼了一張和同學的對話，都覺得已經選不太到符合條件的股票。昨天是本周的第一天，就出現震撼教育，加權指數開盤雖然大幅補漲，最高漲點來到170點，但是搭配台幣貶值，以及一線的AI概念股緯創、廣達、技嘉等…，出現跌停鎖死，盤中最大跌點來到211點，單日高低差超過350點。  
成交量放大到4923億，外資賣超88億，盤後未平倉口數首次轉為3786口空單，融資減幅不大，而原先提醒的「多技術指標二次背離」，KD和MACD也已經出現這樣的現象，短線空方氣盛。  
不過投資人更為關心的，這究竟是一日現象，漲多拉回整理？還是高檔翻空起跌的警訊？一個趨勢要由多翻空，不論技術面跟籌碼面，都是有節奏和演進的，最早出現的領先訊號，技術面有均線扣抵、20日乖離、高檔背離、爆量長黑或長上影線，而籌碼面有外資現貨有買有賣、外資期貨空單5千口以上、融資續增，而這些領先訊號，雖然不完全，但也有8成具備了。  
那麼接下來就要觀察同時訊號，技術面出現3日不創新高、5日線跌破且下彎、10日線跌破且下彎、KD死亡交叉，籌碼面則是外資現貨積極賣、外資期貨空單萬口以上、融資續增，這個部分就尚未完全充分，只有一半符合，不過3到5個交易日，應該就會漸趨明朗。  
但即使出現領先訊號和同時訊號，在沒有跌破關鍵支撐之前，盤勢仍可以視為漲多拉回，或者中性一點的說，還是大箱型區間整理，而確認訊號比較簡單的方法也可以觀察，中期(紅色)和長期(藍色)兩條上升趨勢線。  
至於AI概念股一定有當初沒有跟上車的投資人，正在思考要不要拉回找買點，這部分可能暫時要先停看聽，要看修正的狀況和幅度再做評估，尚未止穩之前不用急著搶短，拉回找買點，不是拉回就買，是拉回出現買訊，搭配多頭趨勢尚未改變。未來或許還有機會出現上漲，但已經過了無腦操作階段，再來真的是給高手做的，功力不夠看戲就好。 (出處：K線捕手楊忠憲粉絲團)  


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import mplfinance as mpf
import numpy as np
```

```python
##### 使用2022年9月12日到2023年7月31日的周K線資料繪製 #####
import pandas as pd

# 載入從「好神通PLUS」輸出的Excel檔
df = pd.read_excel('./data/SINO_TSE_230731.xlsx')

# 保存K棒的基本資訊(開、高、低、收、量)
df_k_line = df.drop(columns=['SMA4','SMA13','MA6','MA12', 'DIF6-12','MACD2','OSC','K(6,3)','D(6,3)'])

# 保存價的均線
df_sma = df.drop(columns=['開盤價','最高價','最低價','收盤價','成交量','MA6','MA12', 'DIF6-12','MACD2','OSC','K(6,3)','D(6,3)']) 

# 將K棒的Columns的名稱由中文改為英文
df_k_line = df_k_line.rename(columns={'時間':'Date','開盤價':'Open','最高價':'High','最低價':'Low','收盤價':'Close','成交量':'Volume'})

# 將價的均線的Columns的名稱由中文改為英文
df_sma = df_sma.rename(columns={'時間':'Date'})

# 將K棒的Date設為Index
df_k_line.set_index(pd.to_datetime(df_k_line['Date'],format='%Y-%m-%d'),inplace=True)
df_k_line = df_k_line.drop(columns=['Date'])

# 將價的均線的Date設為Index
df_sma.set_index(pd.to_datetime(df_sma['Date'],format='%Y-%m-%d'),inplace=True)
df_sma = df_sma.drop(columns=['Date'])
```

```python
def date_to_index(df_in,date_in) :
    number_array = df_in.index == date_in
    idx = 0
    for number_index in number_array :
        if number_index == True :
            break
        idx = idx + 1
    return idx
```

```python
# 人工設定上升趨勢線(藍)
trendline_up1_x = [date_to_index(df_k_line,'2022-10-24'),date_to_index(df_k_line,'2023-05-15')]
trendline_up1_y = [df_k_line.loc['2022-10-24']['Close'],df_k_line.loc['2023-05-15']['Open']]
slope,intercept = np.polyfit(trendline_up1_x,trendline_up1_y,1)
trendline_up1_start_date = '2022-10-24'
trendline_up1_start_price = slope * date_to_index(df_k_line,'2022-10-24') + intercept
trendline_up1_end_date = '2023-07-31'
trendline_up1_end_price = slope * date_to_index(df_k_line,'2023-07-31') + intercept 

# 人工設定上升趨勢線(紅)
trendline_up2_x = [date_to_index(df_k_line,'2023-05-15'),date_to_index(df_k_line,'2023-07-10')]
trendline_up2_y = [df_k_line.loc['2023-05-15']['Open'],df_k_line.loc['2023-07-10']['Open']]
slope,intercept = np.polyfit(trendline_up2_x,trendline_up2_y,1)
trendline_up2_start_date = '2023-05-15'
trendline_up2_start_price = slope * date_to_index(df_k_line,'2023-05-15') + intercept
trendline_up2_end_date = '2023-07-31'
trendline_up2_end_price = slope * date_to_index(df_k_line,'2023-07-31') + intercept 
```

```python
# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定
seq_of_seq_of_points=[
    [(trendline_up1_start_date,trendline_up1_start_price),(trendline_up1_end_date,trendline_up1_end_price)],
    [(trendline_up2_start_date,trendline_up2_start_price),(trendline_up2_end_date,trendline_up2_end_price)]
                     ]

# 設定均線
added_plots={
    "SMA4": mpf.make_addplot(df_sma['SMA4'],width=0.4,color='xkcd:maroon'),
    "SMA13": mpf.make_addplot(df_sma['SMA13'],width=0.8,color='xkcd:sky blue')
            }

# 繪出K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=(1.4,1.4), colors=('xkcd:blue','xkcd:red'), alpha=0.6))
mpf.plot(df_k_line,**kwargs)
```
