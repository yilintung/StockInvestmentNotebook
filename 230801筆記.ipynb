### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】創高拉回，守支撐線 
上周提到兩個重點，第一是7/24-7/28這周還不至於有風險，但本周風險就會提高，第二則是貼了一張和同學的對話，都覺得已經選不太到符合條件的股票。昨天是本周的第一天，就出現震撼教育，加權指數開盤雖然大幅補漲，最高漲點來到170點，但是搭配台幣貶值，以及一線的AI概念股緯創、廣達、技嘉等…，出現跌停鎖死，盤中最大跌點來到211點，單日高低差超過350點。  
成交量放大到4923億，外資賣超88億，盤後未平倉口數首次轉為3786口空單，融資減幅不大，而原先提醒的「多技術指標二次背離」，KD和MACD也已經出現這樣的現象，短線空方氣盛。  
不過投資人更為關心的，這究竟是一日現象，漲多拉回整理？還是高檔翻空起跌的警訊？一個趨勢要由多翻空，不論技術面跟籌碼面，都是有節奏和演進的，最早出現的領先訊號，技術面有均線扣抵、20日乖離、高檔背離、爆量長黑或長上影線，而籌碼面有外資現貨有買有賣、外資期貨空單5千口以上、融資續增，而這些領先訊號，雖然不完全，但也有8成具備了。  
那麼接下來就要觀察同時訊號，技術面出現3日不創新高、5日線跌破且下彎、10日線跌破且下彎、KD死亡交叉，籌碼面則是外資現貨積極賣、外資期貨空單萬口以上、融資續增，這個部分就尚未完全充分，只有一半符合，不過3到5個交易日，應該就會漸趨明朗。  
但即使出現領先訊號和同時訊號，在沒有跌破關鍵支撐之前，盤勢仍可以視為漲多拉回，或者中性一點的說，還是大箱型區間整理，而確認訊號比較簡單的方法也可以觀察，中期(紅色)和長期(藍色)兩條上升趨勢線。  
至於AI概念股一定有當初沒有跟上車的投資人，正在思考要不要拉回找買點，這部分可能暫時要先停看聽，要看修正的狀況和幅度再做評估，尚未止穩之前不用急著搶短，拉回找買點，不是拉回就買，是拉回出現買訊，搭配多頭趨勢尚未改變。未來或許還有機會出現上漲，但已經過了無腦操作階段，再來真的是給高手做的，功力不夠看戲就好。 (出處：K線捕手楊忠憲粉絲團)
