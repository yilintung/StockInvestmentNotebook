【[學習投資的致勝關鍵](https://www.facebook.com/kevinyang20180801/posts/pfbid02yD4rBmepJoWGtDpEzhB54votdtFiFFgDBHPDQxYcqNoabmv4Q6pGcwHhCiGTyeW5l)】  
這麼多年的教學和個人的學習經驗，我發現學習投資進步快速的人，當中或有個體差異，舉凡年齡、性別、職業、學歷、資金…，但普遍都具備以下的共同特質。  
強烈企圖：  
我常常開玩笑能不能學會投資，關鍵在於你有多愛錢，當然可以用稍微文雅一點的說法，就是追求財富的渴望，必須有非常強烈和旺盛的企圖心，才能有棄而不捨的努力，不會遇到一些失敗就半途而廢，反而可以越挫越勇，意志堅定地持續向前。  
沉浸學習：  
沒有人一開始就會投資，肯定都要經過學習，學習方式有很多種，看書、網路、上課…，而且必須複習和消化，有時候可能要循環2-3次，才會成為自己的血肉，不只是專注學習，而是要沉浸學習，如同老僧入定一樣，做到旁邊沒有事物可以干擾你，那種程度的專注。更重要的必須是有系統、有效率的學習，絕對不是很隨興的滑手機、加入自救會社團、看投資八卦文章、轉電視遙控器…，那可以療癒自己，但對投資本身沒有幫助。  
心得筆記：  
每一個學習投資的人，一定會有屬於自己的筆記，我自己將筆記分成3類，第一種是學習筆記，就是把看書上課吸收的知識，整理成容易複習的摘要跟重點，第二種是解盤筆記，對盤是跟個股的觀點分析跟現象，完整地紀錄下來，可以有效精進解盤跟選股。第三種是實戰交易筆記，人的記憶是最不可靠的，只有詳實地把每一次的交易過程，買賣策略、賺賠、得失，詳實記錄下來，才會成為日後交易的養分。  
高頻實戰：  
實戰是檢驗真理唯一的方法，所以學習一段時間之後，一定要進入交易，不只可以驗證所學知識，更能快速累積經驗值，並且找到適合自己的商品、週期和工具。我的經驗是大約100檔，交易能夠累積100檔個股之後，學習效率會大幅提升(某檔股票試單、加碼、買足、減碼、停利或停損，這樣才算是一檔)。  
或許有人會覺得上班已經很累了，哪來這麼多體力跟精神，但投資從來就不是件容易的事情，如果輕輕鬆鬆就能會，那地球上就沒有窮人了，池本克之的《贏在下班後》：休假日做什麼，決定你的收入高低！所以你的時間花在哪裡，為來的成就就在哪裡。共勉之~
