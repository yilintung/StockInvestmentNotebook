* 「量化技術分析系統」架構：  
   - 目前系統只實作『解盤』部分，並且只是配合產出「股票投資筆記」Jupyter Notebook所需內容。  
   - 『解盤』的架構是依據「[專買黑馬股3：K線捕手楊忠憲教你用1張表逮飆股](https://www.books.com.tw/products/0011010836)」第三章的內容。但技術分析工具並不是照單全收而是做了挑選：解盤的前半段都是使用了量化技術分析工具為之（詳情後述），後半段【整體評價】則是使用人工智慧（[多模態模型](https://zh.wikipedia.org/zh-tw/%E5%A4%9A%E6%A8%A1%E6%80%81%E5%AD%A6%E4%B9%A0#%E5%A4%9A%E6%A8%A1%E6%80%81Transformer%E6%A8%A1%E5%9E%8B)）來產生分析結論：  
        
   - 各項量化技術分析工具分別說明於下：  
     * Ｋ線／Ｋ棒  
       使用[talib](https://github.com/TA-Lib/ta-lib-python)進行Ｋ線型態識別（[Pattern Recognition](https://github.com/TA-Lib/ta-lib-python?tab=readme-ov-file#pattern-recognition)）。  
     * Ｋ線圖  
       進行中期與長期的「[量化位階評價](https://vocus.cc/article/62ada936fd89780001fe6208)」。  
     * 成交量  
       價量關係（大量、爆量、天量），參照〔價量關係_230317_筆記.pptx〕。  
     * 型態  
       參照[TradingView](https://tw.tradingview.com/)的[圖表型態](https://tw.tradingview.com/support/solutions/43000706927/)。另外整合底型反轉交易策略的機制，參照〔第一堂課-底型反轉_240809_筆記.pptx〕。  
     * 移動平均線  
       參照「[專買黑馬股3：K線捕手楊忠憲教你用1張表逮飆股](https://www.books.com.tw/products/0011010836)」第三章 第89頁 表5。  
     * ＫＤ指標  
       判斷黃金交叉、死亡交叉、超買與超賣，確認短期多空。  
     * ＭＡＣＤ指標  
       中期多空判斷。  
     * 周Ｋ－周ＫＤ指標  
       判斷黃金交叉、死亡交叉、超買與超賣，確認中期多空。  
   - 整體評價

* 「股票投資筆記」Jupyter Notebook與「K線追擊週報」日報於每個交易日之工作時程：
   - 開盤前(8:00) ～ 開盤(9:00)  
     * 以**剪報**方式將股票老師的盤前快訊與財經新聞紀錄至日報中。  
     * 建立當日的股票投資筆記Jupyter Notebook。先列出今日要進行的工作項目，且若有盤前快訊的圖表先做繪製。  
  - 開盤後先進行交易，空檔時紀錄「K線追擊週報」LINE群組資訊到日報中。  
  - 收盤(13:30)後記錄K線追擊週報LINE群組資訊到日報中。  
  - 14:30後選出觀察股用以記錄於股票投資筆記Jupyter Notebook中。  
  - 15:00後在日報中加入籌碼快訊。週二、周四時就可進行日報完稿並上傳至MEGA雲端硬碟。  
  - 17:30後更新技術分析資料庫，並進行盤後解盤。  
  - 20:00周一、周三、周五紀錄K線追擊週報影音連結至日報，進行日報完稿並上傳至MEGA雲端硬碟。  
  - 21:30後進行觀察股籌碼確認，並總結本日工作進度後將股票投資筆記Jupyter Notebook完稿並發布至GitGub與上傳MEGA雲端硬碟。
