### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】單日大漲，反壓仍多  
先前曾提到16400是關鍵的支撐和壓力，容易在這個點位頻繁來回測試，昨天開高走高，終場上漲203點，收在16557點，成交量也溫和放大到2867億，台幣也微幅回升，不過究竟有沒有真正由空翻多？先從趨勢、位階、型態來做觀察，短線雖然轉強，但中期還是在偏空方的箱型整理，位階屬於中高，型態上仍有盤頭的風險。  
再來是支撐、壓力、轉折，先前提到16400是短線的關鍵轉折，雖然收盤暫時站上，但是還沒站穩，必須符合漲幅、時間、量能，也就是上漲需要300點以上，站穩3-5天，成交量在10日均量之上，所以僅就昨日的表現還言之過早，而且即使真的站上16400，也只是從短中期偏空，轉為以16750箱型整理，從價量累計圖觀察，仍可看到層層反壓，籌碼面也只是稍微偏多，但尚未真正轉強。  
近4個月以來，其實消息面相當凌亂，你會發現對未來的行情，有很多專家大師們看多，但看空的財經達人也不在少數，也因為如此，行情才維持著以16750為中軸的大箱型橫盤整理，呈現多空拉鋸，只是目前是在中軸線的下半區間整理，而且假設一年有240個交易日，多頭空頭盤整平均分配，本來就有1/3的時間，會出現盤整的現象，這是無法避免的，不過一旦市場有集體共識，趨勢就會走得很快。  
短線上站回16400，多方當然可以稍微鬆一口氣，但趨勢尚未改變，中期盤頭危機也沒有解除，只少必須先站上16750才算是偏多方的整理。不要因為一天大漲就過度樂觀，尚未完全化解危機，慎防一日行情，所以再多觀察幾日，讓趨勢更加明朗。藝高人膽大的高手投資人，想趁機順勢搶一下短多，自然是可以，但以短多視之，資金不宜過度放大，因為最關鍵的還是個股表現，昨日指數上漲的個股大多為反彈，或是原先維持多方個股，所以並沒有因此改善選股難度。 (出處：K線捕手楊忠憲粉絲團)
