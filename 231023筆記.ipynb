【[盯盤，不是盯一個數字](https://www.facebook.com/kevinyang20180801/posts/pfbid0znnor9ZafKFU9V3gbGdRtWwi3mqds4Umk9Wt1AazZkXJiDeFbVHRr496qXmzeNiYl)】  
支撐和壓力，並非是一個固定價位，她可能是型態、可能是指標、可能是大量區間，最重要的，她是一個浮動數字，我解盤時只是為了方便大家記憶，所以會去寫出一個數字。既然她是一個動態現象，你就要給予她一些寬鬆的震盪空間，甚至需要跟著走勢做微調。  
而且突破和跌破，還需要搭配其他的要件。比如我說壓力是100，不是有一天漲到101，她就叫突破，至少要到103才算數，而且要站穩3天，還需要帶大量，甚至籌碼也要夠強，這樣才是真的突破。所以不是跟你說100，你就死盯著100，這樣就失去技術分析的意義了。  
而且所謂盯盤，是很單純的事情，就是做好兩件事情。第一件事，昨天有沒有先選好股票，然後觀察這些股票今天是否有發動點/買點。第二件事，如果不打算再買股票，或是不打算提高持股資金比重，那就關注好目前持有的個股支撐或目標價，規畫好離場策略。  
很多散戶習慣盤中東想西想、太多雜念，其實盤中就是專注做好這些事情就可以，因為大多數的研究和功課，都是在盤前甚至昨晚，就要先做好準備，盤中只要專注在交易，不要想太多有的沒有的。  
  
所以很多時候，看到指數或個股，在盤中漲跌就會腦袋空白，都是因為前一天晚上，沒有先做好準備功課，盤中一出現漲跌，自然會不知所措
