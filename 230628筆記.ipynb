### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】短線支撐，醞釀上漲  
昨天加權指數持續走弱，下跌171點，成交量3002億，收在16887點，雖然盤勢不強小跌破20日線，但賣壓已經小很多，多數個股都能維持小紅小黑居多。而且昨天提到「測完20日線支撐，下半周就有機會止穩上漲」，也剛好來到重要支撐，所以只要紅色位置能守穩，一樣維持會出現短線上漲的可能，但上漲是反彈還是回升，還有漲幅大小，就要看到時候的量能跟籌碼。  
所以雖然認為中期修正機率高，但目前短線還屬於多頭回檔，多頭看支撐不看壓力，所以何時翻空也很簡單，等到有一天看到我的解盤，已經改成關注壓力而不是支撐，就知道趨勢已經反轉。趨勢的反轉不會是一天，是從多頭力竭，極短線轉弱，短線翻空，中期翻空，長期翻空，每一種階段的現象轉折工具跟跌幅都不同，現在其實只到第三階段，所以暫時修正空間不大，但倘若後續的訊號再出現，才會代表著更大的跌幅風險。  
不過指數只是控制風險跟調節資金的工具，與個股交易沒有直接關係，所以只要資金比重控制好，把風險鎖住了，指數漲跌輕鬆看待即可，而且周末就說過，「短線上就算跌也是緩跌，對個股干擾會比較小」，我自己目前還是低資金比重，但個股持續作多，還沒有到需要全面一次性出清的狀態，但如果你覺得指數的波動，會影響持股情緒或是會很有壓力，表示持股太滿了，應該要減一些了。 (出處：K線捕手楊忠憲粉絲團)

【[投資和賭博](https://tinyurl.com/yvbb8a7n)】  
賭跟投資很像，甚至可以說他們是同父異母的兄弟，同父是因為兩者都需要運氣，所以很多人會把他們兩者混淆。  
但異母是異在哪邊呢？就是專業技巧跟心理素質的比例了。投資是九成專業靠技巧，一成靠運氣，但賭博剛好是顛倒的。  
你一定會馬上反駁，有些人確實能做到夠十賭九勝。沒錯，如果你的賭技和心理素質能超越運氣，長期下來多賺少賠，所以賭博對進出國際型賭場，真正的專業的職業賭徒而言，就反而是投資了。  
但如果你投資專業跟技巧很弱，常常大賠小賺，那你所謂的投資其實跟賭博沒兩樣。
