* 實驗：模擬手工方式，由兩個轉折點中畫出一條趨勢線

```python
import matplotlib.pyplot as plt
import matplotlib
import mplfinance as mpf
import numpy as np

##### 使用2022年3月11日到2023年7月21日之日線圖資料 #####

# 載入從「好神通PLUS」輸出的Excel檔
df = pd.read_excel('./data/SINO_1515_230721.xlsx')

# 保存K棒的基本資訊(開、高、低、收、量)
df_k_line = df.drop(columns=['SMA5','SMA10','SMA20','SMA60','SMA120','SMA240','MA5','MA10'])

# 保存價的均線
df_sma = df.drop(columns=['開盤價','最高價','最低價','收盤價','成交量','MA5','MA10']) 

# 將K棒的Columns的名稱由中文改為英文
df_k_line = df_k_line.rename(columns={'時間':'Date','開盤價':'Open','最高價':'High','最低價':'Low','收盤價':'Close','成交量':'Volume'})

# 將價的均線的Columns的名稱由中文改為英文
df_sma = df_sma.rename(columns={'時間':'Date'})

# 設定今天日期
today_date = df_k_line.iloc[-1]['Date'].strftime("%Y-%m-%d")

# 將K棒的Date設為Index
df_k_line.set_index(df_k_line['Date'],inplace=True)
df_k_line = df_k_line.drop(columns=['Date'])

# 將價的均線的Date設為Index
df_sma.set_index(df_sma['Date'],inplace=True)
df_sma = df_sma.drop(columns=['Date'])
```

```python
import myutils

# 尋找轉折點
_,_,max_min = myutils.FindingTurningPoints(df_k_line,'close',order=10,real_body=True)
```

```python
##### 試驗：趨勢線程序 #####

turning_points     = max_min
turning_points_len = len(max_min)
trendline_list     = []

# 轉折點連接：高點對高點，低點對低點
for turning_point_idx in range(0,turning_points_len) :
    tl_start_idx  = -1 # 趨勢線起點的索引
    tl_end_idx    = -1 # 趨勢線終點的索引
    tl_type       = '' # 趨勢線類型 ('U' 上升趨勢線，'D'：下降趨勢線)
    for point_idx in range(turning_point_idx, turning_points_len) :
        idx = turning_points.iloc[point_idx].name
        # 連接股價高點與次高點形成下降趨勢線，連接股價低點與次低點形成上升趨勢線
        if turning_points.iloc[point_idx]['Type'] == 'HI' :
            typ = 'D'
        elif turning_points.iloc[point_idx]['Type'] == 'LO' :
            typ = 'U'
        if tl_start_idx == -1 :
            # 趨勢線起點
            tl_start_idx  = idx
            tl_type = typ
        elif tl_end_idx == -1 :
            if typ == tl_type :
                # 趨勢線終點
                tl_end_idx = idx
                x = [tl_start_idx,tl_end_idx]
                y = [turning_points.loc[tl_start_idx]['Price'],turning_points.loc[tl_end_idx]['Price']]
                m,c = np.polyfit(x,y,1)
                trendline_list.append((turning_points.loc[tl_start_idx]['Date'].strftime("%Y-%m-%d"),turning_points.loc[tl_start_idx]['Price'],
                                turning_points.loc[tl_end_idx]['Date'].strftime("%Y-%m-%d"),turning_points.loc[tl_end_idx]['Price'],
                                m,c))
                tl_end_idx = -1

# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定趨勢線
seq_of_seq_of_points = []
for trendline in trendline_list :
    seq_of_seq_of_points.append([(trendline[0],trendline[1]),(trendline[2],trendline[3])])

# 繪出K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10),alines=dict(alines=seq_of_seq_of_points, linewidths=0.4, colors='xkcd:azure', alpha=0.6))
mpf.plot(df_k_line,**kwargs)
```
