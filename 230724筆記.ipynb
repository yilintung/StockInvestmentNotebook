上週加權指數微幅創高，盤中最高一度來到17401，但是由於先前提到的，「多方條件若不夠紮實，還是容易再度出現創高拉回的現象」，創高過程中，原先提到的多方瑕疵都存在，所以確實周二開始出現大幅拉回整理，最近的4個交易日，也形成價跌量增、漲漲量縮的空方價量關係。  
籌碼面法人賣融資增，且外資未平倉多單無法突破5千口，都在3千口以下徘徊，資金面台幣由升轉貶，動能不足。技術面KD和MACD出現雙指標單次背離，上一次6月15日的修正並沒有背離的現象，所以賣壓較小，如果出現背離後的修正，風險係數會增高許多，最後是周KD持續死亡交叉。  
未來觀察3個重點，首先是16593前低支撐，一旦跌破會形成中期頭部，會比先前迷你頭部壓力更大，其次是台幣前低31.4的支撐，能否有守，最後長期指標，下周一收月KD，慎防滑落80以下鈍化區。  
延續近日的策略，指數沒止穩之前，手中個股盯緊支撐即可，離場後可以暫時保留現金，採取漸進式的減碼，修正過程的股票不要急著低接，20日線下彎或是做頭的個股更是是避開。 (出處：K線捕手楊忠憲粉絲團)
